```python
text="""
Continuous electroproduction of formate via CO2 reduction on local symmetry-broken single-atom catalysts
Abstract
Atomic-level coordination engineering is an efficient strategy for tuning the catalytic performance of single-atom catalysts (SACs). However, their rational design has so far been plagued by the lack of a universal correlation between the coordination symmetry and catalytic properties. Herein, we synthesised planar-symmetry-broken CuN3 (PSB-CuN3) SACs through microwave heating for electrocatalytic CO2 reduction. Remarkably, the as-prepared catalysts exhibited a selectivity of 94.3% towards formate at âˆ’0.73â€‰V vs. RHE, surpassing the symmetrical CuN4 catalyst (72.4% at âˆ’0.93â€‰V vs. RHE). In a flow cell equipped with a PSB-CuN3 electrode, over 90% formate selectivity was maintained at an average current density of 94.4â€‰mAâ€‰cmâˆ’2 during 100â€‰h operation. By combining definitive structural identification with operando X-ray spectroscopy and theoretical calculations, we revealed that the intrinsic local symmetry breaking from planar D4h configuration induces an unconventional dsp hybridisation, and thus a strong correlation between the catalytic activity and microenvironment of metal centre (i.e., coordination number and distortion), with high preference for formate production in CuN3 moiety. The finding opens an avenue for designing efficient SACs with specific local symmetries for selective electrocatalysis.
Introduction
Formic acid has long been regarded as an important liquid feedstock for valuable chemicals synthesis and ideal hydrogen carrier material owing to its high volumetric (53.4â€‰gâ€‰Lâˆ’1) and moderate gravimetric (4.4â€‰wt%) hydrogen storage capacity in ambient conditions1,2. Producing formic acid via the electrochemical reduction of carbon dioxide (CO2) is therefore attracting widespread attention3,4,5. Considering both the thermodynamical and kinetic energy barriers, the energy consumption for electroreduction of CO2 to formic acid (
CO2+2H++2eâˆ’â†’HCOOH
, âˆ’0.19â€‰V vs. RHE) is relatively less compared with other products like CH3OH and CH4. In the past years, various metals (e.g., Cu, Sn, Bi) and their alloys have been prepared as electrocatalysts for electrocatalytic CO2 to formic acid6,7,8,9,10. Although significant progress has been made in reducing the overpotential for CO2 electroreduction to formic acid11,12,13,14, achieving the necessary faradaic efficiency (FE) and current density for practical applications remains a challenge15,16,17. Hence, there is a pressing demand to develop highly selective and energy-efficient catalysts for the electroreduction of CO2 to formic acid.
Carbon-supported single-atom catalysts (SACs), most of which usually adopt a prototypical MN4 moiety with local planar-like D4h symmetry, show great promise for accelerating the kinetics of heterogeneous catalysis owing to their unique physicochemical properties18,19,20. Though substantial progress has been made in activity improvement, recent studies reveal that the highly symmetrical MN4 architecture actually imposes serious limitations on the electronic configuration regulation of the active metal site21,22,23. To lift the confinement and better tune the electronic structures of the active sites, various coordination engineering strategies such as tuning the coordination species and the coordination number of the active centres, heteroatom substitution, neighbouring metal interactions are developed24,25,26,27. For example, the coordination-engineered M-Nx/NxLy (L=C, O, P, S) exhibits excellent performance in oxygen-involved electrocatalysis with a medium bond strength (494â€‰kJâ€‰molâˆ’1 for O=O bond)23,28,29,30,31,32,33, and great potentials for activating the molecules with high bond energy such as C=O in CO2 (750â€‰kJâ€‰molâˆ’1)34,35,36,37,38,39. Essentially, local symmetry breaking from the planar-like D4h via coordination engineering can redistribute the electronic (anti-)bonding states on specific d/s/p orbitals of the metal sites along with the energy scale and thereby maximumly regulate their hybridisation interaction in a rationally pre-selected local symmetry. This local symmetry manipulation could thus facilitate the adsorption and activation of CO2 molecular as well as reactive intermediates and optimise the reaction pathways for CO2 reduction reaction (CO2RR) with high activity and selectivity40. Unfortunately, except for the extrinsic addition of various axial ligands, there is no report about the intrinsic local symmetry breaking in M-Nx SACs for electrocatalytic CO2 reduction to date. Moreover, despite the extensive endeavours, the underlying mechanism that governs the structureâˆ’functionality relationship and the universal design principle to manipulate the local coordination symmetry remains unclear.
Herein, the local planar-symmetry-broken CuN3 (PSB-CuN3) SACs were synthesised by a facile and simple microwave process for electroreduction of CO2. The as-prepared PSB-CuN3 delivers a selectivity of 94.3% for electroreduction of CO2 into formate at a potential of âˆ’0.73â€‰V vs. RHE, which outperforms the CuN4 SACs with local planar D4h symmetry (PS-CuN4) with a maximum selectivity of 72.4% at a given potential of âˆ’0.93â€‰V vs. RHE. Impressively, the flow cell equipped with the PSB-CuN3 electrode exhibits over 90% formate selectivity at an average reduction current density of 94.4â€‰mAâ€‰cmâˆ’2 during 100â€‰h continuous operation at an applied potential of âˆ’0.95â€‰V vs. RHE. By combining definitive structural identification with ex-situ/operando X-ray spectroscopy characterisations and systematic theoretical calculations, we reveal that the intrinsic local symmetry breaking from the planar-like D4h to C2v configuration induces an unconventional dsp hybridisation. It causes the catalytic activity unconformable to the widely used d-band centre theory, but strongly correlated with the local environment of the metal centre (i.e., coordination number and geometric distortion), which can serve as a universal descriptor to predict the activities of H2/CO/HCOOH productions for graphene-based SACs, with a high preference for formate production in CuN3 moiety.
Results and discussion
Synthesis and structural characterisation of PSB-CuN3 catalyst
PSB-CuN3 is synthesised via a microwave heating strategy, as schematically illustrated in Fig. 1a. Amine-functionalised graphene nanosheets (AGNs) are used to immobilise the Cu2+ ions through an electrostatic interaction. Then, PSB-CuN3 is obtained through a rapid microwave heating process within five seconds. In comparison to conventional thermal annealing, microwave-induced instant heating leads the local environment in the material to change dramatically, facilitating the formation of a local symmetry-breaking architecture. The ultrathin morphology of as-synthesised PSB-CuN3 can be discerned well by transmission electron microscopy (TEM) and high-angle annular dark-field scanning TEM (HAADF-STEM) images, in which the edges of the two dimentional (2D) nanosheets spontaneously curl upon microwave heating (Fig. 1b, c, Supplementary Fig. 1a). The aberration-corrected HAADF-STEM (AC-HAADF-STEM) image (Supplementary Fig. 1b) shows that no Cu-derived nanoparticles can be observed, which agrees with the X-ray diffraction (XRD) result (Supplementary Fig. 2). Meanwhile, the atomic dispersion of single Cu atoms can be directly identified as the obvious individual bright dots in the high-magnification AC-HAADF-STEM images (Fig. 1d, e). Energy-dispersive X-ray spectroscopy (EDS) mappings analysis shows C, N and Cu elements are homogeneously dispersed on the entire architecture of PSB-CuN3 (Fig. 1f) with a content of 96.7â€‰at %, 2.82â€‰at % and 0.48â€‰at %, respectively (Supplementary Fig. 3). And, the atomic force microscopy (AFM) measurement (Fig. 1g) is performed to present the average thickness of the PSB-CuN3 (~1.61â€‰nm). The N2 adsorption-desorption isotherms shows a high specific surface area of ~510â€‰m2â€‰gâˆ’1 with dominant pore sizes of ~4â€‰nm (Supplementary Fig. 4). The PS-CuN4 catalyst prepared by conventional thermal annealing shows similar physiochemical information (Supplementary Figs. 2â€“6), demonstrating the high controllability for materials synthesis. Subsequently, chemical states of the elements in PSB-CuN3 and PS-CuN4 are investigated by X-ray photoelectron spectroscopy (XPS) analysis (Supplementary Figs. 7â€“9). While three types of graphitic (401.2â€‰eV), pyrrolic (399.4â€‰eV) and pyridinic (398.6â€‰eV) N species for stabilising the Cu atoms exhibit no obvious difference, the Cu 2p spectrum of PSB-CuN3 possesses two dominant peaks at binding energies of 932.0â€‰eV (2p3/2) and 951.8â€‰eV (2p1/2) that are obviously lower than the characteristic binding energies of 934.3 and 954.1â€‰eV for the Cu2+ in PS-CuN4, implying a lower valence state of Cu1+ in PSB-CuN3. Cu LMM spectra are obtained to further demonstrate the chemical valence of the two samples (Supplementary Fig. 10). While PSB-CuN3 is dominated by Cu1+ species located at 571.1â€‰eV, PS-CuN4 is dominated by Cu2+ species located at 572.0â€‰eV41. In addition to the chemical states, Raman spectra show there are a larger number of structural defects in PSB-CuN3 compared with PS-CuN4 (Supplementary Fig. 11). Thus, the above results suggest that the Cu species in PSB-CuN3 catalyst are likely to be dispersed as unusual low-valent mononuclear centres on largely defective graphene nanosheets.
Fig. 1: Synthesis and structural characterisations of PSB-CuN3 catalyst.
a Schematic illustration of the preparation strategy for PSB-CuN3 and PS-CuN4. bâ€“e TEM image (b; scale bar, 200â€‰nm), HAADF-STEM image (c scale bar, 200â€‰nm), AC-HAADF-STEM image (d scale bar, 1â€‰nm) and corresponding intensity maps (e) of PSB-CuN3. f EDS mapping images of PSB-CuN3 (C, red; N, orange; Cu, blue; scale bar, 10â€‰nm). Note that the molybdenum grid was used for all STEM measurements. g AFM image of as-prepared PSB-CuN3 (scale bar, 2â€‰Î¼m), showing measured dimensions of individual nanosheets.


Atomic and electronic structure analysis of PSB-CuN3 catalyst by X-ray absorption fine structure spectroscopy
The coordination environment and chemical state of PSB-CuN3 are further explored by element-selective X-ray absorption fine structure (XAFS) spectroscopy analysis including extended-XAFS (EXAFS) and X-ray absorption near-edge structure (XANES). Figure 2a shows the Cu K-edge EXAFS Fourier-transformed (FT) magnitudes. Both PSB-CuN3 and PS-CuN4 exhibit a major peak at ~1.60â€‰Ã… and a minor satellite peak at ~2.20â€‰Ã…, which is similar to the archetypical profile of pyridinic-N-based MN4C4 motifs21. Concurrently, the EXAFS wavelet-transform analysis is performed to more clearly discriminate the coordination atoms42. Only one intensity maximum at ~4.2â€‰Ã…â€“1 is detected in the two catalysts (Fig. 2b), which confirms the assignment of the major and minor peaks to the Cuâ€“N/C bonding (Supplementary Fig. 12a). However, the two peaks intensities in PSB-CuN3 are obviously weaker than that in PS-CuN4 (Supplementary Fig. 12b), which implies a lower coordination configuration for Cu atoms in PSB-CuN3. Subsequently, quantitative EXAFS curve-fitting analysis is carried out to investigate the coordination configuration (Fig. 2c, Supplementary Figs. 13â€“17; Supplementary Table 1). Coordination numbers of the first N and second C coordination spheres in PS-CuN4 are estimated to be 4.0 and 4.1 at distances of 2.00 and 2.67â€‰Ã…, respectively, supporting the adoption of a pyridinic-N-based CuN4C4 configuration (Supplementary Fig. 17c). By contrast, the best-fit for PSB-CuN3 demonstrates reduced coordination numbers (2.8 for Cuâ€“N and 2.9 for Cuâ€“C) with contracted interatomic distances (1.98â€‰Ã… for the first Cuâ€“N and 2.65â€‰Ã… for Cuâ€“C) for the first and second coordination spheres, suggesting a defective pyridinic-N-based CuN3C3 configuration (Fig. 2c).
Fig. 2: Atomic and electronic structure characterisation by XAFS spectroscopy.
a Fourier-transformed magnitudes of the experimental Cu K-edge EXAFS signals of PSB-CuN3 and PS-CuN4 along with reference samples. The Fourier transforms are not corrected for phase shifts. b EXAFS wavelet transforms for the PSB-CuN3 and reference samples. The contour maxima at ~4.2 and 7.0â€‰Ã…âˆ’1 (dashed lines) are associated with the Cuâ€“N/C and Cuâ€“Cu contributions, respectively. c Cu K-edge EXAFS analysis of PSB-CuN3 with the Cuâ€“N and Cuâ€“C two-body backscattering signals Ï‡2 included in the fit. The calculated (red line) and measured (black dots) spectra show excellent agreement (Supplementary Table 1). d Experimental Cu K-edge XANES spectra of PSB-CuN3, PS-CuN4 and reference samples. e, f Comparison between the experimental Cu K-edge (e) and L3,2-edge (f) XANES spectra of PS-CuN4 and PSB-CuN3 and the theoretical spectra calculated for the depicted CuN4C4 and CuN3C3 structures (Supplementary Fig. 21). g, h The calculated 3d, 4â€‰s and 4p projected densities of states on the Cu sites in CuN4C4 (g) and CuN3C3 (h) structures. i Schematic diagrams of band shifts and hybridisation for the planar CuN4 moiety with local D4h symmetry and the defective CuN3 moiety with lower local C2v symmetry. a.u. arbitrary units.


Due to the higher sensitivity to the 3D arrangement of atoms around the photo-absorber, XANES spectroscopy is applied to better identify the atomic-site structures43,44. Figure 2d shows the hard Cu K-edge XANES spectra for the samples along with references. The PS-CuN4 exhibits a shoulder feature B at ~8987â€‰eV, which is close to that of Cu(NO3)2. It arises from the 1sâ†’4pz transition of a planar-like Cu2+ configuration45,46. However, in the spectrum of PSB-CuN3, a pronounced peak A is observed at a much lower energy of ~8982â€‰eV as in Cu2O, signifying an unusual Cu1+ configuration. Moreover, PSB-CuN3 shows a more flattened profile for the post-edge features C, D and E compared with PS-CuN4. This tendency is also corroborated by the soft XANES results (Supplementary Figs. 18â€“20). And, the Cu L3,2-edge spectra reveal an emergence of two extra post-edge resonant peaks from PS-CuN4 to PSB-CuN3 that coincides with the positions of Cu2O (Supplementary Fig. 20). Those hard and soft XANES differences indicate a peculiar Cu1+ species in PSB-CuN3 and a planar-like Cu2+ species in PS-CuN4, consistent with the XPS results (Supplementary Figs. 9 and 10). Then, as guided by the EXAFS results, we constructed various CuNxCy structural models with different local symmetry (Supplementary Figs. 21 and 22) and resorted to XANES simulation to identify the specific configuration. Figure 2e, f shows that the CuN4C4 moiety with local D4h symmetry (Note: The C4 symmetry axis is perpendicular to the plane) reproduces properly the experimental K- and L3,2-edge XANES profiles of PS-CuN4. In addition, when a N defect is introduced, the obtained CuN3C3 moiety shows a local symmetry breaking from D4h to C2v (Note: The remaining C2 symmetry axis is in the plane) and it produces an excellent agreement to the experimental profiles of PSB-CuN3, particularly for the characteristic peaks A and D at K-edge (Fig. 2e) and for the peaks Fâ€™ and Gâ€™ at L3,2-edge (Fig. 2f). By contrast, more N defects in CuN2C2 moiety lead to a dramatic discrepancy in the profiles (see Supplementary Fig. 23). The probability for the substitution of the N atoms in the first coordination sphere by C atoms can also be excluded (Supplementary Figs. 24 and 25, Supplementary Table 2 and Supplementary Note 1). Therefore, the combination of EXAFS and XANES analyses has unambiguously revealed a dominance of defective CuN3C3 moiety with lower local C2v symmetry and mono-valence implanted in a graphene sheet for PSB-CuN3 catalyst.
Because the K-edge (1sâ†’4p transition) and L3,2-edge (2pâ†’3d transition) XANES can probe the unoccupied density of states on the 4p and 3d orbitals of the Cu sites, the emergent resonant features (A, Fâ€™ and Gâ€™) in PSB-CuN3 catalyst indicate a unique p and d electronic distribution that can be regulated by the local coordination symmetry. Detailed analysis on the density of states (Fig. 2gâ€“i, Supplementary Figs. 26 and 27 and Supplementary Note 2) unveils that the local symmetry lowering from D4h in CuN4C4 to C2v in CuN3C3 can drive an obvious upshift of the anti-bonding 4â€‰s and 4p state to above the Fermi energy and finally trigger a dramatic dsp hybridisation at the energy positions of peaks A/F/G and Fâ€™/Gâ€™, where the p orbital is dominated by the in-plane px/py orbital. The local symmetry lowering to C2v in CuN2C2 promotes a dsp2 hybridisation with the p orbital arising mainly from the in-plane px and py orbitals (Supplementary Fig. 28 and Supplementary Note 2). Accordingly, those anti-bonding state upshifts induced by local symmetry breaking in defective CuNxCy moieties can be pivotal to modifying the adsorption strength of the reactive intermediates and thus optimise the reaction pathways for CO2RR with high activity and selectivity.
Evaluation of CO2RR performance
The intrinsic CO2RR activity of the as-prepared samples was performed in a gas-tight H-type cell by coating a thin electrocatalyst layer on the carbon fibre paper with a catalyst loading of 0.5â€‰mgâ€‰cmâˆ’2. The electrolytes pH (pHâ€‰=â€‰7.3 for CO2 saturated 0.5â€‰M KHCO3) and the potential of the reference electrode were checked before the test to ensure the reliability of the as-obtained electrochemical results (Supplementary Figs. 29 and 30). Linear sweep voltammetry (LSV) is employed to obtain polarisation curves of PSB-CuN3, PS-CuN4 and AGNs electrodes in the Ar and CO2 saturated 0.5â€‰M KHCO3 aqueous solution. As shown in Supplementary Fig. 31, the reduction current increases greatly when the electrolyte solution is saturated with CO2. Supplementary Fig. 32 shows a sharp increased cathode current response started at an onset potential of âˆ’0.41â€‰V vs. RHE observed from the PSB-CuN3 electrode. By contrast, PS-CuN4 exhibits an onset potential (âˆ’0.72â€‰V vs. RHE) under identical conditions. Further, chronoamperometric measurements were carried out at different potentials to quantitively identify product distributions using online gas chromatography and nuclear magnetic resonance (NMR) spectroscopy (Supplementary Figs. 33 and 34). As shown in Supplementary Fig. 35 and Supplementary Table 3, formate is the only liquid product from the CO2RR, with a certain amount of gaseous products including methane, carbon monoxide and hydrogen. No other C1 and C2 liquids (i.e., methanol, ethanol, n-propanol and acetone) were detected (Supplementary Fig. 34). Impressive FEs (HCOOâˆ’) (>90%) were obtained with the PSB-CuN3 throughout the potential range (âˆ’0.75 to âˆ’1.0â€‰V vs. RHE). On the contrary, the maximal FEs of formate for PS-CuN4 and AGNs electrodes are 72.4% and 9.5% under identical testing conditions. This suggests the PSB-CuN3 moiety should be responsible for the high selectivity for formate. Besides, chronoamperometric measurement was carried out to investigate the stability of the as-prepared PSB-CuN3. As shown in Supplementary Fig. 36, the FE of the PSB-CuN3 electrode is ~92% after 16â€‰h electrocatalysis compared with PS-CuN4 (~27%). To investigate CO2 reduction performance under industrial level current of PSB-CuN3, a flow cell equipped with gas diffusion electrodes (GDEs) was fabricated (Supplementary Fig. 37). Figure 3a shows that the GDE with PSB-CuN3 could be operated at a current density of 150â€‰mAâ€‰cmâˆ’2, with modest overpotentials in 0.5â€‰M KHCO3 solution, which is much higher than the PS-CuN4 under the same applied potential. The optimal FEs of PSB-CuN3 electrode is 97.9%, which is higher than the PS-CuN4 (78.8%) (Fig. 3b and Supplementary Table 4). With this FE, a CO2RR to formate partial current density of ~62â€‰mAâ€‰cmâˆ’2 was achieved (Fig. 3c). Moreover, the FE of PSB-CuN3 for HCOOâˆ’ can be maintained over 90% through a wide applied potential range from âˆ’0.65 to âˆ’0.97â€‰V vs. RHE (Fig. 3d, e), exhibiting excellent selectivity for CO2RR-to-formate. It should be stressed that both the selectivity and activity of the PSB-CuN3 are comparable to the previously best-reported CO2RR-to-HCOOâˆ’ catalysts (Fig. 3f and Supplementary Table 5). Turnover frequency (TOF), as a key parameter for CO2RR efficiency assessment, is then calculated based on the standard equation. As shown in Supplementary Fig. 38, the maximum TOF value of PSB-CuN3 is 30825.3â€‰hâˆ’1 at a given potential of âˆ’1.1â€‰V vs. RHE, considerably higher than PS-CuN4 (8558.8â€‰hâˆ’1) and is comparable to the recently best-reported CO2RR-to-HCOOâˆ’ catalysts (Supplementary Table 5). To show the scalability of the as-fabricated flow cell with PSB-CuN3 electrode, the long-term durability was tested under the large current density. As shown in Fig. 3g, the PSB-CuN3 electrode maintains an average current density of 94.4â€‰mAâ€‰cmâˆ’2 with over 90% FEformate during 100â€‰h testing. By contrast, the PS-CuN4 electrode shows 58.3% of the initial reduction current of 25â€‰mAâ€‰cmâˆ’2 with an average FE of 68% (Supplementary Fig. 39). Further structural stability characterisation by AC-HAADF-STEM and XAFS suggest that the Cu atom and its coordinating matrix in PSB-CuN3 are robust enough to withstand the CO2RR operation condition (see below detailed discussion). Thus, the PSB-CuN3 electrode could be readily incorporated into practical CO2RR electrolysers because of its high activity, selectivity and durability at large current densities.
Fig. 3: Evaluation of CO2RR catalytic activity by electrochemical measurements.

a Linear sweep voltammetry curves of PSB-CuN3 and PS-CuN4 in a flow-cell setup using 0.5â€‰M KHCO3 electrolyte at 25â€‰Â°C. b The corresponding formate FEs under different potentials. The maximal formate FE of PSB-CuN3 is 97.9% at âˆ’0.87â€‰V vs. RHE. c Comparison of partial current density for CO2-to-formate conversion between PSB-CuN3 and PS-CuN4. The maximal formate partial current on PSB-CuN3 reached 125â€‰mAâ€‰cmâˆ’2 at âˆ’1.1â€‰V vs. RHE, which represents 3.6 folds higher than that of PS-CuN4. d Chronoamperometric responses of PSB-CuN3 in CO2-saturated electrolyte at different potentials. The potential was corrected with 100% iR compensation. e The corresponding FEs of different products under different potentials on PSB-CuN3. Note that the total FEs are not complete 100% due to systems errors and double layer current of catalysts. The presented values represent the mean, while the error bars indicate the standard deviation, based on three independent measurements. f Comparison of formate FEs and overpotentials for PSB-CuN3 and various catalysts reported at similar conditions (See the details in Supplementary Table 5). g Long-term stability test of PSB-CuN3 at a potential of âˆ’0.95â€‰V vs. RHE.
Mechanism of CO2 activation and reduction by first-principles calculation and operando spectroscopies
To understand the origin of high efficiency toward formate generation on PSB-CuN3 catalyst, first-principles calculations were carried out for the electroreduction process of CO2 on various graphene-based CuNxCy structures. We considered the reaction pathways for CO, formate, methanol and methane productions as well as the competing hydrogen evolution reaction (HER). For the formation of intermediate *OCHO/*COOH, two mechanisms were considered, i.e., a reaction with *H via a CO2 insertion reaction into the metalâ€“hydrogen bond and a direct protonation with H+ from solution47,48. The optimised structures of the intermediates and the Gibbs free-energy diagrams at zero electrode potential show remarkable response to the coordination environment (Supplementary Figs. 40â€“47, Supplementary Tables 6 and 7). On the CuN4C4 and CuN3C3 structures, the different adsorbed states of *COOâˆ’ and *OCOâˆ’ for CO2 via the first electron addition are suggested to be the potential limiting steps for CO and formate productions, respectively, and the *OCOâˆ’ formation is more energetically favoured than *COOâˆ’. By contrast, on the CuN2C2 structure, the potential limiting steps shift to the desorption of CO and formate. The pathways initiated by *H adsorption do not show any preference than the pathways initiated by *COOâˆ’/*OCOâˆ’ adsorption. It is well documented that the reduction of CO2 to CH4 and CH3OH shares similar pathways until the formation of *CO intermediate, and the potential limiting steps are shown to be 
âˆ—+CO2â†’âˆ—COOâˆ’
 for CuN4C4, 
âˆ—COâ†’âˆ—CHO
 for CuN3C3 and CuN2C2 (Supplementary Fig. 47). The magnitudes of the corresponding theoretical limiting free-energy differences (Î”GL) are summarised in Supplementary Fig. 47. While the three distinctly coordinated catalysts exhibit a flipped volcano relationship for CO and formate activities, the predicted HER and CH4/CH3OH activities are simultaneously enhanced by lowering the local symmetry via introducing N vacancy. In comparison to the highly symmetrical CuN4C4, it leads to a much lower Î”GL (0.23â€‰eV) for the formate pathway than the CO (0.68â€‰eV), CH4/CH3OH (0.78â€‰eV) and HER (1.01â€‰eV) pathways in the local symmetry-broken CuN3C3. But, this tendency is reversed by more N vacancies in CuN2C2 which exhibits a preference to the HER pathway. The high selectivity for CO2 conversion to formate in CuN3C3 can be also appreciated from the large difference between the Î”GL for CO2 reduction and HER. The probability for C2 products generation was also considered by examining the first Câ€“C coupling of two *CO species that is crucial for the formation of C2 products (Supplementary Fig. 47). This step is found to have a large Î”GL (1.79â€‰eV on CuN4C4 and 1.67â€‰eV on CuN3C3), indicating a formidable Câ€“C coupling of *CO species for the generation of C2 products. Thus, the HCOOH production is thermodynamically preferred than the generation of other C1 and C2 products and H2 on the CuN3C3 structure at zero electrode potential.
Because the electric field can interact with intermediates with a substantial dipole moment and/or polarizability49,50,51, we further examined the effects of interfacial electric field and pH on the adsorbate bindings on different CuNxCy structures (Supplementary Table 7). A significant response of the adsorbate bonding strength to the electric field effects is revealed in Fig. 4a and Supplementary Fig. 48. Then, an applied potential of âˆ’0.80â€‰V vs. RHE and pHâ€‰=â€‰7.3 is considered and it leads to a remarkable decrease of the free energy for the adsorbates (Fig. 4bâ€“e, Supplementary Figs. 49 and 50), which indicates more stabilised intermediates and decreased endothermicity for the reaction steps. On the CuN3C3 structure, the potential limiting steps for formate and CO productions are shifted to 
âˆ—OCOâˆ’â†’âˆ—OCHO
 and CO desorption, respectively (Fig. 4b), and it produces a considerably lower limiting free-energy difference (Î”GL) of âˆ’0.03â€‰eV for the formate pathway (Fig. 4e), which is in contrast to the comparable Î”GL between the HER (0.51â€‰eV), CO (0.63â€‰eV) and CH4/CH3OH (0.51â€‰eV) pathways, indicating a strong preference to the formate production. By contrast, a much higher Î”GL of 0.62â€‰eV is shown for the formate pathway on the CuN4C4 structure. This prediction agrees well with the impressively higher FE for CO2RR-to-formate on PSB-CuN3 catalyst than on PS-CuN4 catalyst over similar potential range (see Fig. 3). CuN2C2 exhibits an extremely strong preference to the HER pathway with a Î”GL of âˆ’0.68â€‰eV. Similar potential dependence is obtained at an applied potential of âˆ’0.40â€‰V vs. RHE (Supplementary Table 7). All in all, the adsorbate energies corrected for the electric field and pH effects strongly support the highest activity and selectivity toward formate in the local symmetry-broken CuN3C3 moiety.
Fig. 4: First-principles calculations and operando XAFS characterisation on PSB-CuN3 catalysing CO2RR.
a Electric field effects on the adsorbate bindings on the CuN3C3 structure. bâ€“d Gibbs free-energy diagrams for CO2 electroreduction to CO, HCOOH, H2, CH4 and CH3OH on CuN3C3 structure. The highlights indicate the potential limiting steps. e Magnitudes of the theoretical limiting free-energy differences (Î”GL) for CO, HCOOH, H2 and CH4/CH3OH productions. f Differential charge density and projected density of states for the Cu centre and the bound O atom in the adsorbed *OCHO intermediates supported on CuN3C3 structure. White, grey, blue, red and cyan balls represent H, C, N, O and Cu atoms, respectively. The black dashed line represents the Femi energy level. g Adsorption free energies of Î”G*OCOâˆ’, Î”G*COOâˆ’, Î”G*OCHO, Î”G*COOH and Î”G*H vs. the descriptor Ï†. h Thermodynamic relations (volcano) lines for the H2, CO and HCOOH productions as a function of the descriptor Ï†. i Fourier-transformed magnitudes of operando Cu K-edge EXAFS spectra for PSB-CuN3 recorded at various applied potentials vs. RHE in 0.5â€‰M KHCO3 aqueous solution. a.u., arbitrary units.

To understand why the modified electronic structure in PSB-CuN3 catalyst is in favour of CO2 reduction toward formate, the projected density of states for the adsorbed intermediates of *COOâˆ’ and *OCOâˆ’ were obtained. Supplementary Fig. 51 shows the overlap degree between Cu-d and O-p orbitals is generally larger than that between Cu-d and C-p orbitals on each CuNxCy structure, thus indicating a stronger binding of *OCOâˆ’ than *COOâˆ’. For *OCOâˆ’, the overlap degree between Cu-d and O-p orbitals in CuN3C3 and CuN2C2 is obviously increased as compared to CuN4C4. However, the binding of the intermediates in CuN2C2 is so strong that the potential limiting step shifts to the formate desorption. It suggests that the local symmetry-broken CuN3C3 can produce a suitable overlap degree between the Cu-d and O-p orbitals, which can result in a neither strong nor weak binding of the intermediates. Thus, those results unveil that the highest activity and selectivity toward formate in the local symmetry-broken CuN3C3 moiety is strongly correlated with its unusual coordination environment, in good agreement with the experimentally detected excellent activity of the PSB-CuN3 catalyst in electrochemical CO2 reduction (Fig. 3).

The intrinsic coordination characteristics of the active CuNx centres that dominate the adsorbate binding strength and activity are explored. Breaking of the local D4h symmetry to create planar-like CuN3 and CuN2 moieties with lower local C2v symmetry leads to charge redistribution (Fig. 4f) and increased attractive interactions for the intermediates (Fig. 4bâ€“e). The charge redistribution is first manifested as an obvious upshift of the d-band centre of Cu atom in energy relative to the Fermi level from CuN4 to CuN3 moiety, but this tendency is reversed by a dramatic downshift of the d-band centre in CuN2 moiety (Supplementary Fig. 52a). This anomaly leads to the widely used d-band centre model cannot correlate with the increased adsorbate binding strength and activity (Supplementary Fig. 52b, c), which stems from the enhanced dsp orbitals hybridisation near the Fermi level in local symmetry-broken CuNx moieties (Supplementary Fig. 53) as also indicated by experimental XANES results. By contrast, the total electron transfer from Cu and the work function appear to be dominated by the coordination number of nearest-neighbour N atoms (denoted as nN, Supplementary Fig. 54aâ€“c), which matches well with the valence state change obtained by XANES. In addition, approximate linear relationships are observed when nN is plotted against adsorption energies and limiting free-energy differences for different intermediates (Supplementary Fig. 54d, e), which suggests that nN can be a significant factor of the descriptor52. Besides, it is noted that angular distortion has been widely used to describe the stereochemical activity of ns2 electron pairs in metal halide perovskites ABX353. It suggests that the angle distortion between neighbouring Cuâ€“N bonds in the present work can further modify the charge redistribution and binding affinity, and we add it as an additional factor to modify the descriptor as follows:
Ï†=ğ‘›ğ‘Ã—(1âˆ’1ğ‘›ğ‘âˆ‘{ğ‘–=1}ğ‘›ğ‘|90âˆ’ğ›¼ğ‘–90|)
(1)
where nN is the number of nearest-neighbour N atoms, Î±i is the Nâ€“Cuâ€“N bond angle between neighbouring Cuâ€“N bonds and na is the number of the Nâ€“Cuâ€“N bond angles. As shown in Fig. 4g, Supplementary Fig. 55 and Supplementary Table 8, excellent linearity is obtained between Ï† and the adsorption energies for *COOâˆ’, *OCOâˆ’, *COOH, *OCHO and *H. It means that a decrease in Ï† corresponds to an enhancement in binding strength, which particularly causes a tighter binding of *COOâˆ’, *OCOâˆ’, *COOH, *OCHO and *H. This can be attributed to the adoption of a peculiar slant adsorption way (instead of the conventional top-site way, see Fig. 4f) of *OCHO intermediate on the CuNx centres with the successive introduction of N vacancy. It leads to an unusual coupling between the in-plane and out-of-plane 3d, 4s and 4p orbitals of the Cu centre and the 2s and 2p orbitals of the bonded O atom of intermediates (Fig. 4f, Supplementary Figs. 56â€“58). It suggests that the orbital coupling degree and binding affinity between the Cu centre and the intermediates is directly correlated with the local geometric symmetry breaking of the surrounding N atoms (Supplementary Note 3). Figure 4h shows the theoretical limiting free-energy differences versus the descriptor Ï†. While a good linearity with different slopes is shown for H2 production, a flipped volcano relationship is revealed for CO, formate and CH4/CH3OH productions, with the CuN3 moiety lying at bottom of the volcano. The relative order of activity and selectivity from the theoretical prediction reproduces well the experiment data (Fig. 3), thus validating our descriptor Ï†. Hence, the local coordination symmetry breaking, which is quantified as a combination of nearest-neighbour coordination number and bond angle distortion, can serve as a universal descriptor to evaluate the activity of graphene-based single Cu catalysts towards CO2RR.
The dynamic processes for CO2 adsorption and activation towards highly selective production of formate on PSB-CuN3 were further investigated by experimental Cu K-edge XAFS spectroscopy under operando conditions (Fig. 4i, Supplementary Figs. 59â€“66, Supplementary Table 1). The XANES spectra display that the Cu K-edge of pristine PSB-CuN3 initially shifts to higher energy and the intensity of the shoulder peak D also increases when immersing in CO2-saturated 0.5â€‰M KHCO3 solution under open-circuit potential (OCP) bias (Supplementary Fig. 59a). This indicates an increased oxidation state for the Cu sites, which is caused by the charge transferred from low-valent Cu1+ to CO2 molecule to form CO2Î´+ species. This assumption is consistent with a slight contraction of the first major peak from 1.60â€‰Ã… to 1.58â€‰Ã… as well as an elongation of the second satellite peak from 2.20â€‰Ã… to 2.26â€‰Ã… in EXAFS-FT under OCP bias (Fig. 4i). It reflects an additional contribution from the Cuâ€“O bond of the dynamically adsorbed CO2 molecular. During CO2 reduction at âˆ’0.5â€‰V and âˆ’0.8â€‰V vs. RHE, the Cu K-edge XANES shows a shift back to lower energy without obvious profile change, which indicates the adoption of a low oxidation state for the Cu sites during the catalytic stage, corroborating the high activation ability of CuN3C3 centres for CO2 reduction. Meanwhile, the first two peaks in EXAFS-FT, which now consists of the Cuâ€“N bonds and the Cuâ€“O/C bonds from the reaction intermediates (e.g., *OCOâˆ’, *OCHO), demonstrates a continuous shift to shorter length and an increase in intensity (Fig. 4i), corresponding to a contraction of Cuâ€“N/O/C bonds with decreased bond disorder. When the applied potential returns to OCP, both XANES and EXAFS-FT reverse to the initial OCP state. Further quantitative EXAFS curve-fitting analysis reveals a decrease in Cu-N/O/C interatomic distances as well as an increase in coordination numbers during the operating process (Supplementary Fig. 63 and Supplementary Table 1). Similar evolution behaviours for the electronic and atomic structures of the metal centres under the catalytic conditions have also been observed in single-Ni-atom catalyst for CO2 reduction to CO54. Here, the above phenomena consolidate a tighter binding of the reaction intermediates (e.g., *OCOâˆ’, *OCHO) under the catalytic conditions, which is suggested to be responsible for the high catalytic activity and selectivity of the PSB-CuN3.
At the same time, attenuated total reflectance surface-enhanced infra-red absorption spectroscopy (ATR-SEIRAS) was used to probe the reaction intermediates on PSB-CuN3 (Supplementary Fig. 67). When the applied potential is more negative than âˆ’0.8â€‰V vs. RHE, the obvious peak at ~1620â€‰cmâ€“1 exhibits a pronounced rise and red shift, which can be ascribed to the asymmetrical Oâ€“Câ€“O stretching band of dissolved formate. Additionally, two broad bands appeared between 1450 and 1350â€‰cmâ€“1, and they can be assigned to the symmetrical Oâ€“Câ€“O stretching mode of bridge-bonded formate and the Câ€“OH vibration of desorbed formate molecules, respectively55. No signal of *COOâˆ’/*COOH was detected. These findings evidence the preferred formation of *OCOâˆ’/*OCHO intermediates bonded to the PSB-CuN3 catalyst via O.
The impact on the catalytic and structural stability from N coordination number variation is investigated (See detailed discussion in Supplementary Note 4). While PSB-CuN3 shows better electrocatalytic stability than PS-CuN4 in both H-cell and flow cell under the same applied potential, the detailed operando and postmortem XAFS with AC-HAADF-STEM characterisations revealed a structural reconstruction occurs on both PSB-CuN3 and PS-CuN4 during CO2 reduction. Compared with PS-CuN4 with high valence Cu2+, the structure change of low-valent Cu1+ PSB-CuN3 is slight. And, the structural evolution process of PSB-CuN3 is reversible and no atoms aggregation can be observed after CO2 electrocatalysis. On contrary, the dramatic structural reconstruction of PS-CuN4 leads to an irreversible aggregation of fractional Cu atoms. Therefore, the electrochemical stability of PSB-CuN3 is better than that of PS-CuN4, consistent with previous works that the low-valent Ni1+ single-atom catalysts showed no reconstruction into nanoparticles for CO2RR to CO54, but the high-valent Cu2+ single-atom catalysts reconstructed to nanoparticles in the electrochemical reduction of nitrate to ammonia56 and electrocatalytic oxygen reduction57.
In summary, the local planar-symmetry-broken PSB-CuN3 is successfully prepared by a simple and facile method, and further used as the electrocatalysts for CO2RR. Significantly, the as-prepared PSB-CuN3 catalyst exhibits extremely high electrocatalytic activity and catalytic stability in both H-cell and flow cell systems, superior to the highly symmetrical PS-CuN4 and comparable to recently best-reported CO2RR-to-HCOOâˆ’ catalysts. By combining ex-situ/operando EXAFS, XANES and ATR-SEIRAS analyses with theoretical calculations, we unambiguously unveiled that the high catalytic performance for formate production originates from the intrinsic local symmetry breaking from the planar-like D4h to C2v configuration in PSB-CuN3, such that an unconventional dsp hybridisation is emerged. With this anomaly, the catalytic activities are unconformable to the widely used d-band centre theory, whereas the local environment of the metal centre including coordination number and geometric distortion is identified as a newly universal descriptor to the intermediate adsorption and thus catalytic activities for H2/CO/HCOOH/CH4/CH3OH productions in graphene-based SACs. Our study provides a universal and efficient guideline for the rational design and accurate modulation of the active centreâ€™s local symmetry in nonprecious metal SACs for diverse CO2RR applications at the atomic scale.
Methods
Chemicals
CuCl2Â·6H2O (99.99% AR grade) was bought from Alfa Aesar. Nafion solution (5â€‰wt%), natural graphite flakes (100 mesh), ammonia solution (~30%) and KHCO3 (99.95%) were obtained from Sigma Aldrich. All chemicals were used directly without further purification. The ultrapure water (18â€‰MÎ©) used in the experiments was supplied by a Millipore System (Millipore Q).
Synthesis of PSB-CuN3 and PS-CuN4
Graphene oxide (GO) was first synthesised by a modified Hummersâ€™ method58. Then, NH4OH solution (0.5â€‰mL) was added dropwise to a GO aqueous suspension (5â€‰mL, 2â€‰mgâ€‰mLâˆ’1) and stirred for 20â€‰min. The mixed suspension was transferred to a Teflon vessel at 70â€‰Â°C for 5â€‰h under airtight conditions. After cooling and six times washing using water, the purified AGNs were obtained. Subsequently, 80â€‰Î¼L CuCl2Â·6H2O (3â€‰mgâ€‰mLâˆ’1) was added to the well-dispersed AGNs solution under stirring for 1â€‰h. The purified samples were subsequently transferred into a freeze-drying vessel, which had been set at the temperature of âˆ’53â€‰Â°C and vacuum pressure of âˆ’30â€‰Pa. After being freeze-dried, the Cu2+-containing AGNs, together with small amounts (â‰ˆ1â€‰wt%) of thermally reduced (300â€‰Â°C for 1â€‰h in Ar) graphene as catalysts, were sealed in an Ar gas-protected glass vial and then undergo a microwave process (microwave digestion workstation XH-800C) under 1000â€‰W for 5â€‰s59. PSB-CuN3 was collected after a cooling period for 20â€‰min. For comparison, PB-CuN4 was synthesised by the pyrolysis of Cu2+-containing AGNs in an Ar/NH3 mixed atmosphere according to the reported method17.
Physical characterisations
XRD patterns were collected by using a Bruker D8 Advance X-ray powder diffractometer with a Cu KÎ± irradiation source (Î»â€‰=â€‰1.5406â€‰Ã…), and a self-calibration process was performed with a SiO2 internal standard sample prior to target measurement. The sizes and morphologies of the samples were first recorded using a Hitachi H-800 transmission electron microscope. Then, the high-resolution STEM and elemental mappings were carried out on a JEOL JEM-2100F with an electron acceleration energy of 200â€‰kV. A JEM-ARM200F transmission electron microscope operated at 200â€‰keV, equipped with a probe spherical aberration corrector, was used to obtain HAADF-STEM images of the samples. All TEM/STEM samples were prepared by depositing a tiny amount of the samples onto a molybdenum (Mo) grid. Quantitative analysis of metal loading was carried out using inductively coupled plasma atomic emission spectroscopy (TJA RADIAL IRIS 1000 ICP-AES). XPS spectra were performed by a Thermo Scientific K-Alpha spectrometer using a monochromatic Al KÎ± radiation, where the analysis chamber was 1.5â€‰Ã—â€‰10âˆ’9â€‰mbar and the X-ray spot was ~500â€‰Î¼m. The C 1â€‰s peak for adventitious hydrocarbons at 284.8â€‰eV was used for binding energy calibration. Raman measurements were conducted on a Renishaw Microscope System RM2000 with a Renishaw spectrometer of 532â€‰nm. The N2 adsorption/desorption curve was performed at 77â€‰K using a Quadrasorb SI (Quantachrome, USA) surface area analyser.
Electrochemical measurements
The electrochemical measurements were performed on a CHI 760e electrochemical workstation in a three-electrode configuration cell using an as-prepared electrode as the working electrode, platinum wire as the counter electrode and Ag/AgCl (saturated KCl) as the reference electrode in 0.5â€‰M KHCO3 aqueous electrolyte. Nafion117 membrane was inserted between the cathodic chamber and anodic chamber of H-type cell. The catalyst inks were prepared by a mixture of catalyst (5â€‰mg), water (240â€‰Î¼L), isopropanol (750â€‰Î¼L) and 5% Nafion solution (10â€‰Î¼L), followed by ultrasonication for 4â€‰h. An aliquot of 0.1â€‰mL of the catalyst ink was deposited onto carbon paper (HCP120, Shanghai Hesen Electric Co., Ltd) and allowed to dry in air, giving a catalyst loading of 0.5â€‰mgâ€‰cmâˆ’2. Before the measurement, the 0.5â€‰M KHCO3 aqueous solution (pHâ€‰=â€‰7.3) was purged by bubbling Ar for 30â€‰min and then switched to CO2 until saturation. The LSV curves were conducted at a scan rate of 10â€‰mV/s with continuous bubbling CO2. For comparison, cyclic voltammetry curves tested in an Ar (99.99%)-saturated electrolyte were also obtained. All potentials were converted to the RHE scale according to the standard Nernst equation (
ERHE=EAg/AgCl+0.0591Ã—pH+0.197V
). All current densities were normalised to the geometrical area of electrode (1â€‰Ã—â€‰1â€‰cm2). The gas products of CO2RR were checked by online gas chromatography (GC, Supplementary Fig. 33). The 1H NMR spectroscopy was used to analyse and quantify the products in the liquid phase (Supplementary Fig. 34). Samples were prepared by mixing DMSO and electrolyte aliquots. All the potentials were reported vs. RHE and manually corrected by 100% iR compensation, which was obtained through electrochemical impedance spectroscopy (Supplementary Fig. 30 and Supplementary Table 9). EIS measurements were conducted at 25â€‰Â°C using an H-type cell (consisting of two independent anode and cathode chambers separated by a nafion membrane) and a CHI 760e electrochemical workstation. Platinum wire served as the counter electrode, while Ag/AgCl (saturated with KCl) was employed as the reference electrode. The EIS measurement was carried out under open-circuit voltage, and the frequency setting range was from low frequency 1â€‰Hz to high frequency 100,000â€‰Hz. The amplitude was set as 0.005â€‰V. For the flow cell study, catalyst inks were air-brushed onto a gas diffusion layer (GDL) as the cathode electrode with a mass loading of ~0.6â€‰mgâ€‰cmâˆ’2. IrO2 was used as the counter electrode. The IrO2 catalyst ink was prepared by a mixture of commercial IrO2 powder (5â€‰mg), water (240â€‰Î¼L), isopropanol (750â€‰Î¼L) and 5% Nafion solution (10â€‰Î¼L), followed by ultrasonication for 4â€‰h. Then, IrO2 catalyst (0.7â€‰mg/cm2) was loaded onto the GDL as anode. Electrolyte was pumped by a syringe pump (PHD 2000, Harvard Apparatus) over the cathode GDL at a constant flow rate of 10 standard cubic centimetres per minute (sccm). High-purity CO2 gas flowed at a rate of 30â€‰sccm behind the cathode GDL controlled by a mass flow controller (CS200, Beijing Sevenstar flow). Polarisation curves of all studied electrodes are performed at a sweeping potential between 0 and âˆ’1.2â€‰V vs. RHE (all potentials reported here were based on RHE) with a scan rate of 10â€‰mVâ€‰sâˆ’1.
Gas product analysis
Gas products from the cathodic compartment during the electrochemical reactions were analysed using a GC-2014 (Shimadzu) equipped with a BID detector and ShinCarbon ST100/120 packed column. High-purity helium (99.9999%) was used as the carrier gas. The FEs of the gas products were calculated by using the concentrations (ppm) detected by the GC as follows (CO products are selected as an example):
FECO(%)=ğ‘„COQtotalÃ—100%=(ğœˆ60s/min)Ã—(ğ‘¦22400cm3/mol)Ã—ğ‘Ã—ğ¹Ã—100%ğ‘—
(2)
where Î½ is gas flow rate measured by a flowmeter at the exit of the cell at room temperature, y is the measured volume concentration of product, N is the number of electrons required to form a molecule of gas product, F and j are the Faraday constant (96,500â€‰Câ€‰molâˆ’1) and the recorded current, respectively.
Liquid product analysis
Liquid products were analysed by using a 400â€‰MHz NMR spectrometer (Bruker), with DMSO as internal standards. The calibration curve was made by measuring standard solutions of formate. The FEs of liquid products were calculated as follows:
FEHCOOâˆ’(%)=ğ‘„HCOOâˆ’ğ‘„totalÃ—100%=ğ‘›HCOOâˆ’Ã—ğ‘Ã—ğ¹Ã—100%ğ‘—Ã—ğ‘¡
(3)
where nHCOOâˆ’ is the measured amount of formate in the cathodic compartment and t is the reaction time. The HCOOâˆ’ partial current density at different potentials was calculated by multiplying the overall geometric current density and its corresponding FE.
Turnover Frequency (TOF, hâˆ’1) for the CO2 electroredution to formate was evaluated based on the 2-electron pathway:
TOF=ğ¼product/ğ‘ğ¹ğ‘šcatÃ—ğ‘¤/ğ‘€metalÃ—3600
(4)
where Iproduct is the partial current for formate, N is the number of electrons transferred for product formation, F is the Faraday constant (96,500â€‰Câ€‰molâˆ’1), mcat is the catalyst mass in the electrode, w is the metal loading in the catalyst and Mmetal is the atomic mass of Cu (63.5â€‰gâ€‰molâˆ’1).
Soft XAS measurements
Soft C, N K-edge and Cu L3,2-edge X-ray absorption spectra were measured at BL12B station of Hefei National Synchrotron Radiation Laboratory in the total electron yield mode by collecting the sample drain current under a vacuum better than 10âˆ’7â€‰Pa.
Ex-situ and operando XAFS measurements
Hard X-ray absorption spectra at Cu K-edge were acquired under ambient conditions in fluorescence mode using Si(111) double-crystal monochromators at 1W1B and 4B9A beamlines of Beijing Synchrotron Radiation Facility (BSRF, operated at 2.5â€‰GeV with a maximum current of 250â€‰mA). The incident and fluorescence X-ray intensities were monitored by using standard N2-filled ion chambers and air-filled Lytle-type detector, respectively.
A home-made cell collected to a computer-controlled electrochemical analyser was used for the operando XAFS experiments, which were operated under identical conditions as the electrochemical measurements. A catalyst-modified carbon paper was used as working electrode, Pt wire as counter electrode and Ag/AgCl (KCl-saturated) electrode as the reference electrode. CO2-saturated 0.5â€‰M aqueous KHCO3 was used as electrolytes. Operando XAFS spectra were recorded at different potentials to monitor the structural evolutions of the catalysts.
XAFS data analysis
The XAFS raw data were background-subtracted, normalised and Fourier-transformed by the standard procedures with the ATHENA programme44,60, and least-squares curve-fitting analysis of the EXAFS Ï‡(k) data was carried out using the ARTEMIS programme based on the standard EXAFS equation. All fits were performed in the R space with k-weight of 2. The best-fit results are shown in Supplementary Figs. 13â€“17 with the fitting parameter values listed in Supplementary Table 1.
The Cu K- and L-edge theoretical XANES calculations were carried out with the FDMNES code in the framework of real-space full multiple-scattering (FMS) scheme using Muffin-tin approximation for the potential43,61,62. The energy-dependent exchange-correlation potential was calculated in the real Hedinâ€“Lundqvist scheme, and then the spectra was convoluted using a Lorentzian function with an energy-dependent width to account for the broadening due both to the coreâ€“hole width and to the final state width. The CuNxCy moieties were built based on the graphene-enclosed MN4C4 structural motif (Supplementary Fig. 21), which can be regarded as a single Cu atom occupying the divacancy in the graphene lattice with coordination to four pyridinic Ns included in six-member rings. Satisfactory convergence for the cluster sizes had been achieved.
Operando ATR-SEIRAS
Operando ATR-SEIRAS was performed on a Nicolet iS20 spectrometer equipped with a VeeMax III (PIKE technologies) accessory and an HgCdTe (MCT) detector cooled with liquid nitrogen. The electrochemical test was conducted in a custom-made three-electrode electrochemical single cell. A Pt wire and a saturated Ag/AgCl were used as the counter and reference electrodes, respectively. A fixed-angel Si prism (60Â°) coated with Au thin layer was used to load catalysts and serve as the working electrode. The operando ATR-SEIRAS spectra were recorded by varying the potential in CO2-saturated 0.50â€‰M KHCO3. The spectrum recorded at the initial potential of 0â€‰V was used for the background subtraction.
Computational method and models
The first-principles calculations including structural and electronic performances were carried out in the framework of density functional theory (DFT) as implemented in the Cambridge Sequential Total Energy Package (CASTEP)63. The exchange-correlation functional under the generalised gradient approximation (GGA) with norm-conserving pseudopotentials and Perdewâ€“Burkeâ€“Ernzerhof functional was adopted to describe the electron-electron interaction64. The spin polarisation was considered52. The DFT-D2 method of Grimme for the Van der Waals correction was used65. An energy cutoff of 750â€‰eV and a k-point sampling set of 5â€‰Ã—â€‰5â€‰Ã—â€‰1 were tested to be converged. A force tolerance of 0.01â€‰eVâ€‰Ã…âˆ’1, energy tolerance of 5.0â€‰Ã—â€‰10âˆ’7â€‰eV per atom and maximum displacement of 5.0â€‰Ã—â€‰10âˆ’4â€‰Ã… were adopted. The surfaces of the structures were built with a vacuum space of 20â€‰Ã… along the z-direction, which is enough to avoid interaction between the two neighbouring images. The effect of water was considered by using the implicit solvent model with the dielectric constant of 78.54 for water molecule. The structure models were built with a vacuum space of 20â€‰Ã… along the z-direction, which is enough to avoid interaction between the two neighbouring images. Then the intermediates for CO, HCOOH, H2, CH4, CH3OH and C2 products were adsorbed on the active centres. All the atoms were relaxed to release the internal stress of systems. In addition, an electric field was applied to consider its effect on the binding geometry and energy of the adsorbates.
The formation energy of the active centres was estimated by:
Î”ğ¸f=ğ¸sysâˆ’ğ‘›Câˆ—ğœ‡Câˆ’ğ‘›Nâˆ—ğœ‡Nâˆ’ğ‘›Hâˆ—ğœ‡Hâˆ’ğœ‡ğ¶u
(5)
where Esys was the total energy of each system; nC,nN and nH denoted the amounts of C, N and H atoms in each system; Î¼C, Î¼N and Î¼H were the chemical potentials of C, N and H atoms, denoting the energy of per C, N and H atom in graphite, N2 gas and H2 gas.
The adsorption energy Î”Eads of each intermediate on the surface of substrates was defined as:
Î”ğ¸ads=ğ¸interâˆ—âˆ’(ğ¸âˆ—+ğ¸inter)
(6)
where the subscripts inter* and * denoted the adsorption of intermediate on substrates and the bare substrates, Einter denoted the energy of intermediate.
The Gibbs free energy (Î”G) of each chemical reaction without external electric field (i.e., Eâ€‰=â€‰0) was calculated by:
Î”ğºğ‘ğ‘‘ğ‘ ğ¸=0=Î”ğ¸ğ‘ğ‘‘ğ‘ ğ¸=0+Î”ğ‘ğ‘ƒğ¸âˆ’ğ‘‡Î”ğ‘†
(7)
where Î”Eads, Î”ZPE, T and Î”S denoted the calculated total energy, zero point energy, temperature and entropy, respectively. Then, the energy was corrected for the RHE scale and electric field by49,50:
Î”ğºğ‘ğ‘‘ğ‘ =Î”ğºğ‘ğ‘‘ğ‘ ğ¸=0+ğœ‡ğ¸âˆ’ğ›¼ğ¸22âˆ’ğ‘›ğ‘’ğ‘ˆğ‘…ğ»ğ¸
(8)
where E was the external electric field, ne was the amount of protonâ€“electron pairs transferred in each chemical reaction, URHE was the applied potential vs. RHE. The values of Î¼ and Î± were acquired from the second-order polynomial fitting for the energy of each adsorbate across the range of electric fields (see Supplementary Fig. 48). The external electric field was pH-dependent on an RHE scale and estimated by:
ğ¸=ğ¶ğ»(ğ‘ˆğ‘…ğ»ğ¸âˆ’ğœ…ğµğ‘‡ln(10)Ã—ğ‘ğ»âˆ’ğ‘ˆğ‘ƒğ‘ğ¶/ğœ–ğœ–0
(9)
where CHâ€‰=â€‰20â€‰Î¼Fâ€‰cmâˆ’2 refers to Helmholtz capacitance for archetypical carbon material66, Ïµ0 refers to vacuum permittivity (8.85 Ã—10âˆ’12â€‰Fâ€‰mâˆ’1), Ïµâ€‰=â€‰2 refers to dielectric constant of water near a surface (unitless), kB is the Boltzmann constant, Tâ€‰=â€‰300â€‰K and pHâ€‰=â€‰7.3 is kept consistent with experimental conditions, UPZCâ€‰=â€‰0.22â€‰V refers to the potential at the point of zero charge (PZC) vs. SHE, which was measured by differential capacitance curves using impedance methods67.
Five electrochemical reaction pathways were considered47,48,68:
(I)
For formate production,
âˆ—+CO2+eâˆ’â†’âˆ—Oâˆ’OC
(10)
âˆ—OCOâˆ’+H+â†’âˆ—OCHO
(11)
âˆ—OCHO+(H++ğ‘’âˆ’)â†’âˆ—HCOOH
(12)
âˆ—HCOOHâ†’âˆ—+HCOOH
(13)
(II)
For formate production,
âˆ—+(H++ğ‘’âˆ’)â†’âˆ—H
(14)
âˆ—H+CO2â†’âˆ—OCHO
(15)
âˆ—OCHO+(H++ğ‘’âˆ’)â†’âˆ—HCOOH
(16)
âˆ—HCOOHâ†’âˆ—+HCOOH
(17)
(III)
For CO production,
âˆ—+CO2+eâˆ’â†’âˆ—Oâˆ’CO
(18)
âˆ—Oâˆ’CO+H+â†’âˆ—COOH
(19)
âˆ—COOH+(H++eâˆ’)â†’âˆ—CO+H2O
(20)
âˆ—COâ†’âˆ—+CO
(21)
(IV)
For CO production,
âˆ—+(H++ğ‘’âˆ’)â†’âˆ—H
(22)
âˆ—H+CO2â†’âˆ—COOH
(23)
âˆ—COOH+(H++ğ‘’âˆ’)â†’âˆ—CO+H2O
(24)
âˆ—COâ†’âˆ—+CO
(25)
(V)
For hydrogen evolution reaction,
âˆ—+(H++ğ‘’âˆ’)â†’âˆ—H
(26)
âˆ—+(H++ğ‘’âˆ’)â†’âˆ—H
(27)
The electrochemical reaction steps for CH4, CH3OH and C2 productions were referred to in the published works48,60.
"""

main_material = "PSB-CuN3"
```

## 2 Pyridine-N-rich Cu single-atom catalyst boosts nitrate electroreduction to ammonia

```python
text = """
1. Introduction
Ammonia (NH3), as an important raw material for fertilizers, pharmaceuticals, and chemical industries, is also considered a renewable energy carrier with the advantages of no carbon, high energy density (22.5 MJ kgâˆ’1), and easy liquefaction [1], [2], [3]. Moreover, the conventional Haberâ€“Bosch (Hâ€“B) process has dominated the synthetic ammonia industry and contributed to the rapid growth of the global population over the past century [4]. However, the H-B process requires high temperatures (400â€“500 Â°C) and high pressures (100â€“200 atm) to convert nitrogen (N2) and hydrogen (H2) into ammonia, resulting in annual energy consumption of about 5.5 EJ and greenhouse gas emissions of more than 450 million metric tons [5], [6]. Due to the above defects of the Hâ€“B process, electrochemical N2 reduction to NH3 using H2O instead of H2 as a proton source has recently attracted significant attention owing to its mild reaction conditions [7], [8]. Even so, the high dissociation energy of the Ntriple bondN bond (941 kJ molâˆ’1) and limited water solubility of N2 leads to the aqueous-based N2 reduction reaction suffering from inferior efficiency and yield [9], [10], [11]. In contrast, nitrate (NO3-) is more active than N2 for NH3 production due to the relatively lower dissociation energy of the N = O bond (204 kJ molâˆ’1), outstanding solubility, and favorable thermodynamic potential (0.69 V vs. RHE) [12]. Meanwhile, abundant NO3- ions from agricultural and industrial discharge into the global hydrological environment may result in a series of hazards such as human diseases, eutrophication, ecosystem disruption, and even the imbalanced global nitrogen cycle [13], [14], [15]. Therefore, the reduction of NO3- to NH3 can promote the solution of environmental problems caused by NO3- pollution and hopefully can be an alternative to the traditional ammonia synthesis process.
Electrochemical nitrate reduction to ammonia reaction involves a nine-proton coupled eight-electron transfer process, which obviously lowers the kinetic rate and leads to complex intermediates (e.g., NO2-) [12], [16], [17], [18]. Moreover, the sluggish kinetic rate results in a considerable overpotential of NO3RR, and thus the practical working potentials of NO3RR are considerably overlaid with that of the hydrogen evolution reaction (HER) [19], [20], [21], [22]. Therefore, the fierce competition between water/protons and reactants is inevitable during electrochemical NO3RR. To this end, tremendous efforts have been devoted to tailoring structures of electrocatalysts from nano to atomic scale [23], [24], [25], [26]. Compared with nanocatalysts, single-atom catalysts (SACs) demonstrate great potential in NO3--to-NH3 conversion due to their maximum atomic utilization efficiency, unsaturated coordination environment, and tunable electronic structure [27], [28]. Copper-based catalysts exhibit high catalytic activity for NO3RR since Cu atoms with 3d10 orbital are more favorable for NO3- adsorption [29], [30]. Recently, N-doped carbon-based materials are used to anchor copper to achieve uniform dispersion of Cu atoms, serving as Cu-N-C single-atom catalysts [31], [32], [33], [34]. Benefiting from the tunable structure of Cu-N-C SACs, regulating the local coordination environment of the Cu atom is an effective way to optimize the catalytic performance. Some previous reports have made substantial efforts to tune the metal-N-C construction from various dimensions including coordination number, first-shell and second-shell heteroatom doping, dual metal sites, and inter-site distance [24], [35], [36], [37], [38], [39]. Besides, direct regulation of the metal-coordinated N structure is a recently developed strategy, which has been studied in the oxygen reduction reaction (ORR) and the CO2 reduction reaction (CO2RR) [40], [41]. However, for the NO3RR, the effect of the N structure on the selectivity of NH3 and the competing HER is still unclear.
Herein, we report a pyridinic-N-rich coordinated copper single-atom catalyst (PR-CuNC) derived from three-dimensional (3D) semi-interpenetrating polypyrrole-polyethyleneimine (PPy-PEI) hydrogels with adsorbing Cu ions. Without introducing PEI, the structure of coordinated N was transformed to the relatively low pyridinic-N coordinated Cu (CuNC). PR-CuNC exhibits impressive Faraday efficiency of 94.61 % and a yield rate of 130.71 mgNH3 mgCuâˆ’1 hâˆ’1 for electrocatalytic NO3RR, while the activity of CuNC is unsatisfactory. Theoretical evidence suggests that the electronic structure change originating from the Cu-N configuration is responsible for the enhanced catalytic performance. The favorable NO3- adsorption and the lower energy barrier in the nitrate-to-ammonia reduction reaction of Cu-pyridinic-N4 further support the experimental phenomenon from a thermodynamic perspective. The findings in this work highlight a new way to directly regulate the local nitrogen coordination environment for optimizing NO3RR activity at the molecular level.
2. Experimental
2.1. Preparation of PEI-PPy hydrogels and PPy hydrogels
The semi-interpenetrating PEI-PPy hydrogels were prepared by the polymerization of pyrrole by introducing PEI in this strategy. Briefly, pyrrole monomers (0.3 M, 416 Î¼L), PEI (0.101 g), and sodium dodecyl benzene sulfonate (SDBS, 0.1 M, 0.697 g) were dissolved in HCl aqueous solutions (1.0 M, 10 mL) and marked as solution A. Solution B was prepared by adding ammonium persulphate (APS, 0.3 M, 1.397 g) in HCl aqueous solutions (1.0 M, 10 mL). After cooling to 2 Â°C, the above two solutions were rapidly mixed and left overnight at 2 Â°C to ensure complete polymerization. Finally, the as-prepared PEI-PPy hydrogels were alternately washed with deionized (DI) water and anhydrous ethanol several times to eliminate unreacted residual impurities and by-products. As a comparison, PPy hydrogels were also prepared in a similar process, in which PEI was absent in solution A.
2.2. Preparation of PR-CuNC and CuNC
The purified PEI-PPy hydrogels were added into copper chloride (0.2 M) aqueous solution with a continuous stirring for 12 h. Then, the PEI-PPy hydrogels with copper anions were washed several times with DI water and were treated by a freeze-drying process. Afterward, the obtained samples were ground into uniform powders and then calcined at 800 Â°C under flowing argon (Ar) for 2 h with a heating rate of 10 Â°C minâˆ’1. After naturally cooling to room temperature, the powder products were further immersed in HCl solution (0.5 M) for 4 h at 80 Â°C to remove the unstable and inactivated copper clusters and particles. The samples were subsequently washed with DI water until the pH value reached 7 and dried in a vacuum at 80 Â°C. A second calcination was conducted under the same conditions and finally produced PR-CuNC. In contrast, CuNC was prepared via the same process for PR-CuNC unless PPy hydrogels were used. The catalyst (denoted as NC) derived from PPy-PEI hydrogel was prepared following the same process except for the absence of Cu ions adsorption.
2.3. Characterization
Scanning electron microscopy (SEM) was performed on a Helios G4 UC (Thermo Fisher Scientific, USA). High-angle annular dark-field scanning transmission electron microscopy (HAADF-STEM) images were obtained with a JEOL NEOARM probe-corrected transmission electron microscope with aberration correction (Japan). X-ray diffraction (XRD) patterns were recorded with a Rigaku Ultima IV X-ray diffraction system (Japan) using Cu Ka radiation (40 kV / 40 mA, 10Â° minâˆ’1 from 5Â° to 90Â°). X-ray photoelectron spectra (XPS) were acquired with Thermo Scientific K-Alpha (USA); all binding energies were referenced to the C 1 s peak at 284.6 eV to correct the possible shift. X-ray absorption fine structure analysis (XAFS) based on synchrotron radiation was conducted at the BL14W1 beamline of the Shanghai Synchrotron Radiation Facility (China). The corresponding X-ray absorption near edge structure (XANES) and the extended X-ray absorption fine structure (EXAFS) were processed by using the IFEFFIT program package. The content of copper in the samples was determined by inductively coupled plasma atomic emission spectrometry (ICP-OES, Agilent 5110, USA). Ultraviolet-visible (UVâ€“vis) absorption spectra were recorded using a Hitachi UV-2900 spectrophotometer (Japan). An online gas chromatography (GC 9790 II, FULI, China) equipped with a flame ionization detector (FID) and thermal conductivity detector (TCD) was used to monitor the gaseous products. All 1H nuclear magnetic resonance (1H NMR) spectra were collected on a Bruker Avance NEO 600 MHz (GER) with water suppression.
2.4. Electrochemical measurements
Electrochemical measurements were conducted on a DH7003 potentiostat (Donghua test, China) equipped with an H-type quartz electrochemical cell separated by Nafion 115 membrane. Before all tests, the Nafion 115 membranes with a size of 3.5 cm Ã— 3.5 cm were immersed in hydrogen peroxide solution (5 %) at 80 Â°C for 1 h and ultrapure water at 80 Â°C for another 1 h, respectively. The hydrophobic carbon paper was soaked in ethanol and DI water for 15 min with ultrasound treatment to remove surface impurities. The catalyst ink was prepared by adding 3 mg catalyst powder and 20 Î¼L Nafion solution (5 wt%, Dupont, USA) into the mixture of DI water (360 Î¼L) and isopropyl alcohol (120 Î¼L) with ultrasonic treatment for 30 mins. Subsequently, homogeneous catalyst ink (20 Î¼L) was drop-cast on hydrophobic carbon paper with an active area of 0.25 cm2 and a mass loading of 0.48 mg cmâˆ’2. After the other side of the carbon paper without loading catalysts was sealed with an insulating hot-melt adhesive, the final working electrode was made. In addition, a saturated calomel electrode (SCE, 1 M KCl) and a graphite rod were used as the reference electrode and counter electrode, respectively. All potentials referenced to the reversible hydrogen electrodes (RHE) were obtained by the following calibration equation: E(RHE) = E(SCE) + 0.059 Ã— pH + 0.253 V. Prior to electrochemical measurements, pure Ar-gas flow was injected into the 0.1 M KOH with or without 0.1 M NO3- electrolytes for 30 mins to remove oxygen and nitrogen. All electrodes were activated by conducting cyclic voltammetry (CV) tests from +0.5 V to â€“0.9 V vs. RHE for 10 cycles at a scan rate of 100 mV sâˆ’1. Linear sweep voltammograms (LSVs) were acquired from + 0.5 to â€“0.9 V vs. RHE at a scan rate of 10 mV sâˆ’1. The current-time (I-t) curves were recorded by the chronoamperometry methods, which were performed at different given potentials for 30 mins and the corresponding electrolytes were collected to calculate the yield rate and Faraday efficiency of NH3. The electrochemical impedance spectroscopy (EIS) was performed at â€“0.2 V vs. RHE from 105 to 10âˆ’1 Hz.
3. Results and discussion
3.1. The preparation and characterization
PR-CuNC was prepared through the pyrolysis of the PPy-PEI hydrogels with adsorbing Cu ions as illustrated in Scheme 1. Specifically, PEI was introduced in the polymerization of pyrrole to form PPy-PEI hydrogels, where PPy chains were uniformly interpenetrated with PEI chains. Benefiting from abundant primary and secondary amine groups of PEI, PPy-PEI hydrogels show the enhanced capability to anchor Cu ions compared with PPy only. However, the N site (-+NH-) on the pyrrole five-membered ring is the only source of active N sites in PPy and is not conducive to anchoring metal atoms. Cross-linking pyrrole with PEI effectively improved scarce nitrogen active sites and provided diverse coordination environments. Besides, PPy-PEI hydrogels possess a unique 3D semi-interconnected nanostructure with abundant N sites and large surface area, facilitating Cu ions adsorption and preventing aggregated Cu particles during subsequent carbonization. Finally, copper-adsorbed PPy-PEI hydrogels were calcined to produce a PR-CuNC in which the highly dispersed Cu atoms were anchored on the hydrogel-derived carbon substrate.
Scheme 1
Download: Download high-res image (254KB)
Download: Download full-size image
Scheme 1. The synthetic route of the PR-CuNC.

As shown in Fig. 1a, the PR-CuNC possesses a porous granular structure composed of interconnected nanoparticles, which is similar to that of CuNC and NC (Fig. S1). In Fig. S2, the XRD pattern of PR-CuNC displays a broad peak at 25.6Â° corresponding to the (002) plane of graphitic carbon, while no diffraction peaks from Cu-based crystal phases are observed. Thus, there are no aggregated Cu-based compounds existing in PR-CuNC, and Cu species are mainly amorphous. Additionally, CuNC shows similar XRD features as PR-CuNC (Fig. S2). Besides, the Brunauerâ€“Emmettâ€“Teller (BET) surface area of PR-CuNC (240.58â€‰m2 gâˆ’1) is similar to that of CuNC (277.72â€‰m2 gâˆ’1, Table S1). Based on the above characterization results, it can be concluded that the introduction of PEI has little effect on the morphology and crystal plane structures, eliminating the influence of carbon substrate on material properties. High-angle annular dark field scanning transmission electron microscopy (HAADF-STEM) with aberration correction was carried out to identify the distribution of atoms. As depicted in Fig. 1b, the HAADF-STEM image of PR-CuNC exhibits light spots dispersed on the edge and surface of the carbon substrate. Since the atom mass of the Cu element is much larger than that of other elements (C, N, O, see the XPS results in Fig. S3) in the catalyst, it can be determined that the light spots represent isolated Cu single atoms. Energy dispersive X-ray spectroscopy (EDS) mapping analysis further confirms the homogeneous dispersion of Cu, N, and C elements in PR-CuNC, suggesting that Cu atoms were uniformly distributed on the nitrogen-doped carbon substrate.
Fig. 1
Download: Download high-res image (548KB)
Download: Download full-size image
Fig. 1. Morphology, structure, and compositions. (a) SEM, (b) HAADF-STEM, and (c) EDS element mapping images of PR-CuNC. (d) N 1s XPS spectra of PR-CuNC and CuNC. (e) Normalized XANES of Cu K-edge, derivatives of XANES (inset), and (f) EXAFS of Cu foil, Cu2O, and CuO and PR-CuNC. WT plots of Cu K-edge EXAFS spectra for (g) Cu foil, (h) Cu2O, and (i) PR-CuNC.

X-ray photoelectron spectroscopy (XPS) measurement was conducted to determine the elemental composition and valence state of catalysts. According to XPS survey spectra (Fig. S3), the elemental composition and atomic ratios of PR-CuNC, CuNC, and NC were confirmed and summarized in Table S2. The atomic ratio of the N element in PR-CuNC is 8.21 % higher than that in CuNC (6.48 %), which can be attributed to the presence of PEI with rich amine groups. As shown in Fig. 1d, the high-resolution N 1s XPS spectra for PR-CuNC and CuNC can be deconvoluted into four peaks ascribed to pyridinic-N (398.2â€‰eV), pyrrolic-N (399.3â€‰eV), graphite-N (400.8â€‰eV) and oxidized-N (401.7â€‰eV) [41], [42], [43]. The N 1s XPS spectrum of NC is depicted in Fig. S4. The percentages of four nitrogen species are summarized in Table S3 and S4, where PR-CuNC and CuNC both show similar contents of graphite-N and oxidized-N. However, the absolute content of pyridinic-N on PR-CuNC is 3.20 % exceeding that on CuNC (1.99 %), while PR-CuNC shows a lower absolute content of pyrrolic-N (0.76 %) than CuNC (1.16 %). It suggests that the introduction of PEI successfully boosts the pyridinic-N ratio, forming more Cu-pyridinic-N coordination configuration in PR-CuNC. For the Cu 2p XPS spectrum, PR-CuNC (Fig. S5a) exhibits double characteristic peaks of Cu 2p3/2 and Cu 2p1/2, which both are deconvoluted into two peaks corresponding to Cu2+ species (934.55 and 954.40â€‰eV) and Cu1+/Cu0 species (932.40 and 952.14â€‰eV) [44]. The Cu 2p XPS spectrum for CuNC in Fig. S5b shows similar peaks assigned to Cu2+ and Cu1+/Cu0, respectively [45]. The accurate Cu content is further determined by inductive coupled plasma-optical emission spectroscopy (ICP-OES) in Table S5, where PR-CuNC (6.01â€‰wt%) is higher than CuNC (3.69â€‰wt%) probably owing to the strong binding ability of PEI.
Due to the accuracy of XPS being insufficient for analyzing Cu species with low content, the chemical and atomic structure of Cu sites in PR-CuNC were further investigated by synchrotron X-ray absorption spectroscopy (XAS). Fig. 1e illustrates Cu K-edge X-ray absorption near edge structure (XANES) spectra of PR-CuNC and several reference samples. In comparison to Cu2O and Cu foil, the adsorption edge for PR-CuNC is located at a higher energy close to CuO, indicating Cu atoms are in an oxidized state. The first derivative of PR-CuNC is between that of Cu2O and CuO (in the inset of Fig. 1e), which confirms that the valence state of Cu for PR-CuNC is between +1 and +2, consistent with the observation from the Cu 2p XPS of PR-CuNC [21], [28]. The Fourier-transformed X-ray absorption fine structure (FT-EXAFS) spectra were obtained to verify the accurate structure of Cu atoms in PR-CuNC displayed in Fig. 1f. Cu foil reveals a main peak at around 2.24â€‰Ã… corresponding to the Cu-Cu bond. In contrast, PR-CuNC only shows a single peak centered at 1.45â€‰Ã… assigned to the Cuâ€“N bond, differing from the Cu-O bond in CuO and Cu2O [46]. It confirms that Cu atoms in PR-CuNC are separated from each other and coordinated with first-shell N atoms, which is in agreement with the result of the HAADF-STEM. Besides, the best fitting results of EXAFS k and R-space provide detailed structural parameters in Fig. S6 and Table S6. According to the coordination number of 4.54, the Cu atom was bonded by four N atoms with Cu-N bonds, implying the Cu-N4 configurations in PR-CuNC. On the contrary, the model of Cu coordinated with pyrrolic nitrogen shows worse fitting results in Fig. S7, indicating that Cu atoms in PR-CuNC were more coordinated with pyridinic-N, which is consistent with XPS results. Moreover, wavelet transfer (WT) plots in Fig. 1gâ€“i were applied to better visualize the Cu-K edge EXAFS oscillations. Cu foil depicts an intensity maximum at around 8â€‰Ã…âˆ’1 belonging to the Cu-Cu bond. In contrast, the only intensity maximum of PR-CuNC is observed at 4â€‰Ã…âˆ’1 assigning to the Cu-N bond, which slightly differs in the Cu-O bond (about 6â€‰Ã…âˆ’1) in Cu2O [47].
3.2. Electrochemical performance for the NO3RR
The electrochemical measurements for NO3- reduction reaction were conducted in H-cell using Ar-saturated 0.1â€‰M KOH with and without 0.1â€‰M KNO3- as electrolytes. Fig. 2a shows the LSV curves for PR-CuNC, CuNC, and NC. After adding NO3-, the current densities for all samples improve dramatically, demonstrating the occurrences of the NO3RR. Compared with CuNC and NC, PR-CuNC reaches a relatively high current density of 90â€‰mAâ€‰cmâˆ’2 at â€“0.9â€‰V vs. RHE and a low onset potential of +â€‰0.1â€‰V vs. RHE, indicating superior activity for nitrate reduction. Moreover, a small reduction peak at 0.0â€‰V vs. RHE and another wide reduction peak at around â€“0.5â€‰V vs. RHE belong to the NO3--to-NO2- reduction and NO2--to-NH3 process [21], [48]. Additionally, EIS tests were conducted on PR-CuNC and CuNC to measure the charge-transfer resistances during the electrochemical NO3RR process. As shown in Fig. S8, the Nyquist plots displays PR-CuNC with a smaller semicircle diameter and charge transfer resistance (23.21â€‰Î©) than that of CuNC (42.29â€‰Î©) at â€“0.2â€‰V vs. RHE, suggesting a faster kinetics rate for the NO3RR.
Fig. 2
Download: Download high-res image (437KB)
Download: Download full-size image
Fig. 2. Electrocatalytic performance of different catalysts for NO3RR. (a) LSVs of PR-CuNC, CuNC, and NC with and without NO3-. (b) FE of NH3 for PR-CuNC, CuNC and NC. (c) FE of NH3, NO2- and H2 for PR-CuNC. (d) YRA for PR-CuNC, CuNC and NC. (e) YRCu for PR-CuNC and CuNC. (f) The catalytic durability of PR-CuNC.

To further evaluate the NO3RR performance, current densityâ€“time (Iâ€“t) curves of all samples from â€“0.1 to â€“0.6â€‰V vs. RHE were recorded (Fig. S9). Then, the amount of produced NH3 in electrolytes after I-t tests were determined by the indophenol blue spectrophotometric method for calculating Faraday efficiency (FE, Fig. S10). As shown in Fig. 2b, PR-CuNC outperforms all catalysts at almost all potentials, reaching a maximum FE of 94.61 %. In contrast, CuNC displays inferior activity with a maximum FE of 76.99 %, implying that rich Cu-pyridinic-N coordination configuration in PR-CuNC significantly enhances the NO3RR activity. Furthermore, Cu-NC catalysts with higher or lower pyridinic-N content (denoted as HP-Cu-NC and LP-Cu-NC) were obtained by introducing PEI with different concentrations (Fig. S11 and Table S7). Electrochemical nitrate reduction experiments in a 0.1â€‰M KOH and 0.1â€‰M KNO3 electrolyte showed that as the pyridinic-N content increased, the FE also increased, supporting our previous interpretation (Fig. S12). While NC has almost no catalytic capacity for NO3RR, the NH3 production is nearly negligible at less than 30 %, suggesting that Cu sites are the dominant active sites remarkably contributing to the nitrate-to-ammonia conversion. When the applied potential for PR-CuNC is below â€“0.5â€‰V vs. RHE, the Faraday efficiency of NH3 decreases even though the total current density continues to increase, indicating that other side reactions weaken the ammonia production rate. The complex NO3RR process is associated with a variety of nitrogenous byproducts such as NO2-, N2, N2H4, and NOx, accompanied by fierce competition with the HER. It is therefore necessary to detect the FE of major products (NO2- and H2) for further evaluation of selectivity. As depicted in Fig. 2c, NO2- is the primary product at low overpotential for PR-CuNC. With applied a more negative potential, the FE of NO2- gradually decreased, which can be contributed to the NO2--to-NH3 reduction wide peak at near â€“0.5â€‰V vs. RHE (Fig. 2a). Moreover, the competitive HER affects NH3 selectivity from â€“0.4â€‰V, and the maximum Faraday efficiency reaches 6.92 % at â€“0.6â€‰V on PR-CuNC. The same product analysis of CuNC and NC was conducted and shown in Fig. S13. As with PR-CuNC, the major byproduct of CuNC and NC changes from NO2- and H2 as the overpotential rises, while the FE values of H2 in two samples are much higher, reaching 22.38 % and 17.65 % at â€“0.6â€‰V vs. RHE, respectively. Consequently, pyridinic-N-rich Cu sites in PR-CuNC can effectively suppress the HER and NO2- formation during high overpotential, which greatly facilitates the NH3 generation from NO3-. Meanwhile, the yield rates based on the geometric area of electrodes (YRA) (see details in supporting information) of different catalysts were detected at varying potentials (Fig. 2d). PR-CuNC demonstrates a much higher YRA of 3.74â€‰mgNH3 hâˆ’1 cmâˆ’2 at â€“0.5â€‰V vs. RHE than that of CuNC (1.10â€‰mgNH3 hâˆ’1 cmâˆ’2) and NC (0.08â€‰mgNH3 hâˆ’1 cmâˆ’2). To avoid the contribution from the Cu content, the YR was normalized by the mass of Cu labeled as YRCu in PR-CuNC and CuNC (see details in supporting information). As shown in Fig. 2e, PR-CuNC achieves 130.71â€‰mgNH3 mgCuâˆ’1 hâˆ’1 at â€“0.5â€‰V vs. RHE, which outperforms CuNC with the YRCu of 71.87â€‰mgNH3 mgCuâˆ’1 hâˆ’1. Combining the XPS results, the enhanced FE and YR of PR-CuNC over CuNC could be ascribed to the increased pyridinic-N structure. Notably, the NO3RR performance of PR-CuNC at a relatively low overpotential is superior to most reported single-atom catalysts which be summarized in Table S8.
The catalyst stability during the electrochemical process of NO3RR is a crucial factor in measuring the quality and applicability of catalysts. Therefore, ten consecutive electrolysis experiments were carried out at â€“0.5â€‰V vs. RHE using the same electrode and further calculated the FE and YRCu to investigate the durability of PR-CuNC. After 10 cycles, FE is consistently higher than 85 % and YR remains stable above 120â€‰mgNH3 mgCuâˆ’1 hâˆ’1 (Fig. 2f), demonstrating convincingly the excellent electrochemical recyclability of PR-CuNC. The initial improvement of catalytic performance might be attributed to the catalyst on the electrode surface being fully activated after electrolysis and then gradually reaching a steady state (Fig. S14). In addition to electrochemical stability, the structural stability of the PR-CuNC was also evaluated. As shown in Fig. S15, SEM images showed that the catalyst after tests still exhibited a porous granular structure, which was similar to PR-CuNC before tests. Besides, no aggregation of Cu atoms was observed in TEM images (Fig. S16 and S17), suggesting the morphology of the catalyst barely changed. According to XPS survey spectra (Fig. S18), the area of the double characteristic peaks assigned to Cu1+/Cu0 increased after NO3RR tests, while that of Cu2+ became smaller, implying the reduction of Cu2+ to Cu1+/Cu0 during the NO3RR.
Additionally, the 15N isotope labeling proton nuclear magnetic resource (1H NMR) was performed to verify the precision of the NH3 measurements and eliminate the possible ammonia contamination from the environment. Electrochemical tests were conducted using 14NO3- and 15NO3- as nitrate sources, respectively. Fig. S19a exhibits different 1H NMR spectra of ammonia synthesized from different nitrate sources. 14NH4+ owns triple peaks with a spacing of 52â€‰Hz, while 15NH4+ shows characteristic double peaks with a spacing of 72â€‰Hz. Meanwhile, maleic acid was used as an internal standard for quantifying the generated ammonia. As shown in Fig. S19b, the yield of produced 15NH4+ (17.04 Î¼mol) is very close to that of 14NH4+ (16.92 Î¼mol), implying that the NH3 determined in this work only derived from the electrocatalytic nitrate process. Moreover, the amount of ammonia quantified by 1H NMR (16.92 Î¼mol) is comparable to that detected by the UV-Vis method (16.96 Î¼mol), confirming the reliability of the results from indophenol blue spectrophotometry.
3.3. Theoretical understanding of the catalytic mechanism
To further understand the mechanistic insights of the superior nitrate selectivity exhibited by the pyridinic-N-rich Cu SACs, density functional theory (DFT) calculations were conducted to investigate the structure-activity relationship. XANES and EXAFS results indicate the center Cu atom is isolated and coordinated to the surrounding N. Herein, we constructed two Cu-N4 catalyst models, in which Cu single atoms are coordinated with four pyridinic-N (Cu-pyridinic-N4) and four pyrrolic-N (Cu-pyrrolic-N4), respectively. The charge density differences of the above two Cu-N4 models are illustrated in Fig. 3aâ€“b, which depict obviously different charge distributions. Compared to Cu-pyrrolic-N4, Cu-pyridinic-N4 exhibits a change in electron depletion around the Cu atom and a stronger electron accumulation around C atoms bonded to N. Moreover, the partial densities of states (PDOS) of Cu-pyridinic-N4 and Cu-pyrrolic-N4 were presented in Fig. 3c. Cu d orbital of Cu-pyridinic-N4 has obvious energy level splitting near the Fermi level and the negative shift of the active center compared to that of Cu-pyrrolic-N4. Due to apparent hybridization between Cu d and N p orbitals as well as the higher overlap regions of electron density at the bonding orbitals, the Cu-pyridinic-N4 type is more favorable to electron transfer and is more active for adsorbed NO3-. The different distributions of state density demonstrate that the electronic structure of Cu-N4 is regulated by the types of coordination N.
Fig. 3
Download: Download high-res image (290KB)
Download: Download full-size image
Fig. 3. DFT calculations. The charge density difference of (a) Cu-pyrrolic-N4 and (b) Cu-pyridinic-N4. (blue: accumulated electron density; yellow: depleted electron density; the isosurface value: 0.002 e Ã…âˆ’3). (c) PDOS of Cu d and N p of the Cu-pyrrolic-N4 and Cu-pyridinic-N4 sites. (d) The Gibbs free energy diagram of nitrate reduction to ammonia and water dissociation (the top right) on Cu-pyrrolic-N4 and Cu-pyridinic-N4 sites.

The catalytic details for NO3RR are obtained by the Gibbs free energy diagram. The element steps including nine elementary steps and eight intermediates as considered based on the previously reported mechanism [1], [12], [49]. Moreover, the geometry structures of all intermediates generated on the Cu-N4 were optimized before thermodynamic calculations (as the bottom of Fig. 3d). The first step in adsorbing NO3- to give *NO3 is considered crucial, as it matters whether the reaction can begin spontaneously. As shown in Fig. 3d, the adsorption energy of NO3- (Î”GNO3) substantially declines to â€“2.06â€‰eV on Cu-pyridinic-N4, while the Î”GNO3 declines to â€“2.53â€‰eV on Cu-pyrrolic-N4, indicating that Cu-pyridinic-N4 was more advantageous for NO3- adsorption. Then, as two N-O bonds on *NO3 break in sequence to form *NO2 and *NO, the energy gradually decreases. However, a large energy barrier needs to be overcome in the subsequent hydrogenation process, referring to the step from *NO to *NHO. Cu-pyridinic-N4 shows a lower Î”G*NHO of +â€‰0.07â€‰eV than Cu-pyrrolic-N4 (+0.51â€‰eV), implying that Cu-pyridinic-N4 is more active for the protonation of *NO. Subsequently, two successive protonation steps of *NHO and cleavage of N-O bonds generate *NH2, followed by a hydrogenation process to obtain adsorbed *NH3. Finally, *NH3 undergoes a desorption step from the catalyst surface to produce the free NH3, completing the whole nitrate reduction reaction to ammonia. Overall, the NO3RR steps are exothermic except for the *NHO formation, which is considered a rate-limiting step. The larger Î”GNO3 and the lower energy barrier of the rate-limiting step indicate that the Cu-pyridinic-N4 structure is more favorable for ammonia synthesis from NO3-. The Gibbs free energy diagrams of HER on Cu-pyridinic-N4 and Cu-pyrrolic-N4 were analyzed in the top right of Fig. 3d. The higher Î”GH2O of Cu-pyridinic-N4 (0.39â€‰eV) than that of Cu-pyrrolic-N4 (0.32â€‰eV), suggesting a more favorable capability for capturing H2O molecules effectively. Furthermore, the decomposition of *H2O is considered a rate-determining step, where the larger energy barrier (Î”G=0.52â€‰eV) observed on the Cu-pyridinic-N4 demonstrates the suppressed HER compared with the Cu-pyrrolic-N4 surface (Î”G=0.35â€‰eV).
4. Conclusion
In summary, a hydrogel-derived pyridinic-N-rich Cu single-atom catalyst was synthesized in this work by successfully regulating the structure of coordination N atoms. The absolute content of pyridinic-N increases from 1.99 % to 3.20 % by forming semi-interpenetrating PPy-PEI hydrogels with abundant N sites. Efficient NO3--to-NH3 reduction was achieved on PR-CuNC at low overpotentials, reaching a maximal Faraday efficiency of 94.61 % and a yield rate of 130.71â€‰mgNH3 mgCuâˆ’1 hâˆ’1. According to NO3- reduction experiments and DFT calculations, we disclose that the high NO3RR activity and selectivity for NH3 of PR-CuNC originated from the single-atom Cu sites coordinated with four pyridinic-N atoms. Importantly, the different charge density differences and PDOS of Cu-pyrrolic-N4 and Cu-pyridinic-N4 demonstrate the change in electron structure between Cu and N atoms. Compared to Cu-pyrrolic-N4, the Cu-pyridinic-N4 has a more advantageous NO3- adsorption ability and lower energy barrier for protonating *NO to *NHO during the NO3RR reaction steps. This study highlights a strategy to directly regulate the local nitrogen microenvironment of single-atom catalysts at the molecular scale. This work also provides a deeper understanding of the essential relationship between the coordination environment of the atomic site and catalytic behavior for various catalytic reactions.

"""

main_material = "pyridinic-N-rich coordinated copper single-atom catalyst (PR-CuNC)"
```

```python
text = """
Ammonia (NH3), as a clean energy carrier, has been perceived as a promising candidate to replace gaseous hydrogen, given its high gravimetric and volumetric energy densities in a liquid state. (1âˆ’3) NH3 is important not only in the field of energy supply, but it also plays an important role in the development of fertilizers, medications, explosives, dyes, resins, and plastics. (4âˆ’6) One of the most prominent advantages of ammonia is that it can be compressed to a liquid at a pressure as low as 8 bar under room temperature. (6) Giving credit to the huge efforts of previous researchers, a mature system of liquid ammonia storage, transport, and infrastructure has been well developed. However, the traditional NH3 production of the Haberâ€“Bosch process requires high energy input and high pressure, consumes lots of natural gas, and releases greenhouse gas into the environment. (7) Considering carbon neutrality is the future vision for humans, diminishing and even eliminating the carbon footprint is urgently required.
The electrochemical nitrogen reduction reaction (ENRR) has been considered a promising alternative strategy to produce NH3, given its benign reaction conditions, renewable energy source, and environment-friendly reaction precursor (N2 and water). Major efforts have been devoted to investigating the various catalysts which could provide a high yield as well as high selectivity for NH3. (8âˆ’14) Among all types of catalysts, single atom catalysts (SACs) stand out from the crowd due to their distinctive merits, for instance, high atom utilization, exceptional catalytic performance, and well-defined active center. (15âˆ’22) Numerous SACs, such as N-doped carbon, (8,23) Ru, (24,25) Au, (26) Fe, (13,27) and rare earth metals (Y and Sc) (10) have been investigated for the ENRR to produce NH3. Besides varying the different dopants as the active center, the coordination environment (coordinated atom, number, vacancy) also plays an important role in the performance enhancement. (28âˆ’32) Meanwhile, the HER is the representative competitive reaction in the ENRR process. Designing the catalyst with an electron-deficient active center is also a feasible way to improve the performance. (33) However, simultaneously improving the Faradaic efficiency (FE) and yield is still challenging for the single-atom center because of the multiple reaction intermediates during the ENRR process. To deal with this issue, single-atom-dimer (SAD) catalysts with a dual single-atom active center emerge as a promising strategy. Theoretically, various homonuclear and heteronuclear SAD catalysts with metal dimers confined in 2-dimensional (2D) expanded phthalocyanine have been investigated for the NRR. (34) Strain-induced bi-Ti3+ pairs on anatase TiO2 as the active center for N2 fixation have been reported, and the ammonia yield rate was 8.90 Î¼g hâ€“1 cmâ€“2 at âˆ’0.45 V vs the reversible hydrogen electrode (RHE). (16) Not only in the field of the NRR, but SAD catalysts have also been applied in CO2 reduction reactions (CO2RRs) and oxygen reduction reactions (ORRs) as the electrocatalysts and photocatalysts. (17âˆ’19,35,36) Nevertheless, the performances of the ENRRs based on previous research are still insufficient, and more investigations for modifying the electronic structure and the environment of SAD catalysts are still required. As for the choice of metal center for dimer catalysts, vanadium is selected based on the outstanding performance of its bulk materials. Vanadium carbide (VC) with periodic anionic vacancies was reported with the NH3 yield rate of 23.2 Î¼g hâ€“1 mgâ€“1 and FE of 18.3%. (37) Vanadium nitride (VN) showed a yield rate of 1.1 Ã— 10â€“10 mol sâ€“1 cmâ€“2 with a long stability of 116 h. (38) Vanadium phosphide (VP) performed with the yield rate of 8.35 Ã— 10â€“11 mol sâ€“1 cmâ€“2 and the FE of 22%. (39) Vanadiumâ€“nickel oxynitride showed a yield rate of 6.78 Î¼g hâ€“1 cmâ€“2. (40) Vanadium dimer supported on phthalocyanine (V2-Pc) was theoretically proved to effectively suppress the competitive HER. (34) Though the bulk materials of vanadium have been widely investigated for the NRR, the single atom or dimer structure deemed to have better performance has not been stepped into deeply.
Herein, inspired by the above discussion, we present an oxygen-bridged vanadium dimer (V-O-V) catalyst on N-doped carbon (O-V2-NC) configuration for the ENRR. The oxygen bridge over the V dimer results in an electron-deficient active site of V for efficient N2 adsorption, and the V single-atom dimer could further modulate the multiple reaction intermediates during the ENRR process. By using the X-ray absorption fine structure technique (XAFS), in combination with scanning transmission electron microscopy (STEM), high-resolution transmission electron microscopy (HRTEM), and X-ray photoelectron spectroscopy (XPS), the existence of V-O-V was experimentally confirmed. Density functional theory (DFT) calculations found the structural evolution of O-V2-NC during the N2 protonation process, which involves the V-O-V structure splitting into V-O and V during the protonation and regenerating to the original V-O-V after the release of NH3 from the active center. Due to the efficient N2 adsorption and charge transfer induced by oxygen-bridged V-V atoms, O-V2-NC exhibits a considerable FE of 77% at 0 V (vs RHE) and a comparatively high NH3 yield of 26 Î¼g hâ€“1 mgâ€“1 at âˆ’0.4 V (vs RHE). 15N isotope experiments and 1H NMR further proved that NH3 was indeed produced from the NRR.
Results and Discussion
Synthesis and Structural Characterizations
O-V2-NC was synthesized by a simple template-assisted pyrolysis method, as shown in Figure 1. First, melamine cyanurate (MCA), as a template and nitrogen source, was prepared by mixing melamine and cyanuric acid. Then, vanadium ions were trapped by the amino of polyaniline and the reaction proceeded with calcination under high-temperature treatment. Lastly, acid etching was applied to remove nanoparticles to obtain a vanadium single-atom dimer. Details of the synthesis process are described in the Experimental Section. The as-synthesized O-V2-NC material without acid etching shows distinct vanadium oxide peaks, which are matched with PDF#64-9474, while after acid etching, only amorphous carbon peaks appeared, which is similar to the NC sample without any vanadium precursor (Figure 2a). Raman spectra are performed under the excitation wavelength of 532 nm. Two peaks at 1350 and 1580 cmâ€“1 correspond to the D and G bands. The intensity ratios of the D band to G band (ID/IG) for O-V2-NC and NC are 1.042 and 1.078, respectively. A weak broad 2D peak around 2760 cmâ€“1 indicates the poor crystallinity of the samples (Figure S1). The materials prepared by the above method appear as a carbon sheet structure. It is also worth noting that the surface of O-V2-NC is crumpled, and a porous structure is observed after acid treatment (Figure S2). The O-V2-NC clearly presents a porous structure of O-V2-NC. The pore size varies from 2 to 15 nm, and the average pore size is around 5.12 nm (Figure S3). Element mapping images present that vanadium atoms are distributed uniformly over the sample (Figure 2c). Combined with the results of XPS, ICP, and EDS, the average amount of vanadium, around 1.32 wt %, is calculated (Figure S4). Nitrogen adsorption and desorption performance are recorded to investigate the surface and porous properties of the samples. As shown in Figure 2b, the nitrogen take-up of O-V2-NC is much larger than that of NC. O-V2-NC and NC both exhibit an isotherm curve similar to type I. A steep uptake appeared at very low relative pressure, suggesting considerable adsorbentâ€“adsorbate interactions (strong nitrogen adsorption in this case). The strong adsorption capacity is crucial to capture the reactant and make it easier to perform the hydrogenation reaction subsequently. (41,42) An H4 hysteresis loop is observed, which indicates that this catalyst possesses a micropore and plate-like structure. (43) Barrettâ€“Joynerâ€“Halenda (BJH) data are extracted from the adsorption part, and a sharp peak emerges at the pore diameter of 3.9 nm. Brunauerâ€“Emmetâ€“Teller (BET) results further prove that O-V2-NC possesses a higher monolayer nitrogen adsorption. And the surface area of O-V2-NC is higher than that of NC by around 2.5 times (Table S1). This implies the significance of the preparation method for constructing the porosity and high surface area of nanosheets, which enables O-V2-NC to capture N2 more efficiently during the electroreduction process.
Figure 1

Figure 1. Schematic of the synthesis procedure of MCA and O-V2-NC.

Figure 2

Figure 2. (a) XRD patterns of O-V2-NC and NC before and after acid etching. (b) N2 adsorption and desorption plots of O-V2-NC and NC. (c) STEM and EDS mapping of O-V2-NC showing the uniform dispensing of V, O, N, and C. (d) Aberration-corrected HAADF-STEM of O-V2-NC, yellow circle for dimer atom, red circle for a single atom. (e) Intensity profile obtained from (d). (f) Aberration-corrected HAADF-STEM of V2-NC (O-V2-NC after H2 treatment. Conditions: 250 Â°C, 2 h, 5 wt % H2. (g) Intensity profile obtained from (f).

To investigate the microenvironment of vanadium single-atom dimers, XPS and XAFS measurements are conducted. As shown in Figure S5, for N 1s of O-V2-NC, four peaks at 398.29, 399.4, 400.6, and 401.99 can be assigned to pyridinic N, pyrrolic N, graphitic N, and oxidized N, respectively. (12) Comparably, these four peaks are found in the N 1s spectra of NC with a similar peak ratio distribution, which suggests that O-V2-NC and NC possess almost the same nitrogen species and nitrogen environment. The Vâ€“O bonds at 529.5 eV is assigned in the O 1s spectra of O-V2-NC (44) (Figure 3a), and two peaks at 522.94 and 515.75 eV are assigned to the V4+ 2p1/2 and V4+ 2p3/2 (45) (Figure 3b). As shown in Figure 3dâ€“e, not only do the Vâ€“N/Vâ€“O bonds appear in the O-V2-NC sample but the clear Vâ€“V bond is also observed from the EXAFS results. Nevertheless, the Vâ€“V bond of O-V2-NC is longer than that of V foil. To further prove the existence of V-V dimers, Aberration-corrected HAADF-STEM is conducted, the clear V distribution is obtained, and the V dimer is circled by a red rectangle (Figure 2d). Inevitably, single atoms also exist along with the dimer. Nevertheless, the amount of dimer still exceeds that of single atoms. The intensity profile in Figure 2e shows that the distance between Vâ€“V is around 2.70 Ã…, which is consistent with the EXAFS result. Given that there are no nanoparticles or nanoclusters observed in XRD and TEM, the V-V dimers are hypothesized to be in the O-V2-NC configuration. In addition, considering the Vâ€“O bondâ€™s presence and the Vâ€“V bondâ€™s length that is longer than the directly bonded Vâ€“V bond, the V-V dimers are proposed to be bridged by the O atom for the O-V2-NC structure (V-O-V). And the depicted structure is exhibited in Figure S6.
Figure 3

Figure 3. (a, b) XPS spectra of O-V2-NC and V2-NC (a) O 1s and (b) V 2p. (c) Experimental V K-edge XANES spectra of O-V2-NC, V2-NC, and reference samples. The inset zooms in over the half-edge jump feature. (d) Experimental V K-edge FT-EXAFS spectra of O-V2-NC, V2-NC, and reference samples. (e) WT-EXAFS of O-V2-NC, V2-NC, and reference samples at the V K-edge.

To further prove the existence of the V-O-V structure experimentally, the O-V2-NC sample is treated under 250 Â°C for 2 h in a 5% H2 gas environment to remove the bridged O in V-O-V (the sample named V2-NC). As shown in Figure 3a, the samples without H2 treatment clearly show the existence of a Vâ€“O bond at 529.5 eV, even after the Ar etching of 30 s before XPS, while the Vâ€“O bond disappears after the H2 treatment, which indicates that the O is not captured from the atmosphere but uniformly distributed inside the whole structure. Consistently, the V 2p3/2 peak also shifted to lower binding energy from 515.75 to 515.2 eV after H2 treatment, which is attributed to the removal of O from V-O-V (Figure 3b). Figure 3c shows the XANES spectra of various vanadium species, The inset zooms in over the half-edge jump feature. The valence state of V is also correlated from the fitting of the V K-edge XANES energy at the half-edge jump. (46,47) As shown in Figure S7, the oxidation states of V2-NC and O-V2-NC are +4.2 and +4.33, which is consistent with the XPS data that V species mainly exist in the form of V4+. The XAFS analysis also reveals the shift of the bond length of Vâ€“V toward the shorter distance for V2-NC, further demonstrating the structural evolution from V-O-V to V-V (Figure 3d). The aberration-corrected HAADF-STEM shows the uniformly distributed V atoms after the H2 treatment, and the length between the Vâ€“V bond is around 2.49 Ã… (Figure 2fâ€“g), which is also consistent with the XAFS result. Additionally, Figure S8 reveals that the C-O and C-N species still exist in the structure of V2-NC, and the ratio is consistent with that of O-V2-NC, which means that this H2 treatment is a comparatively mild condition and that it does not dramatically change the structure of the carbon substrate. To further prove the structural evolution from V-O-V to V-V, the EXAFS fitting is carried out to obtain the coordination information on V. Figure S9 shows the V K-edge FT-EXAFS of O-V2-NC and V2-NC with the corresponding EXAFS fitting curves. Table S2 summarizes the EXAFS fitting parameters at the V K-edge for O-V2-NC and V2-NC. From this result, the coordination number (CN) of V-N for O-V2-NC and V2-NC is around 3.4, suggesting that each V is coordinated by 3.4 N on average, which is consistent with the predicted structure (Figure S6). Compared with the O-V2-NC, the CN of V-O for V2-NC is reduced to 0.7 and the Vâ€“V bond distance for V2-NC decreased to 2.51 Ã…. That also indicates the structural evolution from V-O-V to V-V. Figure S10 exhibits the XPS spectra of O 1s and V 2p of the V2-NC post structure (the sample of V2-NC is treated at 0 V vs RHE for 30 min under the same NRR environment). The peaks at 522.5 and 515.9 eV are assigned to V4+ 2p1/2 and V4+ 2p3/2. The peaks at 521.2 and 514.1 eV are assigned to V2+ 2p1/2 and V2+ 2p3/2. The peak at 519.6 eV is assigned to the V satellite. The Vâ€“O bond at 529.8 eV appeared again, and the V 2p3/2 peak shifted to the high binding energy of 515.9 eV, which indicates that the O atom attached to the V after the experiment. That further experimentally proves that the V-O-V is the final stabilized structure during the NRR process.
Electrochemical N2 Reduction
The electrochemical NRR experiments were carried out in the H-type cell separated by a Nafion 115 membrane, with 30 mL of 0.1 M HCl as the electrolyte, on the electrochemical workstation (CHI 660C) under room temperature and ambient pressure. Prior to applying any potential, pure N2 was purged into 30 mL of electrolyte for 30 min. The linear sweep voltammetry was scanned at the rate of 5 mVÂ·sâ€“1. The N2 reduction reaction was conducted at 0 V, âˆ’0.1 V, âˆ’0.2 V, âˆ’0.3 V, âˆ’0.4 V, and âˆ’0.5 V for 1 h with continuous N2 feed-up. Unless otherwise noted, all the potentials were calibrated with the reversible hydrogen electrode (RHE). Comprehensive electrochemical measurements were conducted to investigate the electrocatalytic activity of the NRR. Figure 4a shows the linear sweep voltammetry (LSV) curve of O-V2-NC under Ar and N2 feed-up. The current density under N2 saturated conditions increases compared to that under Ar saturated conditions, which means the NRR occurred accompanied by the hydrogen evolution reaction (HER). The indophenol blue method was applied to qualitatively and quantitatively measure the electrocatalytic produced NH3. As shown in Figure S11, a series of standard concentrations of ammonia solution was prepared, and the UVâ€“vis absorbance was recorded. The concentrationâ€“absorbance curve of NH3 with a series of standard concentrations shows a good linear relation (y = 0.445x + 0.0429, R2 = 0.9997). Watt and Chrispâ€™s method was employed to exclude hydrazine production (Figure S12). As shown in Figure 4b, the NRR performance is conducted under different applied potentials from 0 V to âˆ’0.5 V (vs RHE). The NH3 yield rate shows growth with the increase of applied voltage and reaches the highest of 26.12 Î¼g hâ€“1 mgâ€“1 at a voltage of âˆ’0.4 V vs RHE. The highest FE of 77.2% is observed at the voltage of 0 V, which indicates that most energy is used to form ammonia even though with low current density. However, the FE drops down at applied high voltage, which is ascribed to the enhanced HER with the raising of applied voltage, especially under acidic conditions. The chronoamperometry results and UVâ€“vis absorbance of the electrolyte after electrochemical reduction catalyzing by O-V2-NC under different voltages are recorded, as shown in Figure S13. For comparison, the NRR performance of NC is also tested at 0 V, including the NH3 yield rate of 2.3 Î¼g hâ€“1 mgâ€“1 and FE of 23%, which are much lower than those of O-V2-NC (Figure S14). Therefore, V-V dimer atoms play the crucial role of active center to convert N2 to NH3 during the electrocatalytic process, while a comparatively high FE is achieved by NC, which was attributed to the high content of pyridinic and pyrrolic N species. Previous research reported that the high content of pyridinic and pyrrolic N was important for promoting nitrogen reduction to ammonia in nitrogen-doped carbon. (48,8) Consequently, based on this special structure of the carbon matrix, the addition of porosity and V-V dimer atoms in O-V2-NC could more efficiently facilitate the NRR performance. Hydrazine, as a byproduct of nitrogen reduction, is also monitored during the electrocatalytic NRR process. As shown in Figure S15, there is undetectable hydrazine in the electrolyte after 1 h of constant potential reduction. Control experiments under the Ar gas and open circuit potential (OCP) conditions are conducted. No NH3 generation was observed from the control experiments (Figure S16). Different amounts of VCl3 were dissolved in 0.1 M HCl and the absorption spectra were obtained by the indophenol blue method to investigate the effect of the V ions on the colorimetric method. As shown in Figure S17, 10 ppm of the V ion sample, which is 2,000 times higher than 5 ppb (V leaching after 10 cycles under âˆ’0.4 V, detected by ICP), shows a tiny hump. But 0.1 and 0.5 ppm, which are 20 and 50 times higher than 5 ppb, respectively, did not show any absorption changes. Therefore, only under the high concentration of metal ions can it be possible to affect the absorbance of the colorimetric method.
Figure 4

Figure 4. (a) LSV spectra of O-V2-NC under N2 and Ar. (b) NRR performance of O-V2-NC under different overpotentials (0 âˆ¼ âˆ’0.5 V vs RHE). (c) Stability test of O-V2-NC under 0 V (10 cycles = 10 h). (d) Stability test of O-V2-NC under âˆ’0.4 V (10 cycles = 10 h). (e) 1H NMR analysis of the electrolyte fed with 14N2 and 15N2 after 1 h of electrolysis at âˆ’0.4 V. (f) NH3 yield rate and FE of O-V2-NC under âˆ’0.4 V detected by a different method.

To confirm if the NH3 in the electrolyte is indeed from the nitrogen reduction, a 15N isotope labeling experiment was carried out. (24,49,50) 15N2 is continuously fed into the electrolyte for 1 h during the electroreduction process at a fixed voltage of âˆ’0.4 V. The product obtained from the 15N2 experiment is qualitatively analyzed by 1H NMR. As shown in Figure 4e, the 14NH4+ containing electrolyte shows they typical three peaks with 14JNâ€“H of 52.1 Hz, and the 15NH4+ containing electrolyte shows the typical two peaks with 15JNâ€“H of 72.8 Hz, while in the case of 15N2 as the feed gas, the NMR peak becomes broader, and a small peak at 6.94 ppm appears. That is attributed to the trace amount of 14N2 inside the electrolyte, which cannot be eliminated completely. Three different methods are conducted for quantitative analysis of NH3 yield and FE. To obtain the corresponding calibration curves for quantitative analysis, various concentrations of 14NH4+ and 15NH4+ are prepared, and the NMR tests are carried out (Figures S18 and S19). The NMR pattern is calibrated by a TMS peak shift of 0 ppm. Maleic acid is also used as the internal standard for accurate analysis, whose peak appears at 6.3 ppm (Figure S20). This O-V2-NC also shows excellent stability over 10 cycles under 0 V with negligible reduction of yield and FE (Figure 4c). Figure S21 exhibits the NH3 yield rate and FE of O-V2-NC under 0 V detected by a different method (product of 14N as feed gas is checked by indophenol blue and product of 15N as feed gas is checked by NMR), which shows the comparative results. To further investigate the dimer structure stability of O-V2-NC, the long-time performance test is conducted under âˆ’0.4 V for 10 cycles and the characterizations of the material structure are also carried out. Figure S22 exhibits the charge density and UVâ€“vis absorption of electrolyte of O-V2-NC under âˆ’0.4 V for 10 cycles. Figure 4d shows the NH3 yield and FE of O-V2-NC under âˆ’0.4 V for 10 cycles. This catalyst can keep good stability even under the high applied potential. The NH3 yield rate and FE under âˆ’0.4 V are 26 Î¼g hâ€“1 mgâ€“1 and 4.6%. Figure 4f presents the NH3 yield rate and FE of O-V2-NC under âˆ’0.4 V detected by a different method, further confirming the reliability of the employed indophenol blue method. The Aberration-corrected HAADF-STEM and XPS are carried out to investigate the structural change of the O-V2-NC after the stability test. The catalyst still shows the dimer structure along with the single atoms and the distance between the V-V dimer is 2.77 Ã… (Figure S23). The Vâ€“O bond is maintained after the stability test (Figure S24). These results also demonstrate the excellent stability of the dimer structure in O-V2-NC. Compared with other catalysts used in acidic conditions, O-V2-NC presents a higher FE at low applied potential and efficient NH3 yield at high potential (Table S3).
V samples with different amounts are synthesized to study the amount-dependent performance. VNC-5 and VNC-20 are prepared by adjusting the V precursor amount (0.05 mmol for VNC-5 and 0.2 mmol for VNC-20). The synthesis process is the same as that of O-V2-NC. The XRD result indicates the successful preparation of the V single atom (Figure S25). The STEM image also shows the uniform distribution of the V atoms, which are separated far away from each other due to less amount of V (Figure S26). The V amounts for VNC-5 and VNC-20 detected by ICP-MS are 0.11 and 0.591 wt %. The ENRR performances for VNC-5 and VNC-20 are conducted under the same conditions as that of O-V2-NC. Figures S27 and S28 exhibit the charge density and UVâ€“vis absorption of the electrolyte from VNC-5 and VNC-20 under various applied potentials. With the increase of V amount, the ENRR performance is also enhanced. Similar phenomena are also observed from the results of VNC-5 and VNC-20. That is, the FE decreased with the higher applied potential, which is attributed to the facilitated HER in the acidic conditions (Figures S29 and S30). To further investigate the advantages of dimer catalysts on the performance, V single-atom-based material (V coordinated by phthalocyanine, VPC) with a similar V amount as O-V2-NC is applied to the ENRR. From the ICP-MS, VPC contains 3.4 wt % of V, so 3.8 mg of VPC mixed with 6.2 mg of NC makes the sample contain around 1.3 wt % of V (VPC-NC). The ENRR experiment is carried out under the same conditions as those of O-V2-NC. Figure S31 shows the charge density and UVâ€“vis absorption of the electrolyte from VPC-NC under various applied potentials. The highest NH3 yield (12.5 Î¼g hâ€“1 mgâ€“1) of VPC-NC increased compared with those of VNC-5 and VNC-20, but it is still much lower than that of O-V2-NC (Figure S32). This result suggests that the designed dimer structure (V-O-V) is more favorable for the ENRR process and further improves the performance. However, to distinguish the effect on the performance between a single atom and dimer more precisely, the coordination environment should also be considered, which requires more input effort to design the catalyst in the future.
Theoretical Calculations
DFT calculation was further performed to provide a theoretical view of the possible NRR mechanism. First, the formation energies of V2N6O and V2N6 are compared to investigate the stability of the proposed configuration, as shown in Table S4. The formation energies calculated for V2N6O and V2N6 are around âˆ’3.37 and 1.67 eV, respectively, which indicates that the V2N6O structure is more thermodynamically stable than V2N6. Figure S33 shows the Gibbs free energy (Î”G) of H-adsorption over O (O-*H) and V (V-*H), as well as N2 adsorption over V (V-*N2). The Î”G of O-*H is 0.442 eV, which is much more positive than the Î”G of V-*H of 0.200 eV, implying that H is preferentially adsorbed on the V sites. Meanwhile, the Î”G of adsorbing N2 over V (V-*N2) is also calculated. The similar free energy between V-*H (0.200 eV) and V-*N2 (0.202 eV) indicates that H and N2 would compete for adsorbing over V sites. This could be the reason for the decrease in nitrogen reduction performance at the high applied potential. The NRR selectivity is also estimated by calculating the free energy difference between the H* and N2H* (Î”G(H*) â€“ Î”G(N2H*)). A positive Î”G(H*) â€“ Î”G(N2H*) value (>0) suggests a great preference for hydrogenation of N2* and retains good selectivity. (34) In our case, the value of Î”G(H*) â€“ Î”G(N2H*) is âˆ’0.5 at the limiting potential UL (âˆ’0.64 V). It indicates a comparatively low preference for the hydrogenation of N2*, which is consistent with our experiment result of the lower FE at a high applied potential. This is attributed to the trade-off between the adsorption of H* and N2H*, as catalysts with strong binding for N2H* also involve the strong potential for H* adsorption, leading to poor selectivity under the reaction conditions. (51) Figure S34 demonstrates the depicted structure of adsorbing H and N2 on O-V2-NC. The potential limiting step for both the distal and alternating pathways is the last step (*NH2 â†’ * + NH3) with an energy barrier of 0.64 eV. Nevertheless, there is one more barrier that needs to be overcome in the distal process (*NNH2 â†’ *N), which makes the alternating pathway more favorable (Figure 5). Interestingly, the structural change of O-V2-NC is noticed by DFT calculation during the protonation process (Figure S35). At the step of the first protonation, the oxygen connected with two vanadium atoms (V-O-V) would break down to V-O and V due to the hindrance of *NNH, leaving the vanadium connected *NNH alone (V-*NNH) instead of V-O-V*NNH. This transformed structure lasts until all the nitrogen species are released from the vanadium and regenerate to the original V-O-V form for the next circle.
Figure 5

Figure 5. Free energy diagram and structure model of intermediates of the NRR process on O-V2-NC with a distal pathway (i) and an alternating pathway (ii).

Conclusion
In this work, we developed a NRR dimer catalyst (O-V2-NC) with a peculiar local configuration in which the V dimer sites are bridged by one O atom (V-O-V). At 0 V vs RHE, the O-V2-NC exhibits a FE as high as 77% for NH3 production with a yield of 9.97 Î¼g hâ€“1 mgâ€“1 and a relatively efficient yield rate of 26 Î¼g hâ€“1 mgâ€“1 with the FE of 4.6% achieved at âˆ’0.4 V vs RHE. The advantages of the O-V2-NC coordination environment favor the NRR while suppressing hydrogen coupling, which is experimentally and theoretically certified. Interestingly, a loop evolution of the structure is found during the N2 reduction process of the DFT calculation, well explaining the excellent phenomena of activity and selectivity. This work gives further insights into investigating the role of varying structures during the catalytic process. In addition, this research will aid in exploring the single-atom dimer catalysts for selectively converting N2 to NH3 under benign conditions as well as the CO2 reduction reaction and nitrate reduction reaction, which possess the multi-intermediate and competitive HER reaction in the process. Not limited to this catalyst, other similarly designed catalysts with the proper electro-deficiency and coordination environment can also be applied to investigate the effect on the catalytic performance.
Experimental Section
Reagents and Materials
Vanadium chloride (VCl3, 97%), aniline (C6H7N, >99.5%), cyanuric acid (C3H3N3O3, 98%), ammonium persulfate ((NH4)2S2O8, â‰¥98%), phthalocyanine (C32H18N8, 98%), p-(dimethylamino)benzaldehyde (C9H11NO, 99%), sodium hypochlorite (NaClO, 11%â€“14%), sodium nitroprusside (Na2[Fe(NO)(CN)5]Â·2H2O, 99%), 15N2 (98 atom % 15N), maleic acid (C4O4H4, â‰¥99%), and dimethyl sulfoxide-d6 ((CD3)2SO, 99.9 atom % D, contains 0.03% (v/v) TMS) were purchased from Sigma-Aldrich. Ethanol (C2H6O, â‰¥99.9%), isopropanol (C3H8O, â‰¥99.9%), hydrofluoric acid (HF), sodium hydroxide (NaOH), salicylic acid (C7H6O3), sodium citrate (C6H5Na3O7Â·2H2O), and ammonia chloride (NH4Cl) were purchased from Sumchun Chemicals. All the reagents were used without further treatment.
Synthesis of Melamine Cyanurate (MCA)
The same amounts of cyanuric acid and melamine were dissolved in 100 mL of DI water separately, and then the melamine solution was poured into the cyanuric acid solution. The white precipitate appeared immediately. The mixed solution was continuously stirred for 2 h, filtered and washed with DI water 3 times, and dried under 50 in the vacuum oven.
Synthesis of Oxygen-Bridged Vanadium Single-Atom Dimer over N-Doped Carbon (O-V2-NC) and N-Doped Carbon (NC)
O-V2-NC was synthesized by simply modifying the previous procedure. (52) MCA was suspended in 100 mL of DI water under sonication for 10 min. 0.4 g of aniline and 0.4 mmol of VCl3 were added to the above suspension and stirred for proper dissolution. 0.52 g of APS (ammonium persulfate) was dissolved into 2 mL of DI water and dropped into the vanadium precursor solution. The reaction was conducted at room temperature for 12 h, and then the temperature was changed to 100 Â°C to evaporate all the solvent. The black powder was collected and treated under 800 Â°C for 4 h in a vacuum furnace. To obtain the vanadium single-atom product, the high-temperature furnace-treated sample was soaked in 1 M HF for 12 h, washed with a large amount of DI water and ethanol, collected by filtration, and dried in a vacuum oven overnight under 60 Â°C. The NC sample was synthesized according to the above procedure without adding any metal precursor. The NC sample was also soaked in 1 M HF for 12 h to ensure a similar microenvironment.
Synthesis of Vanadium Single Atom over N-Doped Carbon by Adjusting the V Amount (VNC-5, VNC-20)
VNC-5 and VNC-20 were prepared by the same method as the O-V2-NC except for the amount of the V precursor (0.05 mmol for VNC-5 and 0.2 mmol for VNC-20).
Synthesis of Vanadium Single Atom over Phthalocyanine (VPC)
VPC was synthesized by modifying the previous method. The V precursor and PC were mixed inside 20 mL of DMF and reflux under 120 Â°C for 6 h under the N2 protection. Then the precipitate was filtered and washed with ethanol and DI water 3 times and dried under 50 in the vacuum oven.
Cathode Preparation
5 mg of catalyst and 40 Î¼L of Nafion solution (5%) were dispersed into a 960 Î¼L mixture solution of DI water and IPA (isopropanol) by sonication for 1 h. 100 Î¼L of suspension was dropped onto the carbon clothes with the size of 1 Ã— 0.5 cm2. The mass loading of the catalyst was 1 mg cmâ€“2.
Electrochemical NRR Measurement
The NRR performance was carried out in the H-cell separated by a Nafion 115 membrane. The Nafion membrane was pretreated by sonication in 3 wt % H2O2 for 1 h, DI water for 30 min, and 0.5 M H2SO4 for 1 h and further cleaning with DI water several times. The electrochemical measurement was conducted in the three electrodes system, of which prepared cathodes were used as the working electrode and Ag/AgCl and Pt mesh were used as the reference electrode and counter electrode, respectively. All the potentials applied against Ag/AgCl were converted to RHE by E (vs RHE) = E (vs Ag/AgCl) + 0.197 V + 0.059 Ã— pH. Prior to applying any potential, pure N2 was purged into 30 mL of electrolyte of 0.1 M HCl for 30 min. The linear sweep voltammetry was scanned at the rate of 5 mVÂ·sâ€“1. The N2 reduction reaction was conducted at 0 V, âˆ’0.1 V, âˆ’0.2 V, âˆ’0.3 V, âˆ’0.4 V, and âˆ’0.5 V vs RHE for 1 h with continuous N2 feed-up.
Determination of Ammonia by the Indophenol Blue Method
The indophenol blue method was applied to determine the concentration of produced ammonia. (24) Typically, solution A contained 1 M NaOH with 5 wt % of salicylic acid and 5 wt % of sodium citrate. Solution B contained 0.05 M sodium hypochlorite. Solution C contained 1 wt % of sodium nitroprusside. In detail, 2 mL of solution A was added to 2 mL of the electrolyte after electrolysis; then 1 mL of solution B and 0.2 mL of solution C were added. The mixed solution was incubated under a dark environment for 2 h before the UVâ€“vis absorption measurement. The concentration was measured according to the absorbance at the wavelength of 655 nm. The calibration curve was built by using a standard NH4Cl solution with a series of concentrations in 0.1 M HCl.
Determination of Ammonia by NMR
40 Î¼L of sample in 0.1 M HCl, 40 Î¼L of maleic acid (MA, 200 ppm) in 0.1 M HCl, and 320 Î¼L of DMSO-d6 (contains 0.03% v/v TMS) were added and mixed sufficiently inside the NMR tube. The amount of the produced NH4+ was quantified by 1H nuclear magnetic resonance measurements (1H NMR; Agilent 500 MHz). The calibration curves of 14NH4Cl and 15NH4Cl were carried out over different concentrations of NH4+ ions in 0.1 M HCl: 0, 1, 3, 5, and 10 ppm.
Determination of Hydrazine
The hydrazine produced was determined by the method of Watt and Chrisp. The color reagent was prepared by dissolving 0.559 g of p-(dimethylamino)benzaldehyde into the mixture of 3 mL of HCl (12 mol Lâ€“1) and 30 mL of ethanol. 2 mL of color reagent was added to the 2 mL of the electrolyte after electrocatalysis, of which the absorbance at 455 nm was measured after 10 min of incubation. The calibration curve was plotted by using a standard hydrazine hydrate solution with a series of concentrations in 0.1 M HCl.
Calculation of Ammonia Yield and Faradaic Efficiency
NH3 yield was calculated by the equation:
NH3yield(Î¼ghâˆ’1mgâˆ’1)=ğ‘NH3Ã—ğ‘‰ğ‘¡Ã—ğ‘š
The Faradaic efficiency (FE) was calculated by the equation:
ğ¹ğ¸(%)=3Ã—ğ¹Ã—ğ‘NH3Ã—ğ‘‰17Ã—ğ‘„Ã—100%
CNH3): the measured NH3 concentration (Î¼g mLâ€“1)
V: volume of the electrolyte (30 mL)
t: reduction time (1 h)
m: mass loading of catalyst (mg)
F: Faradaic constant (96485 C molâ€“1)
Q: quantity of applied electricity (C)
15N2 Isotope Labeling Experiment
The 15N2 isotope labeling experiment was conducted to further ensure the electroreduction of N2. 15N2 was used as the feeding gas during the electroreduction process for 1 h. The electrolyte was concentrated, and the produced 15NH4+ was identified by 1H NMR located at the chiral material core facility center of Sungkyunkwan University.
Density Functional Theory (DFT) Calculation
DFT calculations were conducted through the Vienna ab initio Simulation Package (VASP) with the projector augment wave method. A generalized gradient approximation of the Perdewâ€“Burkeâ€“Ernzerhof (PBE) functional was used as the exchange-correlation functional. The cutoff energy was set as 500 eV, and structural relaxation was performed until the convergence criteria of energy and force reached 1 Ã— 10â€“5 eV and 0.02 eV Ã…â€“1, respectively. A vacuum layer of 15 Ã… was constructed to eliminate interactions between periodic structures of surface models. The van der Waals (vdW) interaction was amended by the DFT-D3 method of Grimme.
The formation energy of N6V2O and N6V2 was calculated by the following formula:
Î”ğ¸form(N6V2O)=ğ¸(N6V2O)âˆ’ğ¸(G)+10ğ¸bulk(C)/ğ‘›1âˆ’2ğ¸bulk(V)/ğ‘›2âˆ’3ğ¸gas(N2)âˆ’ğ¸gas(O2)/2Î”ğ¸form(N6V2)=ğ¸(N6V2)âˆ’ğ¸(G)+10ğ¸bulk(C)/ğ‘›1âˆ’2ğ¸bulk(V)/ğ‘›2âˆ’3ğ¸gas(N2)
where Î”Eform, E(N6V2O), E(N6V2), E(G), Ebulk(C), Ebulk(V), Egas(N2), and Egas(O2) represent the formation energy of doped graphene and the total energy of N6V2O, N6V2, pure graphene, bulk C and V, and N2 and O2 in the gas phase, respectively. n1 and n2 represent the number of atoms in bulk C and V, respectively.
The adsorption energy (Î”Eads) of N2 adsorption on the surface is defined as
Î”ğ¸ads=ğ¸(N2âˆ—)âˆ’ğ¸(âˆ—)âˆ’ğ¸(N2)
where E(*N2) and E(*) are the total energy of surface systems with and without N2, respectively, and E(N2) is the energy of an isolated N2 molecule. According to this definition, negative adsorption energy suggests that the adsorption process is exothermic and the adsorption system is thermodynamically stable. Contrarily, a positive value corresponds to an endothermic process.
The Gibbs free energy was calculated as
Î”ğº=Î”ğ¸+Î”ğ¸ZPEâˆ’ğ‘‡Î”ğ‘†
where Î”E, Î”EZPE, and Î”S are the electronic energy, zero-point energy, and entropy difference between products and reactants. The zero-point energies of isolated and absorbed intermediate products were calculated from the frequency analysis. The vibrational frequencies and entropies of molecules in the gas phase were obtained from the National Institute of Standards and Technology (NIST) database.
"""

main_material = "oxygen-bridged vanadium dimer (V-O-V) catalyst on N-doped carbon (O-V2-NC)"
```

```python
text = """
Introduction
Rechargeable zincâ€“air batteries (ZABs), as a new generation of green energy conversion devices, have been extensively explored due to the abundance of the zinc anode and high theoretical energy density (1086 W h kgâˆ’1).1â€“8 However, their large-scale application is still restricted by the sluggish kinetics of the oxygen reduction reaction (ORR), as well as the high-cost and scarce benchmark Pt/C catalysts.9,10 Alternatively, single-atom catalysts (SACs), prominently featuring Mâ€“Nx sites (where M refers to Fe, Co, Ni, etc.) on carbon substrates, have been validated to be the most promising contributors to the ORR owing to their maximized atom-utilization efficiency.11â€“14 Among these, Feâ€“N4â€“C catalysts exhibit the most exceptional ORR potential.15â€“18 Nevertheless, the symmetric D4h electron distribution in the Feâ€“N4 site endows oxygen intermediates with strong binding energy, resulting in unsatisfactory ORR catalytic activity.19â€“23 Therefore, the accurate manipulation of geometry and electronic structure of Feâ€“Nâ€“C catalysts to optimize the binding energy of key intermediates is significant but challenging.
The adsorption strength of intermediates on Feâ€“N4 active sites is primarily determined by the d-band centers of Fe atoms, heavily depending on their coordination environments and 3d orbitalsâ€™ electronic configurations.19,24â€“28 To regulate the electronic distribution and d-band energy levels of the surrounding metal sites, tremendous efforts have so far been devoted to adjusting the electron-withdrawing/donating properties of carbon carriers in the second coordination shell (CS) or further outside. For instance, the introduction of heterogeneous heteroatoms (B, P, S, O, etc.) with different atomic radii and electronegativity can distinctly improve the charge distribution on the carbon matrix surface.16,29â€“32 In this view, Yin et al. noticed that long-distance P atoms could drive electron delocalization of edge Feâ€“N4 and enhance ORR activity in acidic media.33 However, such long-range electron modulation may be limited by the strength of absorbing/donor moieties and the spatial interaction distance, leading to a diminished effect of electron delocalization.34,35 Hence, efficient modulation of the first CS has been proposed and extensively studied.36â€“40 For instance, ligand atoms have been directly anchored to host atoms, aimed at modulating the microenvironment or the electronic spin state of metal active centers for tailored d-band centers and intrinsic activity of Feâ€“Nâ€“C catalysts.41â€“44 In this respect, Li et al. induced the electron rearrangement and spin state transition of Feâ€“N3S catalysts by incorporating S atom-coordination to accelerate the nitrogen reduction reaction.42 Nevertheless, most studies have focused on coordination modulation or heteroatom substitution in the Feâ€“N4 plane while there is only a handful of reports on the modulation of the coordination number and geometrical structure design of metal centers. Furthermore, the promotion mechanism of electrocatalysis by the synergistic coordination effect of the first and second shells is still not clear.

Herein, a nuanced strategy was developed to meticulously tailor the coordination milieu and spin state of Fe sites. An innovative Feâ€“N4SP configuration with a high coordination structure was specifically designed through a multi-shell coupling strategy by decorating with axial-bridged sulfur atoms in the first CS surrounded by phosphorus atoms in the second CS. Density functional theory (DFT) calculations predicted broken electron symmetry of Feâ€“N4 sites using a high-coordination Feâ€“N4SP model along with a depressed d-band center and elevated spin density, resulting in declined *OH adsorption energy. Guided by the theoretical model, a single atom catalyst with Feâ€“N4SP sites on the N, P, and S ternary co-doped hollow carbon nanocage (Feâ€“N4SP/NPS-HC) was experimentally and precisely synthesized. Detailed structural characterization confirmed the critical role of highly coordinated configurations in tailoring the d-band centers, spin polarizations, and electronic structures. Consequently, the obtained Feâ€“N4SP/NPS-HC catalyst exhibited superior ORR activity, improved reaction kinetics, and enhanced stability in both alkaline and acidic media. When used in ZABs, the Feâ€“N4SP/NPS-HC exhibited high power density and amazing extra-long durability, superior to most reported catalysts even noble metal catalysts, demonstrating the potential practical application value.

Results and discussion
Theoretical prediction of Feâ€“N4SP model
In this study, the restricted ORR activity of conventional Feâ€“Nâ€“C catalysts was explored by setting up a standard four-coordination model (Feâ€“N4) as a foundation for DFT computations (Fig. S1 of the ESIâ€ ). Initially, the Gibbs free energy profiles were calculated for the ORR's basic steps on the standard Feâ€“N4 model at variable pH levels. The step bearing the most significant thermodynamic hurdle was deemed the rate-determining step (RDS) of the overall ORR pathway. In Fig. S1a and b (ESIâ€ ), the *OH desorption step on the standard Feâ€“N4 model's surface emerged as the RDS in the ORR, pointing to the suboptimal activity of the Feâ€“N4 sites stemming from the overly robust binding between the *OH intermediate and Fe sites. Nevertheless, the Sabatier principle suggested that the catalyst's adsorption for reaction intermediates should neither be too strong nor too weak. Over-adsorption can lead to a detrimental buildup of reaction intermediates on the catalyst's surface, thereby obstructing the reaction. The binding energy of the intermediate on the Fe site would be contingent upon its d-band center position. Consequently, fine-tuning the d-band center to weaken the adsorption of *OH could significantly improve ORR catalysis in Feâ€“Nâ€“C materials. Meanwhile, the d-band center of the same element would be closely related to the width of its d-orbitals. As illustrated in Fig. 1(a), the broadening of the d-band resulted in an elongated and narrowed shape (state Î± to state Î²) since the change in the bandwidth did not affect the total number of energy levels. Under constant electron filling number, the entire band shifted downward after narrowing, accompanied by a drop in the d-band center (state Î² to state Î³). Harnessing this principle, we aim to augment the coordination atoms (without replacements) of the Fe sites in Feâ€“Nâ€“C single-atom catalysts. In this case, higher coordination numbers would induce more intense d-orbital interactions of the single Fe atom, thereby widening the d-band and lowering the d-band center, conducive to weakening the adsorption of *OH.
image file: d3ee03183g-f1.tif
Fig. 1 Theoretical prediction. (a) Effect of energy bandwidth on d-band center position (Îµd: d-band center). (b) D-band centers, spin magnetic moments, and Bader charges of Feâ€“N4 and Feâ€“N4SP. Partial density of states (PDOS) for Fe 3d of (c) Feâ€“N4 and (d) Feâ€“N4SP. (e) The optimized structure of ORR intermediates on the Feâ€“N4SP model and free energy diagrams of Feâ€“N4SP at different electrode potentials in an acidic electrolyte.	
Accordingly, the Feâ€“N4SP configuration was designed to increase the coordination number of Fe sites by introducing axially bridged S atoms in the first CS and P atoms in the second CS. As expected, the d-band center of Feâ€“N4SP shifted to âˆ’1.9124 eV, a value 0.9 eV lower than the d-band center of the standard Feâ€“N4 configuration (âˆ’0.9802 eV, Fig. 1(b)â€“(d)). Furthermore, the computed ORR free energy diagrams of the Feâ€“N4SP model are displayed in Fig. 1(e). At U = 0[thin space (1/6-em)]V (vs. RHE), every reaction step in the Feâ€“N4SP configuration moved downhill, suggesting spontaneous and thermodynamically exothermic ORR processes. Notably, the free energy displayed an identical trend under alkaline correction, suggesting no altered RDS of the ORR process by pH adjustment (Fig. S2, ESIâ€ ). The limiting potential of Feâ€“N4SP was identified as 0.92[thin space (1/6-em)]V (Fig. 1(e)), highlighting the enhanced ORR activity that can be credited to the adsorption energy refinement of *OH intermediates by the highly coordinated structures in both acidic and alkaline environments. Moreover, we also developed the Feâ€“N4S and the Feâ€“N4P models to illustrate the functionality of sole S or P on the Feâ€“N4 configuration. As shown in Fig. S3 and S4 (ESIâ€ ), the individual doping of S or P also optimized the adsorption of *OH to a certain extent. However, the overpotentials for both models did not exceed 0.76 V, which is significantly less efficient for the ORR compared to the Feâ€“N4SP model (0.92 V). This finding underscored the synergistic effect between the axially bridged S atoms and the second CS-doped P atoms, where sulfur increased the coordination number at the Fe center, and phosphorus disrupted the square planar symmetry of the Feâ€“N4 configuration. In particular, sulfur positioned axially and phosphorus situated at a near-neighbor location together contributed to a charge redistribution of the Feâ€“N4 sites (Fig. S5, ESIâ€ ). Consequently, this distinctive high-coordination Feâ€“N4SP configuration not only promoted the downward shift of the d-band center but also diminished the electronegativity and increased the spin density of the Fe center (Fig. 1(b)), conducive to facilitating the desorption of *OH intermediates.

Preparation and morphological characterization of Feâ€“N4SP/NPS-HC
Based on the theoretical predictions, we precisely synthesized N, P, and S ternary co-doped hollow carbon nanocages with the Feâ€“N4SP configuration (Feâ€“N4SP/NPS-HC), followed by performance testing and mechanistic validation. The preparation procedure of Feâ€“N4SP/NPS-HC is schematically illustrated in Fig. 2(a). Initially, ZIF-8/FePc@PZS was synthesized by adding iron phthalocyanine (FePc) into the typical zeolite imidazole framework (ZIF-8) followed by coating with a uniform layer of phosphonitrile polymer (PZS) on its surface. This method allowed the encapsulation of FePc in the cavity of ZIF-8 without a change in the crystal structure (Fig. S6 and S7, ESIâ€ ) and the thickness of the PZS shell layer is about 20 nm (Fig. 2(b)). Afterward, Feâ€“N4SP/NPS-HC was fabricated by one-step pyrolysis and acid etching of ZIF-8/FePc@PZS, in which doped N served as the effective anchoring sites for the Fe atom, while abundant P and S elements in PZS were synchronously doped into the carbon shell. Additionally, the S element was also directly and axially bonded to the Fe atom to modulate the electronic structure of Feâ€“N4. The transmission electron microscopy (TEM) and high-resolution TEM (HRTEM) images of Feâ€“N4SP/NPS-HC in Fig. 2(c), (d) and Fig. S8 (ESIâ€ ) displayed a hollow carbon nanocage without any lattice striations, revealing Fe moieties present as single atoms or tiny clusters. Such a unique hollow structure would maximize the accessibility of the active sites to the electrolyte, as well as facilitate ion transport and mass transfer during electrocatalysis.45â€“47 Furthermore, only a broad carbon peak (at 24.4Â°) was observed in the X-ray diffraction (XRD) pattern, proving the complete etching of FeP caused by pyrolysis (Fig. S8b, ESIâ€ ). The aberration-corrected high-angle annular dark-field scanning TEM (HAADF-STEM) also confirmed uniformly dispersed Fe species as individual atoms throughout the hollow carbon shell (Fig. 2(e)). Additionally, energy-dispersive X-ray spectroscopy (EDX) mapping illustrated the presence of well-distributed C, N, P, S, and Fe elements throughout the entire carbon matrix (Fig. 2(f)). Among them, the elemental contents of Fe, P, and S were determined to be 1.33%, 3.90%, and 2.15%, respectively (Table S1, ESIâ€ ).
image file: d3ee03183g-f2.tif
Fig. 2 Synthesis and morphological characterizations. (a) Schematic illustration of the synthesis process of Feâ€“N4SP/NPS-HC. TEM images of (b) ZIF-8/FePc@PZS and (c) Feâ€“N4SP/NPS-HC. (d) HRTEM, (e) HAADF-STEM, and (f) corresponding EDX mapping images of Feâ€“N4SP/NPS-HC.	
Formation mechanism of the hollow structure
To gain insights into the formation of hollow structures and the influence of heteroatom doping, a series of controlled experiments were conducted and the corresponding morphologies are summarized in Fig. S9 (ESIâ€ ). Note that details of the experiments can be found in the ESI.â€  Obviously, Feâ€“N4/NC and Feâ€“N4P/NPC without S-doping depicted a similar porous solid polyhedron, with Feâ€“N4P/NPC accompanied by slight surface collapse, while Feâ€“N4S/NS-HC displayed a hollow nanocage (Fig. S10, ESIâ€ ). Thus, the S element played a key role in the formation of hollow structures. Further validation was implemented by thermogravimetric (TG) and XRD analyses. In Fig. S11a (ESIâ€ ), ZIF-8/FePc@PZS depicted a preferential decomposition at about 250 Â°C, while ZIF-8/FePc started to break down at 500 Â°C. Besides, the XRD pattern exhibited ZIF-8/FePc with the retained intact lattice structure of ZIF-8, while ZIF-8/FePc@PZS showed significant decomposition at 400 Â°C (Fig. S11b and c, ESIâ€ ). As the temperature rose to 600 Â°C, ZIF-8/FePc@PZS completely transformed into an amorphous carbon structure, while ZIF-8/FePc still retained the characteristic peak of ZnO. Therefore, the PZS shell contributed to the decomposition of ZIF-8/FePc. Note that the TG curve of ZIF-8/FePc@PZM highly coincided with ZIF-8/FePc, while ZIF-8/FePc@MZS also decomposed at 250 Â°C but at a lower decomposition rate than that of ZIF-8/FePc@PZS (Fig. S11a, ESIâ€ ). Thus, S played a vital role in the formation of hollow structures, while P accelerated the reaction to some extent.
Based on this information, a formation mechanism of the hollow structure was proposed and the details are shown in Fig. S11d (ESIâ€ ). During the pyrolysis process, ZIF-8/FePc underwent decomposition induced by S2âˆ’ ions from the PZS shell, releasing Zn2+ ions. The difference in radii between Zn2+ and S2âˆ’ resulted in a lower inward transport rate of S2âˆ’ than the outward diffusion of Zn2+, invoking the Kirkendall effect.48 The internal ZIF-8/FePc gradually decomposed and ultimately formed a hollow structure as a function of the occurrence of the Kirkendall effect. Simultaneously, the FePc molecules trapped in the ZIF-8 cavity were reduced in situ to form atomically dispersed Feâ€“N4 sites.

Coordination configurations and spin states of catalysts
The chemical valences and local coordination environments of Fe sites were clarified by X-ray absorption spectroscopy (XAS). The Fe K-edge absorption of Feâ€“N4SP/NPS-HC was situated between those of FePc and Fe2O3, elucidating an average valence of isolated Fe atoms of +2 to +3, as shown in the X-ray absorption near-edge structure (XANES, Fig. 3(a)). The coordination configurations of the Fe site were investigated by k3-weighted Fourier transform of Fe K-edge extended X-ray absorption fine structure (FT-EXAFS) spectroscopy in the R-space (Fig. 3(b)). Only the peak of Feâ€“N4SP/NPS-HC can be observed near 1.5 Ã…, consistent with FePc and corresponding to the predominant Feâ€“N scattering. The absence of the peak position at 2.2 Ã… demonstrated mainly anchored Fe atoms by N atoms instead of the Feâ€“Fe bond. It is noteworthy that a Feâ€“S peak can be detected in Feâ€“N4SP/NPS-HC via the quantitative least-squares curve-fitting, with coordination numbers of Feâ€“N and Feâ€“S calculated as 4.0 and 1.1 (Fig. 3(c), Fig. S12â€“S14 and Table S2, ESIâ€ ), respectively. In order to pinpoint the exact coordination structures of Feâ€“N4SP/NPS-HC, we considered all possible models for the position of coordination S atoms according to EXAFS fitting results, including four in-plane sites and the axial position. And the total energies of these models were calculated by the DFT (Fig. S15, ESIâ€ ), which indicates that the energy is the lowest and the structure is the most stable when sulfur is positioned axially. This convincingly demonstrated each Fe atom coordinated to four in-plane nitrogen atoms and one axially oriented sulfur atom, unambiguously fitting with the Feâ€“N4SP configuration, where the shortened Feâ€“S bond implied the effect of P atoms in the second CS (Table S2, ESIâ€ ). The unique axial S-coordination and adjacent P-coordination moved the central Fe slightly out of the Feâ€“N4 plane, thereby breaking the symmetry interactions of the original Feâ€“N4 sites.49,50 Compared to Feâ€“N4/NC, a negative shift of 0.3 eV accompanied by a decrease in the L3/L2 ratio of Feâ€“N4SP/NPS-HC were seen in Fe L2,3-edge XANES (Fig. 3(d)). Both indicated the declined valence state of the Fe center, consistent with DFT calculations (Fig. 1(b)). This revealed the strong electron transfer between Fe atoms and axial S atoms, as well as surrounding doped P atoms through dâ€“p orbital hybridization. Additionally, the wavelet transform (WT) analysis shown in Fig. 3(e) illustrated only a maximum intensity at about 5 Ã…âˆ’1, corresponding to the Feâ€“N bond. The deviation in the contour toward âˆ¼9 Ã…âˆ’1 can be attributed to the Feâ€“S contribution. Overall, the Feâ€“N4SP configuration was identified as the most probable structure of the Feâ€“N4SP/NPS-HC catalyst. Specifically, the Fe species were stabilized by the four N atoms as isolated atoms, while being modified by the S atom in the first CS combined with the P atom in the second CS.
image file: d3ee03183g-f3.tif
Fig. 3 Atomic and electronic structure analysis. (a) Fe K-edge XANES, (b) FT-EXAFS spectra, and (e) WT contour plots of Feâ€“N4SP/NPS-HC and reference materials. (c) FT-EXAFS fitting curve of Feâ€“N4SP/NPS-HC. (d) Fe L-edge XANES and (f) EPR spectra of Feâ€“N4SP/NPS-HC and Feâ€“N4/NC. (g) The magnetic hysteresis loops of Feâ€“N4SP/NPS-HC and Feâ€“N4/NC at room temperature (300 K). (h) Temperature-dependent magnetic susceptibilities of Feâ€“N4SP/NPS-HC and Feâ€“N4/NC.	
Electron paramagnetic resonance (EPR) measurements were carried out to highlight the importance of 3d orbital electron arrangement in ORR activity. Note that in this technique, the g factor would directly be related to the paramagnetic absorption signal of the unpaired spin electron.51 As shown in Fig. 3(f), Feâ€“N4SP/NPS-HC exhibited the highest g value compared to Feâ€“N4P/NPC and Feâ€“N4/NC, indicating increased unpaired 3d electron numbers of the Fe center through doping of heteroatoms, especially axial S atoms. The higher saturation magnetization also meant more unpaired electrons in Feâ€“N4SP/NPS-HC (Fig. 3(g)). The zero-field cooling (ZFC) temperature-dependent magnetic susceptibility tests were then performed to further explore the d-electrons spin states of Fe sites. The effective magnetic moments of Feâ€“N4SP/NPS-HC and Feâ€“N4/NC calculated by fitting the curve of Ï‡âˆ’1â€“T according to the Curieâ€“Weiss law were 3.02 Î¼eff and 1.73 Î¼eff (Fig. 3(h)). Obviously, the highly coordinated configuration triggered by axial S atoms and proximal P atoms intensified the spin polarization in the 3d orbitals of Fe atoms, validating the DFT calculations (Fig. 1(b)). Thus, high-energy eg orbitals with more electron filling can be obtained in Feâ€“N4SP/NPS-HC for facilitated penetration of oxygen antibonding Ï€-orbitals, thereby promoting O2 activation and reduction.52â€“55

Modulation of the Fe content and porosity
The influence of Fe sites was studied by preparing Feâ€“N4SP/NPS-HC with different Fe contents. As depicted in Fig. S16 (ESIâ€ ), all as-prepared Feâ€“N4SP/NPS-HC with different Fe contents exhibited similar hollow nanocage structures, further proving the unaffected structure of the catalysts by adding FePc. As the amount of added FePc increased, the metal phase gradually transformed from the FeP phase to the Fe2P phase, and all metal phosphides were removed after acid leaching (Fig. S17, ESIâ€ ). The Fe contents of Feâ€“N4SP/NPS-HC-0.5, Feâ€“N4SP/NPS-HC, Feâ€“N4SP/NPS-HC-1.5, and Feâ€“N4SP/NPS-HC-2 determined by inductively coupled plasma optical emission spectrometry (ICP-OES) were 0.63, 1.33, 0.39, and 0.21 wt.% (Table S3, ESIâ€ ), respectively. Notably, the addition of excess FePc led to a decline in Fe content in the catalyst due to the weakened spatial confinement effect of ZIF-8, insufficient in limiting the migration of excessive FePc during pyrolysis and consequently triggering agglomeration. Most Fe moieties were removed as phosphate iron during subsequent acid etching, resulting in a lower Fe content in the catalysts. Interestingly, FePc was involved in modulating not only the densities of metal sites but also the specific surface area and porosity of the catalysts. The Brunauerâ€“Emmettâ€“Teller (BET) measurements shown in Fig. S18 and Table S4 (ESIâ€ ) revealed an increment in the specific surface area as a function of the amount of added FePc. Moreover, pore diameters of Feâ€“N4SP/NPS-HC-0.5 and Feâ€“N4SP/NPS-HC were predominantly distributed between 2 and 4 nm, while larger mesopores (7â€“10 nm) started to form after the addition of 150 mg FePc, further confirming the aggregation of excess FePc. A large number of aggregated metal compounds were etched, leaving abundant hierarchical micro/mesopores in the carbon matrix. The large specific surface area and abundant pores could enhance the contact between the active site and electrolyte, as well as accelerate electron and mass transport.56,57
Electrocatalytic ORR performance of catalysts
The unique coordination configuration and electronic environment of the as-prepared catalysts prompted the investigation of their ORR properties in alkaline and acidic electrolytes in a three-electrode system. The linear sweep voltammetry (LSV) curves of as-prepared catalysts evaluated in 0.1 M KOH are displayed in Fig. 4(a). The Feâ€“N4SP/NPS-HC presented a more positive half-wave potential (E1/2) of 0.912 V than Feâ€“N4P/NPC (0.858 V), Feâ€“N4/NC (0.828 V), Pt/C (0.889 V), and most published state-of-the-art Fe-based electrocatalysts (Table S5, ESIâ€ ), indicating the excellent ORR activity of Feâ€“N4SP/NPS-HC in alkaline media. Since a smaller Tafel slope would mean faster reaction kinetics, Feâ€“N4SP/NPS-HC with a much smaller Tafel slope of 39.18 mV decâˆ’1 than others would indicate the highly coordinated configuration contributing to an accelerated ORR process (Fig. 4(b)). Undeniably, the kinetic current density (Jk) is also a key parameter to evaluate the ORR activity. Feâ€“N4SP/NPS-HC showed a record Jk value of 60 mA cmâˆ’2 at 0.85 V, exceeding those of Feâ€“N4P/NPC, Feâ€“N4/NC, and Pt/C by 8.5-, 20-, and 4-fold, respectively (Fig. 4(c)). Furthermore, the effect of the hollow structure was explored by preparing Feâ€“N4/NPSC with a solid structure. As expected, Feâ€“N4SP/NPS-HC exhibited better ORR performances, indicating the superiority of the hollow structure (Fig. S19, ESIâ€ ).
image file: d3ee03183g-f4.tif
Fig. 4 Electrocatalytic performance of the ORR. (a) ORR polarization curves at a rotation rate of 1600 rpm, (b) Tafel plots, (c) comparison of E1/2 and Jk at 0.85 V, (d) electron-transfer number and H2O2 yield, and (e) double-layer capacitance (Cdl) values of Feâ€“N4SP/NPS-HC, Feâ€“N4P/NPC, Feâ€“N4/NC, and Pt/C in 0.1 M KOH. (f) LSV curves of Feâ€“N4SP/NPS-HC before and after 10k cycles in 0.1 M KOH. (g) ORR polarization curves at a rotation rate of 1600 rpm, (h) electron-transfer number and H2O2 yield, and (i) comparison of E1/2 and Jk at 0.7 V of Feâ€“N4SP/NPS-HC, Feâ€“N4P/NPC, Feâ€“N4/NC, and Pt/C in 0.1 M HClO4.	
The in-depth ORR details of Feâ€“N4SP/NPS-HC are demonstrated in Fig. S20 (ESIâ€ ). Compared to the N2-saturated electrolyte, the pronounced reduction peak of cyclic voltammetry (CV) at 0.895 V illustrated good intrinsic reactivity in the O2-saturated electrolyte. The first-order reaction kinetics and a 4-electron transfer pathway were extrapolated by the parallel linear relationships in Kouteckyâ€“Levich (Kâ€“L) plots. Furthermore, rotating ring disk electrode (RRDE) tests confirmed a transfer electron number of Feâ€“N4SP/NPS-HC close to 4 accompanied by a lower H2O2 yield (below 6%) than other catalysts, indicating its higher selectivity toward the 4-electron mediated ORR process (Fig. 4(d)). Afterward, the electrochemical double-layer capacitance (Cdl) was used to estimate the electrochemical surface areas (ECSA). As shown in Fig. 4(e) and Fig. S21 (ESIâ€ ), the Cdl value of Feâ€“N4SP/NPS-HC was identified as 19.8 mF cmâˆ’2, superior to those of Feâ€“N4P/NPC (14.02 mF cmâˆ’2), Feâ€“N4/NC (9.87 mF cmâˆ’2), and Pt/C (15.25 mF cmâˆ’2). This result implied that the hollow structure endowed Feâ€“N4SP/NPS-HC with maximum exposure of active sites, thereby facilitating the ORR reaction. Besides, the catalyst stability was assessed by the LSV after cycling and chronoamperometric response (iâ€“t) curves. As presented in Fig. 4(f), only 17 mV of decay in E1/2 was recorded after 10k cycles for Feâ€“N4SP/NPS-HC. Moreover, the current of Feâ€“N4SP/NPS-HC still remained above 95% after continuously operating for 70 h, while Pt/C decayed to 68% only after 37 h (Fig. S20d, ESIâ€ ). All of the above results demonstrated the excellent ORR activity, reaction kinetics, and stability of Feâ€“N4SP/NPS-HC in the alkaline media.

The broadened usefulness of Feâ€“N4SP/NPS-HC was further investigated by evaluating the ORR performance in acidic media. As shown in Fig. 4(g)â€“(i), Feâ€“N4SP/NPS-HC also exhibited a striking ORR activity in 0.1 M HClO4 with an E1/2 value of 0.814 V and a Jk value of 46.7 mA cmâˆ’2 at 0.7 V, values much higher than those of Feâ€“N4P/NPC (E1/2 = 0.782 V, Jk = 15.7 mA cmâˆ’2) and Feâ€“N4/NC (E1/2 = 0.769 V, Jk = 14.3 mA cmâˆ’2), even exceeding those of commercial Pt/C (E1/2 = 0.785 V, Jk = 16.5 mA cmâˆ’2). Compared to other catalysts, the smallest Tafel slope (49.72 mV decâˆ’1) and the largest Cdl (36.72 mF cmâˆ’2) of Feâ€“N4SP/NPS-HC also implied faster reaction kinetics and a larger electrochemically active area (Fig. S22a, b and S23, ESIâ€ ). Meanwhile, RRDE experiments confirmed a high-efficiency 4eâˆ’ ORR pathway with a transfer of 3.96 electrons per process accompanied by an H2O2 yield below 3% (Fig. 4(h)). Compared to Pt/C exhibiting 69% current retention attenuation after 52 h, Feâ€“N4SP/NPS-HC showed a high current retention of 97% after 76 h and only 12.2 mV attenuation after 10k cycles, indicating outstanding long-cycle durability in an acidic electrolyte (Fig. S22c and d, ESIâ€ ). In addition, unlike the Pt/C, the CV curves and the current density of Feâ€“N4SP/NPS-HC remained essentially unchanged after methanol crossover, inferring a favorable methanol tolerance in both acidic and alkaline environments (Fig. S24, ESIâ€ ). Significantly, the performances of the proposed Feâ€“N4SP/NPS-HC catalyst also compared well with previously reported single-atom catalysts (Table S6, ESIâ€ ). All these results corroborated the predominance of the designed theoretical model.

The effects of the metal site density on the catalytic ORR performance of Feâ€“N4SP/NPS-HC with different Fe contents were also studied. Both the half-wave potential, the Tafel slope, and kinetic current density were positively correlated with the Fe content and far exceeded those of HPS-HC without Fe species (Fig. S25, ESIâ€ ), confirming the critical role played by isolated Feâ€“N4 sites in the ORR activity. Therefore, modulating the electronic structure and spin state of the Feâ€“N4 sites is essential to enhance ORR activity. Besides, the impacts of the calcination atmosphere on ORR activity were also compared in Fig. S26a and b (ESIâ€ ). Unlike results under an Ar atmosphere, Feâ€“N4SP/NPS-HC obtained in Ar/NH3 exhibited a better E1/2 value and a smaller Tafel slope. Furthermore, X-ray photoelectron spectroscopy (XPS) spectra were recorded to reveal the differences in chemical composition between Feâ€“N4SP/NPS-HC obtained under an Ar/NH3 and an Ar atmosphere. The high-resolution N 1s spectra of Feâ€“N4SP/NPS-HC and Feâ€“N4SP/NPS-HC-Ar both displayed three peaks at 401.3, 400.4, and 398.5 eV, corresponding to graphitic-N, pyrrolic-N, and pyridinic-N, respectively (Fig. S26c, ESIâ€ ). Meaningfully, the results showed that a small amount of ammonia contributed to the enhancement of the N content in the catalyst, especially the percentage of pyridinic-N and pyrrolic-N (Fig. S26d, ESIâ€ ), which were also considered important ORR active sites in previous reports.58â€“60 Overall, the single Fe atoms coordinated with pyridinic-N and pyrrolic-N served as predominant catalytic active sites. Also, the enhanced ORR activity can be attributed to the near/long-range electronic regulation of the axial S-coordination and adjacent P-coordination, synergizing the unique hollow structure.

Performance in zincâ€“air battery applications
Inspired by the excellent ORR performance, a Znâ€“air battery (ZAB) was prepared with Feâ€“N4SP/NPS-HC as the catalyst and tested to verify its practical application value. The schematic illustration is shown in Fig. 5(a) and further details can be found in the ESI.â€  As expected, the ZAB based on Feâ€“N4SP/NPS-HC showed an open circuit voltage of 1.42 V and a specific capacity of 893 mA h gZnâˆ’1 at a current density of 5 mA cmâˆ’2, outperforming ZAB based on Pt/C (1.39 V and 752 mA h gZnâˆ’1, Fig. S27, ESIâ€ ). Moreover, Feâ€“N4SP/NPS-HC-based ZAB possessed a better current density than others at the same voltage (Fig. 5(b)), and the peak power densities (225 mW cmâˆ’2) were 1.2-, 1.5-, and 1.3-fold higher than those of Feâ€“N4P/NPC, Feâ€“N4/NC, and Pt/C, respectively. These data were also superior to recently reported ZABs (Fig. S28 and Table S7, ESIâ€ ). The galvanostatic discharge curves of Feâ€“N4SP/NPS-HC and Pt/C at different current densities revealed a decline in discharge potential plateau as the current density increased (Fig. 5(c)). However, the voltage of Feâ€“N4SP/NPS-HC was permanently higher than that of Pt/C. Feâ€“N4SP/NPS-HC still maintained a voltage plateau of 1.12 V even at a high current density of 100 mA cmâˆ’2. As the current density gradually decreased, the voltage plateau also returned to the initial state, indicating the remarkable rate performance of Feâ€“N4SP/NPS-HC. Significantly, unlike the obvious voltage drop of Pt/C-based ZAB at 130 h, Feâ€“N4SP/NPS-HC-based ZAB still maintained an ultra-stable voltage window after continuous operation for over 320 h (approximately 2880 cycles), revealing superb long-term durability (Fig. 5(d) and Fig. S29, ESIâ€ ). It is worth mentioning that the charging voltage of Feâ€“N4SP/NPS-HC-based ZAB is significantly lower than that of Pt/C-based ZAB, which can be attributed to the excellent oxygen evolution reaction (OER) activity of Feâ€“N4SP/NPS-HC (Fig. S30, ESIâ€ ). Besides, as shown in Fig. 5(e), the charge/discharge voltage gap was only 0.704 V in the initial 10 cycles and the round-trip efficiency reached as high as 62.9%. Note that the voltage gap only increased by 2% and 9% at the 1505th and 2875th cycles, respectively. Also, the cycling efficiency still remained above 60% even after the 2875th cycle (Fig. 5(f) and (g)). In sum, the extra-long stability would greatly be important for the practical use of Feâ€“N4SP/NPS-HC-based ZAB.
image file: d3ee03183g-f5.tif
Fig. 5 Zincâ€“air battery performance. (a) Schematic illustration of a Znâ€“air battery. (b) Discharge polarization and power density curves of Feâ€“N4SP/NPS-HC, Feâ€“N4P/NPC, Feâ€“N4/NC, and Pt/C. (c) Galvanostatic discharge curves of Feâ€“N4SP/NPS-HC and Pt/C at different current densities. (d) Charge/discharge cycling stabilities of Feâ€“N4SP/NPS-HC and Pt/C at a current density of 5 mA cmâˆ’2. (e)â€“(g) Charge/discharge cycle curves and round-trip efficiency of different cycles.	
Conclusions
In summary, a meticulous multi-shell synergistic coordination strategy was successfully used to construct Feâ€“N4SP configurations with an elevated coordination number. DFT calculations and detailed structural characterization demonstrated that the precisely tailored interaction of axial S-coordination and adjacent P-coordination effectively induced electronic delocalization and decreased the d-band center of the Feâ€“N4 site, accompanied by an increase in the spin state. This strategy contributed to the activation and reduction of oxygen, as well as the desorption of *OH intermediates in the RDS, endowing Feâ€“N4SP with an enhanced ORR intrinsic activity. The optimal Feâ€“N4SP/NPS-HC catalyst displayed an impressive high half-wave potential (E1/2, 0.912 V in alkaline media and 0.814 V in acidic media) coupled with excellent kinetic current density (Jk, 60 mA cmâˆ’2 at 0.85 V in alkaline media and 46 mA cmâˆ’2 at 0.7 V in acidic media), along with an outstanding maximum power density of 225 mW cmâˆ’2 and continuous stable operation up to 320 h when used as the ZAB catalyst. This work unveils the critical role of the coordination micro milieu and spin state on the ORR catalytic performance, promising to guide the design and optimization of high-efficiency single-atom catalysts from the electronic level.
"""

main_material = "Feâ€“N4SP/NPS-HC"
```

```python
# text = """
# The oxygen evolution reaction (OER), which is considered as the key limiting reaction in various electrochemical devices such as water splitting1â€“6 and rechargeable metalâ€“air batteries,7â€“10 has received significant research attention. The OER is a kinetically sluggish reaction, involving the transfer of four proton-coupled electrons towards the formation of an O[double bond, length as m-dash]O bond.11 Intensive efforts have been devoted to the understanding of the OER in an alkaline medium, wherein the interaction between the intermediate species and the local structure of the active sites have been investigated.12â€“14 According to the adsorption evolution mechanism, the OER activity is dominated by the adsorption energetics of the oxo-intermediates, whereby the optimization of adsorption is limited by the scaling relation.15 Specifically, the scaling relation of the OER is associated with the binding energies of the main intermediate species, namely OH* and OOH*, to the active sites.16 Thus, decoupling the scaling relation of the intermediate species through the surface structure design on the catalyst is considered to be the best strategy for the OER.17â€“19
# However, the development of OER catalysts remains a challenge due to the high overpotentials required to drive the multi-electron and multi-proton redox reaction, which also accelerate the destruction of the highly active surface structures. It has been reported that the surface coverage of the intermediate species strongly affects the reaction kinetics at a low overpotential.20 Only at elevated potentials will the reaction intermediates gradually desorb from the metal sites by forming an oxygen molecule. Recently, numerous non-precious metal catalysts have been reported for OER electrocatalysis in alkaline conditions, mainly including Co-based, Ni-based, and Fe-based materials.21,22 Among them, Coâ€“Fe dual-metal catalysts are found to possess high OER activity in alkaline conditions. Generally, the complicated OER catalytic process is associated with the participation of surface metallic active sites.23â€“28 Thus, identifying the synergistic effect between Fe and Co is critical for the comprehensive understanding of the reaction mechanism to design new bimetal-based catalysts with high efficiency. Hu et al. revealed the formation of the Coâ€“Fe active site structure during electrochemical activation through operando X-ray absorption spectroscopy, confirming the presence of Coâ€“Fe dual-atom sites.29 However, the synergetic mechanism of the dual-active sites in the electrocatalytic process is far from being clearly understood due to the complex catalytic interface and limited monitoring methods. Most studies on Co/Fe dual-atom catalysts are devoted to identifying the active sites and the underlying catalytic mechanism.30â€“33 However, the real catalytic mechanism, which is associated with the evolution of the Co/Fe dual-metal sites during the catalytic process, is ambiguous, and thus should be seriously considered.

# Herein, a Co/Fe dual-metal-site catalyst was prepared using the reverse-micelle assembly-pyrolysis strategy, which exhibited a superior performance towards the OER. The overpotential of the optimized catalyst at 10 mA cmâˆ’2 was determined to be as low as 240 mV, which is lower than that of the benchmark commercial IrO2 catalyst. Experimental evidence suggests that the synergistic effect between Co and Fe activates the generation of OOH*, whereby the Co and Fe sites act as the active site and catalytic promoter, respectively. The FesiteOOH sites originate from the oxidation of Fe2+ in alkaline media at a high potential, which can further trigger the formation of CositeOOH active sites. Density functional theory (DFT) calculations confirmed the synergistic effect between the Fe and Co single-atom-sites on the improved OER kinetics. The pinned Co-3d orbitals showed high electron transfer efficiency, while the Fe-3d orbitals protected the Co sites to sustain high electrocatalytic activity during the OER. The insight into the synergistic effect of the Fe and Co sites in Co/Fe-SNC gained from this study will shed light on the design of highly efficient OER electrocatalysts.

# Results and discussion
# The preparation of the Co/Fe-SNC catalyst is illustrated in Scheme 1. Firstly, the well-dispersed Fe(OH)x/Co(OH)x@SNC precursor was obtained via a microemulsion-co-precipitation reaction, with cetyltrimethylammonium bromide (CTAB), melamine (M), trithiocyanuric acid (TCA) and metal salt (CoCl2 and FeCl2Â·4H2O) in the presence of N,N-dimethylformamide (DMF) followed by the addition of polyetherimide (PEI), which was stirred overnight. Then, the as-prepared Fe(OH)x/Co(OH)x@SNC precursor powder was pyrolyzed at 800 Â°C to stabilize the metal-N/S coordination structure, whereby the Fe and Co atoms were coordinated to the N or S species hosted in the carbon substrates, resulting in atomically dispersed Co/Fe dual-sites. Finally, acid leaching was carried out to remove the metal or metal oxide nanoparticles, which could ensure the exposure of the atomically dispersed metal sites, avoid the blockage of the porous structure and improve the electrical conductivity of the materials.
# image file: d2ee03930c-s1.tif
# Scheme 1 Schematic illustration of the preparation of Co/Fe-SNC800.	
# The morphology, microstructure, and composition of the catalysts were systematically characterized by electron microscopy. The scanning electron microscopy (SEM) and transmission electron microscopy (TEM) images (Fig. 1a and b, respectively) reveal that the Co/Fe-SNC800 catalyst shows a typical open hollow sphere with a hierarchical porous structure consisting of micropores and mesopores, which was also confirmed by the Brunnerâ€“Emmettâ€“Teller (BET) measurements (the inset in Fig. 1b and Fig. S1, ESIâ€ ). The pore size curves indicate that all the catalysts possess a dominant micro-meso-pore distribution with a pore size ranging from 5 to 50 nm. Co/Fe-SNC800 shows higher microporous structures (Fig. S1b, ESIâ€ ), leading to its higher surface area (242.6 m2 gâˆ’1) than that of its single-metal counterparts, i.e., Fe-SNC800 (126.9 m2 gâˆ’1) and Co-SNC (67.3 m2 gâˆ’1). Moreover, abundant cross-linking channels can be observed, which are separated by the ultrathin nano-walls with a thickness in the range of 2.5âˆ¼4.5 nm, as shown in Fig. 1c. The fluffy porous structures result in a high degree of defects, which was verified by Raman spectroscopy (Fig. S2a, ESIâ€ ). The low ID/IG value (0.918) of Co/Fe-SNC800 may result from the reconstruction of defect structures induced by favorable anchoring of the Co/Fe bimetal sites at the edge/corner-rich structures. Furthermore, peaks associated with the vibrational modes of FeOx or CoOx cannot be observed at the low wavenumber region, suggesting the absence of metal oxides in the catalysts34 (Fig. S2b, ESIâ€ ), which is consistent with the X-ray diffraction (XRD) results (Fig. S3, ESIâ€ ). The XRD pattern of the catalysts displays a broad peak centered at 24.5Â° and a weak peak at 43.5Â°, which can be indexed to the (002) and (100) crystallographic planes of carbon, respectively, while no crystalline inorganic phases can be detected. The metal contents of the catalyst were determined by inductively coupled plasma optical emission spectrometry (ICP-OES). As shown in Table S2 (ESIâ€ ), Co/Fe-SNC800 contains 1.18 wt% Fe and 1.66 wt% Co metals, suggesting that the total metal content is higher than that of Fe-SNC800 (1.09 wt% Fe) and Co-SNC800 (1.24 wt% Co), respectively. The energy-dispersive spectroscopic (EDS) mapping indicates that the C, N, S, O, Co, and Fe atoms are uniformly distributed on the carbon matrix (Fig. 1d). Besides, the high-angle annular dark field scanning transmission electron microscopy (HAADF-STEM) image confirms the presence of the dual-metal sites, as shown in Fig. 1e and Fig. S4 (ESIâ€ ), which are displayed as two adjacent bright dots circled with pink marks. The representative dual-atom site in Region 1 in Fig. 1f is evidenced in the raster graphic. The intensity profile of the dual-atom sites in Region 1 reveals the presence of a diatomic feature, with a bond length of 3.479 Ã… (Fig. 1g). It is inevitable that the dual-atom-site can co-exist with other metallic species on the surface, including single-atom-sites, triple-atom-sites, and clusters. To assess the possibility of the occurrence of dual-atom metallic sites, statistical analysis was carried out by counting 35 randomly picked metallic species in Fig. 1e and Fig. S4 (ESIâ€ ), with the percentage of each species displayed in the histogram, as shown in Fig. 1h. It should be noted that when discriminating the single-atom-site from the associated metallic atoms in the dual-atom or triple-atom, the atoms more than 9 Ã… apart were treated as isolated sites, whereas most of the dual-site atoms show distances in the range of 3â€“6 Ã…, consistent with the standard used in the literature.35 It can be seen that the percentage of dual-atom metallic sites is approximately 60.00% as the dominant metallic sites in Co/Fe-SNC800. Furthermore, the statistics histogram of the measured distances between the adjacent metal atoms in the dual-atom sites shows a dominant distance of 3â€“6 Ã… (Fig. 1i).


# image file: d2ee03930c-f1.tif
# Fig. 1 Electron microscopy analysis of Co/Fe-SNC800. (a) SEM image, (b and c) HAADF-STEM images (inset in (b) shows the pore distance for Co/Fe-SNC800 obtained from BET) and (d) corresponding elemental mapping images of Co/Fe-SNC800, (e) colored HAADF-STEM images, (f) corresponding intensity profile along the arrow in Region 1 of e, (g) distance between two adjacent metal atoms in Region 1 of (e), (h) diagram for the percentages of single, double, triple atoms and clusters and (i) distance distribution of two adjacent metal atoms in dual-atom sites.	
# As shown in Fig. S5 (ESIâ€ ), the Fourier-transform infrared spectra (FTIR) display vibration bands located at low wavenumbers (<1000 cmâˆ’1), which are ascribed to the metalâ€“nitrogen bond.36 The bands at around 900â€“650 cmâˆ’1 can be assigned to the Coâ€“N or Feâ€“N vibrations. Both bands confirm the coordination between Co/Fe and N in the catalysts. As shown in the high-resolution X-ray photoelectron spectroscopy (XPS) core-level N 1s spectra (Fig. S6a, ESIâ€ ), the negative shifts in the pyridinic N and metalâ€“N of Co/Fe-SNC800 compared with that of Fe-SNC800 and Co-SNC800 can be attributed to the coordination between the dual-atom metal (Fe, Co) and the adjacent N atoms.37,38 Besides, the highest metalâ€“N component in Co/Fe-SNC800 suggests the highest content metalâ€“N active sites due to the coexistence of Fe and Co (Fig. S6b, ESIâ€ ). As shown in Fig. S7 (ESIâ€ ), â€“Câ€“Sâ€“Câ€“ is the potential dominant coordination for S in the catalysts, and the slight shift in the peaks associated with S 2p3/2 S2p1/2 in Co/Fe-SNC800 suggests the bonding between the metal atoms and S atoms.39,40 As shown in Fig. S8 (ESIâ€ ), the Fe 2p peak of Co/Fe-SNC800 shifted to a low binding energy, while the Co 2p peaks shifted to a high binding energy compared to that of Fe-SNC800 and Co-SNC800, respectively, which is ascribed to the electron-coupling effect between the adjacent Fe and Co atoms.

# To obtain insights into the atomic-level local electronic and coordinative environments in the catalysts, X-ray absorption near edge structure (XANES) and extended X-ray absorption fine structure (EXAFS) analyses were performed. The k-edge XANES spectra of Fe for Fe-SNC800 and Co/Fe-SNC800 are very similar to that of FePc, but obviously different from that of Fe2O3 and Fe foil (Fig. 2a). Similar trends were also observed in the Co k-edge XANES spectra (Fig. 2b), possibly because the square planar geometry of Fe and Co, like in FePc or CoPc, is distorted by the S doping. The k-edge absorption energy of Fe in Co/Fe-SNC800 is located between that of Fe2O3 and FePc, while the Co k-edge absorption energy is between that of Co2O3 and CoPc, suggesting that the oxidation state valances of both Fe and Co in Co/Fe-SNC800 are between +3 and +4. The plots of valence vs. energy are shown in Fig. S9 and 10 (ESIâ€ ), verifying that the valences of Fe and Co in Co/Fe-SNC800 are approximately +3.7 and +3.5, respectively. The Fourier-transformed (FT) k2-weighted EXAFS spectra of Fe for Co/Fe-SNC800 (Fig. 2c) shows the main peak at 1.63 Ã…, corresponding to the Feâ€“N/S bond, which is slightly larger than the bonding length of Feâ€“N in FePc (1.53 Ã…) due to the structural distortion. Fig. 2d suggests that Coâ€“N/S are the nearest neighbouring atoms (1.47 Ã…), which is still slightly different from the Coâ€“N bond length in CoPc, indicating the distorted Coâ€“N coordination. In Co/Fe-SNC800, no metal-metal bonds (e.g., Coâ€“Co, Feâ€“Fe, and Coâ€“Fe coordination) can be observed, indicating that the isolated Fe and Co atoms are atomically dispersed without obvious aggregation. The wavelet transform (WT) further unambiguously reveals the isolated metal site in all the obtained catalysts (Fig. S11, ESIâ€ ).


# image file: d2ee03930c-f2.tif
# Fig. 2 XAFS analysis for Fe-SNC800, Co-SNC800 and Co/Fe-SNC800 together with Fe foil, Co foil, iron phthalocyanine (FePc) and cobalt phthalocyanine (CoPc) as references. (a) Fe and (b) Co k-edge NEXAFS spectra; (c) Fe and (d) Co k-edge EXAFS analysis in R space; (e) C and (f) N k-edge NEXAFS spectra; and (g) Fe and (h) Co k-edge fitting spectra of (i) the fitting catalytic site.	
# As shown in Fig. 2e, the two k-edge features of C located at âˆ¼286.3 and 288.7 eV can be assigned to the Câ€“N bonds and Feâ€“C/Coâ€“C bonds, respectively.41â€“43 Compared to Fe-SNC800, the metalâ€“C bonds in Co/Fe-SNC800 almost disappeared, indicating the absence of the Feâ€“C or Coâ€“C coordination in Co/Fe-SNC800. As shown in Fig. 2f, the k-edge XANES spectra of N show three peaks in the lower energy region, corresponding to the pyrrolic N (âˆ¼399.10 eV), Feâ€“N or Coâ€“N coordination (âˆ¼400.2 eV) and graphitic N (âˆ¼401.8 eV), respectively.41,44 The metalâ€“N (Feâ€“N and Coâ€“N) coordination component of Co/Fe-SNC800 is 2-fold higher than that of Fe-SNC800 or Co-SNC800, owing to the Coâ€“Fe dual-metal coordination with the adjacent N atoms. Two sharp peaks at 2472.5 and 2482.83 eV and a broad peak at 2497.9 eV were observed in the k-edge XANES spectra of S, which can be assigned to the transition of S 1s to the S-C state, S5+ and S6+, respectively (Fig. S12, ESIâ€ ).45â€“47 Particularly, the absence of a pre-edge peak related to the Feâ€“S bonds in the energy region of âˆ¼2469â€“2471 eV indicates that there is no direct bonding between the Fe and S atoms in the catalysts.48,49 As shown in Fig. 2g and h, the R-space fitting results of Fe and Co k-edge reveal that the Fe and Co atoms in Co/Fe-SNC800 are atomically dispersed, whereby the configuration of the dual-metal site is dominated by FeN3-CoN2S coordination (Fig. 2i), respectively.

# The OER activity of the catalysts was evaluated in O2-saturated 1.0 M KOH at a scan rate of 2 mV sâˆ’1. As shown in Fig. 3a, Co/Fe-SNC800 exhibited an onset potential (Eonset) of 1.42 V (vs. RHE, for all potential values) at a current density of 0.1 mA cmâˆ’2. Moreover, the overpotential of Co/Fe-SNC800 for the OER at a current density of 10 mA cmâˆ’2 was found to be as low as 240 mV, which is superior to that of Fe-SNC800 and Co-SNC800, even lower than that of the commercial IrO2 catalyst. The electrocatalytic activity Co/Fe-SNC800 substantially surpassed that of most of the recently reported dual-atom catalysts (Table S3, ESIâ€ ). The OER performances of the Co/Fe dual-metal-site catalysts were tuned by varying the Co/Fe precursor ratio. As shown in Fig. S13 (ESIâ€ ), it can be found that too high or too low ratios of Co/Fe precursor had a negative impact on the OER activities. The highest mass activity (MA) value of Co/Fe-SNC800 with the optimized Co/Fe ratio (1/1) was determined to be 3.51 mA cmâˆ’2 gâˆ’1 (Fig. 3b). The effect of S-doping was evaluated by comparing the OER activity of Co/Fe-NC800 and Co/Fe-SNC800. As shown in Fig. S14a (ESIâ€ ), Co/Fe-NC800 showed lower OER activity than that of Co/Fe-SNC800, which is attributed to the inappropriate electronic structure of the active site in the absence of S atom. Additionally, the unfavorable morphology (Fig. S14b, ESIâ€ ) is also responsible for the low OER activity of Co/Fe-NC800. Tafel analysis was performed to assess the catalytic kinetics of the rate-determining step. The Tafel slope values are in the range of 40â€“60 mV decâˆ’1, indicating high surface reactivity.50Fig. 3c shows that the calculated Tafel slope of Co/Fe-SNC800 (47.92 mV decâˆ’1) is close to that of the commercial IrO2 (41.58 mV decâˆ’1), but lower than that of Fe-SNC800 and Co-SNC800, implying its fast OER kinetics. According to eqn (S1) and (S2) (ESIâ€ ), the electrochemical surface area (ECSA) was calculated through the electrochemical double-layer capacitance (Cdl) from the CVs in potential range of 1.2â€“1.3 V at different scan rates (Fig. S15, ESIâ€ ). As shown in Fig. 3d, Co/Fe-SNC800 shows the highest Cdl value of 20.8 mF cmâˆ’2. Interestingly, Co-SNC800 also shows a high Cdl value of 18.0 mF cmâˆ’2, which is similar to that of Co/Fe-SNC800 but much higher than that of Fe-SNC800, suggesting that the Co sites may play the key role in the adsorption and activation of the reactants. As shown in Table S4 (ESIâ€ ), the ECSA value of Co/Fe-SNC800 is 0.0645 gcatâˆ’1, which is higher than that of Fe-SNC800 (0.0149 gcatâˆ’1) and Co-SNC800 (0.0513 gcatâˆ’1). Subsequently, the metal active site density (MASD) was calculated using eqn (S3) (ESIâ€ ) to evaluate the intrinsic activity of the metal sites and confirm the improved catalytic activity of the Co/Fe dual-metal sites. The MASD value of Co/Fe-SNC800 (1.15 Ã— 1023 molmetalâˆ’1) is approximately 2-fold that of Fe-SNC800 (4.05 Ã— 1022 molmetalâˆ’1) and Co-SNC800 (5.02 Ã— 1022 molmetalâˆ’1) (Table S4, ESIâ€ ). The outstanding electrocatalytic activity of Co/Fe-SNC800 is also confirmed by its high turnover frequency (146 sâˆ’1 @ 10 mA cmâˆ’2, denoted as TOF, as described in SI), which is 2-fold that of Co-SNC800 (63 sâˆ’1@10 mA cmâˆ’2), as shown in Fig. 3e. Moreover, Co/Fe-SNC800 also shows an outstanding TOF at high overpotentials.


# image file: d2ee03930c-f3.tif
# Fig. 3 Electrochemical characterization of Fe-SNC800, Co-SNC800, Co/Fe-SNC800 and reference catalysts in 1 M KOH. (a) LSV curves at 1600 rpm. (b) Mass activity normalized by mass of Fe and Co (ICP-OES) in the catalysts. (c) Tafel plots. (d) Plots of (Jaâ€“Jc)/2 vs. scan rate v. (e) TOF curves derived from LSV curves. (f) Production of H2O2 and the transfer electron number. (g) Plots of long-term stability. (h) Nyquist plots and (i) Bode plots.	
# The catalytic OER pathway was further investigated using the rotating ring disk electrode (RRDE) approach. The ring electrode potential was fixed at 1.5 V to monitor the generation of H2O2 generated the 2eâˆ’ pathway, with the working mechanism illustrated in Fig. S17a (ESIâ€ ). As shown in Fig. S17b and c (ESIâ€ ), Co/Fe-SNC800 displays a higher Idisk and lower Iring than that of Fe-SNC800 and Co-SNC800, indicating the lower generation of H2O2 on Co/Fe-SNC800, respectively. Co/Fe-SNC800 demonstrates an electron transfer number (n) value of >3.5 with <25% H2O2 generation at potentials higher than 1.47 V (@10 mA cmâˆ’2), verifying the dominant 4eâˆ’ OER pathway on Co/Fe-SNC800 (Fig. 3f). Therefore, the Co/Fe dual-atom-site speeds up the conversion of H2O2 or eliminates the 2e OER pathway, facilitating the selective production of O2. Meanwhile, the RRDE was also used to detect the O2 at the Pt ring electrode by setting it at the ORR potential (0.4 V), with the working mechanism illustrated in Fig. S18a (ESIâ€ ).51,52 The ORR current on the ring electrode increased fast upon the onset of the OER at the disk electrode and kept increasing with an increase in the disk potential, confirming the efficient production of O2 (Fig. S18b, ESIâ€ ). The Co/Fe-SNC800 catalyst showed a higher Iring, and hence more efficient O2 generation than that of Fe-SNC800 and Co-SNC800, indicating its outstanding 4eâˆ’ OER activity. Chronopotentiometry was performed to evaluate the long-term stability of the catalysts. Co/Fe-SNC800 with the highest OER activity showed an almost negligible decrease in activity by maintaining the stable OER current at 20 mA cmâˆ’2 for 26 h (Fig. 3g). As shown in Fig. 3h, the Nyquist plot of Co/Fe-SNC800 shows a smaller quasi-semicircle than that of Co-SNC800, implying the lower electron transfer resistance in Co/Fe-SNC800, and hence its superior conductivity.

# As shown in Fig. 3i, the Bode plots of the three catalysts, which indicate the frequencyâ€“responsive reaction processes, show distinct electrochemical reaction processes on the different catalytic surfaces under 1.6 V. Based on the above-mentioned findings, the Bode plots under varying potentials were recorded to identify the electrochemical processes during the OER catalytic reaction. As shown in Fig. 3i and 4aâ€“c, three types of peaks are observed in the Bode plots under various applied potentials, which can be ascribed to three different surface reaction processes,53,54 as follows: (1) the adsorption capacitance in the high-frequency range in the vicinity of 104 Hz, originating from the most rapid relaxation process due to the adsorption of oxygen species (such as OHâ€“); (2) the faradaic process at the intermediate-frequency range of 102 Hz, which is associated with the charge transfer impedance response; and (3) the double-layer capacitance in the low-frequency range below 10 Hz, which is controlled by the diffusion process. As shown in Fig. 4a, Fe-SNC800 exhibits an obviously high-frequency relaxation process, indicating the strong adsorption of oxygen species. In the electrochemical process, the Fe sites can be easily oxidized into FesiteOOH in alkaline media viaeqn (1), which may block the catalytic sites and hinder the subsequent catalytic process.55 The Bode plots of Co-SNC800 and Co/Fe-SNC800 show their sensitive dependence on the potential. As shown in Fig. 4b, Co-SNC800 exhibits a conversion from diffusionâ€“faradaic mixed process at low potentials of 1.15â€“1.30 V to faradaic process at intermediate potentials of 1.40â€“1.60 V. The relatively weak OHâˆ’ adsorption on the Co site in Co-SNC800 compared with the Fe site in Fe-SNC800 results in straightforward desorption from the Co site (eqn (S10), ESIâ€ ), making it difficult to form CositeOOH active species (eqn (S11), ESIâ€ ). Co/Fe-SNC800 displays an interfacial conversion from a pure diffusion process to faradaic process, as shown in Fig. 4c. The co-catalytic processes occur on both the Fe and Co sites in Co/Fe-SNC800, i.e., the excessive OHâˆ’ species are strongly adsorbed on the Fesite to form FesiteOOH (eqn (1)), followed by reaction with the adjacent Cosite, further promoting the formation of highly active species CositeOOH (eqn (2)). The formation of FesiteOOH can lead to the transformation of the electronic structure around the Co/Fe dual-metal site, which makes the adjacent Co atoms unstable with high free energy. Thus, the Co atom tends to capture the ligand groups from the adjacent FesiteOOH at low potential under alkaline OER conditions to ensure its own stability by achieving the conversion from FesiteOOH to CositeOOH.56â€“60 Thus, a reasonable speculation for the OER catalytic mechanism on Co/Fe-SNC800 is proposed as the following eqn (1)â€“(3). As shown in Fig. S19 (ESIâ€ ), the co-catalysis between Fe and Co not only releases the excessive adsorbed OHâˆ’ on the Fe sites but also accelerates the formation of CositeOOH active species, which can improve the OER activity through the synergistic effect.

 
# Fesite + 3OH âˆ’ 3eâˆ’ = FesiteOOH + H2O	(1)
 
# FesiteOOH + Cosite = CositeOOH + Fesite	(2)
 
# CositeOOH + OHâˆ’ âˆ’ eâˆ’ = Cosite + O2 + H2O	(3)

# image file: d2ee03930c-f4.tif
# Fig. 4 OER catalytic mechanism on Fe-SNC800, Co-SNC800 and Co/Fe-SNC800. Bode plots of (a) Fe-SNC800, (b) Co-SNC800 and (c) Co/Fe-SNC800 under various applied potentials. (d) EPR signals for OOH* without electric field. (e) CV curves in N2-satruated 0.1 M KCl with 0.05 M K3[Fe(CN)6]/K4[Fe(CN)6], (f) at a scan rate of 50 mV sâˆ’1 and (g) at various scan rates in the range of 10 to 100 mV sâˆ’1. (g) Plot of Eaâ€“Ecvs. scan rate for Co/Fe-SNC800. (h) Color HAADF-STEM images (the four pictures were captured in less than 2.0 s) and (i) corresponding intensity profiles along the arrow in Region 1.	
# Electron paramagnetic resonance (EPR) was conducted as a semi-quantification method to assess the formation of OOH* active species as the key intermediate species in the 4eâˆ’ pathway of OER. Fig. 4d shows the EPR signal of the OOH* radicals in the absence of an electric field. The characteristic peak intensity of the OOH* radical reflects the relative concentration of the free radicals using the same test conditions. The highest OOH* radical concentration of Co/Fe-SNC800 indicates its outstanding activity in generating OOH* active species. The concentration of OOH* increased with an increase in the potential (Fig. S20, ESIâ€ ). Furthermore, the adsorption and desorption of O2 were probed by O2-temperature-programmed desorption (O2-TPD). As shown in Fig. S21 (ESIâ€ ), the peak in the range of 400â€“550 Â°C can be attributed to the desorption of the chemically adsorbed O. The area of the desorbed O normalized by the MASD, which is known as the O affinity coefficient, was employed to assess the oxygen desorption capacity per metal site (Table S4, ESIâ€ ). The Fe sites show a higher O affinity coefficient than that of the Co sites. The Co/Fe dual-sites in Co/Fe-SNC800 show a moderate chemisorption O coefficient, which accelerates the reaction rate of the OER. The current response of the [Fe(CN)6]3âˆ’/[Fe(CN)6]4âˆ’ redox pair was used as a descriptor to further assess the structureâ€“function relationship of the catalysts in the OER. As shown in Fig. 4e, the pronounced redox peaks observed in the range of 0.8â€“0.9 V can be ascribed to the [Fe(CN)6]3âˆ’/[Fe(CN)6]4âˆ’ redox pair. The highest current response of Co/Fe-SNC800 suggests the fastest electron transfer (I = Q/t, where I represents the current, Q is the total charge, and t is the reaction time). The lowest potential difference (Eaâ€“Ec) and current difference on Co/Fe-SNC800 indicate that fast electron transfer and approximately reversible process occur (Fig. S22, ESIâ€ ). The Eaâ€“Ec value, derived from the CV curves at various scan rates (Fig. 4f and Fig. S23a and b, ESIâ€ ), was plotted against the scan rate, as shown in Fig. 4g and Fig. S23c and d (ESIâ€ ). Different from the theoretical relationship that the Eaâ€“Ec value is independent of the scan rate, the calculated Eaâ€“Ec value is positively proportional to the scan rate, which can be attributed to the electron transfer between the metal atoms in the Co/Fe-SNC800 catalyst. The extrapolated Eaâ€“Ec value at 2 mV sâˆ’1 was found to be 56.78 mV vs. the ideal reversible electrode at 56.52 mV, suggesting that the electron transfer at a scan rate of 2 mV sâˆ’1 on Co/Fe-SNC800 is almost ideally reversible owing to the small charge transfer resistance. Additionally, the new redox peak observed in the range of 1.05âˆ¼1.25 V for Co-SNC800 can be attributed to the Co2+/Co3+ redox pair. However, the peaks of the Co2+/Co3+ redox pair disappear in Co/Fe-SNC800, possibly due to the electronic coupling between Fe and Co. The doping of Fe in Co-based materials decreases the surface Co2+/Co3+ ratio and the Co2+/Co3+ redox behavior, resulting in the disappearance of the Co2+/Co3+ redox peaks. In addition, the [Fe(CN)6]3âˆ’ in the electrolyte can chelate with the incompletely coordinated Co atom in the FeN3-CoN2S site on the surface of Co/Fe-SNC800, which can also contribute to the disappearance of the Co2+/Co3+ redox peaks.61â€“69

# The previous results strongly indicated the electronic coupling between the Co and Fe sites from various aspects, which was further revealed in the microscopy investigations. Thus, the time-continuum HAADF-STEM strategy with a series of images was employed in a time sequence to look into the structural evolution of the metal sites. The continuously captured images in 2.0 seconds on Region 2 (Fig. 4h) reveal a variation in the distance between the two adjacent metal atoms, indicating that there is no chemical bonding correlation between two adjacent metal sites. Further analysis of the intensity profile suggests that the two adjacent metal atoms vibrate elastically within a certain range (Fig. 4i and Fig. S24, ESIâ€ ), which presents further evidence of the existence of dual-atom sites (Coâ€“Fe dual-single-atom) with weak interactions, rather than the accidental combination of two independent single-atom sites. These dual-atom sites verified the electronic coupling effect between Fe and Co. The adjacent metal atoms can self-organize their configuration through electronic coupling, and thus their electronic structure is based on that of the adsorption species, which can effectively optimize the adsorption and desorption of the atoms or molecules and the active intermediates.70â€“72

# Subsequently, first-principle density functional theory calculation was employed to understand the mechanism of the remarkable OER performances of Co/Fe-SNC. For the bonding orbitals and anti-bonding orbitals near the Fermi level (EF), it can be noticed that Co/Fe-SNC exhibits strong orbital coupling, especially near the Co and Fe SA sites (Fig. 5a). The Coâ€“Fe SA sites form highly electroactive regions with electron-rich features. The orbital coupling involves the hybridizations of dz2, dx2âˆ’y2, and dyz of the Co and Fe sites to support the highly efficient electron transfer. In comparison, the coupling between the SA sites is absent in Co-SNC and Fe-SNC, leading to weakened site-to-site electron transfer in their lattice (Fig. 5b and c). Moreover, Fe-SNC displays stronger anti-bonding orbitals than bonding orbitals, which further reduces its overall electroactivity. The detailed electronic structures of the catalysts were displayed through the projected partial density of states (PDOSs). In the case of Co/Fe-SNC, the Co-3d orbitals exhibit a sharp peak close to the Fermi level at EV of âˆ’0.33 eV (EV = 0 eV), indicating its high electroactivity for the OER with strong electron depletion capability (Fig. 5d). The Fe-3d orbital shows egâ€“t2g splitting, which pins the Co-3d orbitals at the highly active position to guarantee a remarkable OER performance. Meanwhile, the strong overlapping among the N-2p, C-2p, and S-3p orbitals acts as the electron supply center to achieve efficient site-to-site electron transfer in Co/Fe-SNC. In the case of Co-SNC, it was observed that different Co sites display distinct electronic structures (Fig. 5e). The Co sites coordinated with S are beneficial for high electroactivity, while the coordination with N sites leads to a downshift in the 3d orbitals, resulting in decreased electroactivity. In comparison, both types of Fe-3d orbitals exhibit evident egâ€“t2g splitting near EF, revealing an electron transfer barrier for the OER process (Fig. 5f). The p orbitals of the SNC structure show very little change among the different electrocatalysts. To further evaluate the electroactivity, the d-band center was compared, where Co/Fe-SNC possesses superior electroactivity because it possesses the highest d-band center at EV of âˆ’0.86 eV (Fig. 5g). The d-band center of Fe/SNC and Co/SNC is EV = âˆ’0.88 eV and EV = âˆ’1.01 eV, respectively. Interestingly, we observed that the integration of Fe and Co SA sites in Co/Fe -SNC induced an increase in their d-band center, which can be attributed to the synergistic effect induced by the dâ€“d orbital coupling. The 3d orbitals of Fe protect the Co sites during the OER to maintain the high electroactivity towards the reactants and intermediates. The PDOS of the key adsorbates during the OER are shown in Fig. 5h and i. Co/Fe-SNC showed a slightly stronger linear correlation than that of Co-SNC for the intermediate conversion process from H2O to O2, which corroborates the more efficient electron transfer process with experimentally discovered lower overpotential for Co/Fe-SNC. Moreover, the main deviation in the linear correlation of PDOS evolution occurs at O* and OOH*, which indicates that the conversion of O* to OOH* is the potential rate-determining step (RDS) of the OER.


# image file: d2ee03930c-f5.tif
# Fig. 5 3D contour plot of bonding and anti-bonding orbitals near the Fermi level in (a) Co/Fe-SNC, (b) Co-SNC, and (c) Fe-SNC. Blue isosurface = bonding orbitals and green isosurface = anti-bonding orbitals. PDOS of (d) Co/Fe-SNC, (e) Co-SNC, and (f) Fe-SNC. Pink balls= Co, orange balls = Fe, blue balls = N, yellow balls = S, and grey balls = C. (g) Comparison of d-band center for Co/Fe-SNC, Co-SNC and Fe-SNC. PDOSs of key adsorbates of the OER on (h) Co/Fe-SNC and (i) Co-SNC. Reaction energy change of Co/Fe-SNC and Co-SNC under (j) U = 0 V and (k) U = 1.23 V.	
# From an energetic perspective, we demonstrated a comparison of the binding energy and energy cost of OOH* transfer in Fig. S25 (ESIâ€ ). The adsorption of OH* and O* is more energetically favored on the Fe sites, which is consistent with the experimental characterization (Fig. S25a, ESIâ€ ). The transfer of OOH* from the Fe site to the Co sites is overall spontaneous with a negative energy change of âˆ’0.81 eV (Fig. S25b, ESIâ€ ). Moreover, the transient state of OOH* transfer only shows an energy cost of 0.11 eV, which guarantees the fast transfer of OOH* to promote the OER performance. These results support the reaction mechanisms characterized by experiments. Finally, the reaction energy change of the OER for both Co/Fe-SNC and Co-SNC were further compared. Under U = 0 V, for both electrocatalysts, the OER process exhibited an overall uphill trend for the energy change (Fig. 5j). The largest energy barrier of the OER is the conversion from [2OHâˆ’ + H2O + O*] to [OHâˆ’ + H2O + OOH*] as the RDS, which is consistent with electronic structure results. Co/Fe-SNC has an energy barrier of 1.46 eV for the RDS, which is much smaller than that of Co-SNC, confirming the improved electroactivity of Co/Fe-SNC. After applying the equilibrium potential of 1.23 V, the overpotential of the OER was estimated to be 0.23 eV and 0.38 eV for Co/Fe-SNC and Co-SNC, respectively, agreeing well with the experimental results (Fig. 5k). Meanwhile, the reaction energy change of the comparison samples Fe/Co-NC and Fe/Co-SC was also revealed (Fig. S26, ESIâ€ ). Notably, both catalysts show higher energy costs than the Co/Fe SNC catalyst, revealing their synergistic contributions to the electroactivity. Co/Fe-NC shows the RDS at the conversion from OH* to O*, while Co/Fe-SC demonstrates the RDS at the initial adsorption of OH* (Fig. S26a, ESIâ€ ). Under the equilibrium potential, the overpotential of Co/Fe-NC and Co/Fe-SC is 0.43 V and 0.37 V, respectively (Fig. S26b, ESIâ€ ). These results further revealed that the introduction of both S and N atoms is important to achieve superior OER performances.

# Conclusions
# In summary, a Co/Fe-SNC dual-metal catalyst was designed and synthesized in this work via the reverse micellar assembly strategy followed by pyrolysis. The optimized catalyst Co/Fe-SNC800 showed high electrocatalytic activity towards the OER with an overpotential of 240 mV at 10 mA cmâˆ’2, as well as long-term stability with a negligible decrease in activity at 20 mA cmâˆ’2 for 26 h. The mechanism for the catalytic enhancement was systematically investigated through experiments, whereby the Cosite was found to initiate the activation of the reactants into intermediate species. The electron-coupling between Fe and Co not only could reduce the adsorption of oxygen species on the Fe sites but also accelerate the formation of CositeOOH active species. Fesite served as the auxiliary site to expedite the conversion of the intermediate species of CositeOOH into the final product. DFT calculation revealed the optimized electronic structure induced by the dâ€“d orbital coupling between the Fe and Co sites, which increased the electroactivity towards the conversion of intermediate species and sustained a robust valence state of active sites for the OER process. The synergistic effect of the Fe and Co SA sites led to a significantly reduced overpotential for the OER on Co/Fe-SNC. This work presents critical insight into the role of the Fe sites in the Co/Fe dual-metal catalyst for the OER, as well as a new proof-of-concept methodology for the precise control of the synergistic effect of dual-metal sites.
# """

# main_material = "Coâ€“Fe dual-atomic catalyst"
```

```python
# Define prompts for questions
extraction_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an advanced language model specializing in the extraction and analysis of electrocatalyst properties for material science. Your task is to meticulously examine the experimental section of the provided text and extract quantitative information on material properties of the {main_material}. Focus on key metrics such as numerical values, units, and experimental conditions, and disregard extraneous descriptive or contextual information that does not contribute to property characterization."),
    ("system", "Based on your expert knowledge and the content provided, answer the following questions regarding material properties:"),
    ("system", "0. What are the material composition, i.e., the percentage of each element within the material? Please extract the chemical composition in the form of chemical symbols and percentages. For example:If the material is a compound, provide it as follows: Chemical A: 70, Chemical B: 30. If it's an alloy, provide the atomic ratio. Ensure the symbols are correctly formatted, e.g., Fe: 80, O: 20"),
    ("system", "1. What are the active sites, coordination atoms, coordination numbers, bond lengths, and valence states? Make sure you double check the element in the chemical formula of the {main_material} such that you dont omit any element in your response. This will be the most important information to extract."),
    ("system", "2. What is the morphology of the material?"),
    ("system", "3. What are the atomic states, including the substrate material, elemental composition, and ID/IG ratio?"),
    ("system", "4. Provide information about the DOS (density of states), including d-band center, Fermi level, electron filling, and charge transfer."),
    ("system", "5. What is the onset potential and the type of reaction (e.g., HER) and the electrolyte information? If there are multiple reactions, please list all of them."),
    ("system", "6. What is the maximum Faradaic efficiency, and under what overpotential and current density conditions?"),
    ("system", "7. What is the yield rate of the material?"),
    ("system", "8. What is the total yield over the entire test duration?"),
    ("system", "9. Provide details of the stability test, including time and current density."),
    ("system", "10. What is the selectivity of the main product, including product name, potential and current density conditions?"),
    ("system", "11. What is the selectivity of the side product, including product name, potential and current density conditions?"),
    ("system", "12. What is the rate-determining step, and under what overpotential condition?"),
    ("system", "13. Describe the charge transfer around the active site, including direction and values."),
    ("system", "14. Important: Do not extract any information that is not related to the main material {main_material} reported in the paper."),
    ("system", "15. If you cannot find the information from the text, please report it as NA. Do not make up any information."),
    ("human", "The experimental section of the paper is provided below: {text}. Please answer the above questions using the provided information.")
])

def extract_data_from_paper(abstract, prompt=extraction_prompt, main_material=None):
    print(f"Extracting data from paper with abstract: {abstract[:100]}...")  # Debugging log
    llm = ChatOpenAI(temperature=0, model="o1-preview")
    
    # Combine prompt and model to create a runnable chain
    runnable = prompt | llm.with_structured_output(schema=Paper)
    
    # Invoke the runnable chain with the context needed for prompt placeholders
    response = runnable.invoke({"text": abstract, "main_material": main_material})
    
    print(f"Extraction response: {response}")  # Debugging log
    return response
```

```python
class ValidationResponse(BaseModel):
    accept: bool = Field(default=False, description="Indicates if the extracted output is acceptable based on a defined accuracy threshold.")
    materials: Optional[Paper] = Field(default=None, description="If the output is not acceptable, this field contains the corrected version of the extracted data.")
def validate_extraction_output(output, text):
    print("Validating extraction output...")  # Debugging log
    validation_prompt = ChatPromptTemplate.from_messages([
        ("system", "You are an advanced language model tasked with validating the accuracy and completeness of extracted data from an electrocatalyst material science paper. Please review the output provided and indicate if all required fields have been correctly filled and if the values seem consistent with the domain-specific knowledge."),
        ("human", "The extracted data is provided below: {output}. The original experimental section text is: {text}. Please review and provide feedback on the accuracy and completeness.")
    ])
    llm = ChatOpenAI(temperature=0, model="gpt-4o")
    formatted_prompt = validation_prompt.format_prompt(output=output, text=text)
    validation_runnable = validation_prompt | llm.with_structured_output(schema=ValidationResponse)
    validation_response = validation_runnable.invoke({"output": output, "text": text})
    print(f"Validation response: {validation_response}")  # Debugging log
    return validation_response
```
